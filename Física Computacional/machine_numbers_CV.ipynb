## 2.1.1 Tipos de errores en el análisis numérico

* **Error experimental:** precisión del experimento, fundamental en el análisis estadístico.  


* **Error de redondeo:** los ordenadores sólo permiten utilizar números dentro de un subconjunto de los números reales: los números máquina $A\subset \mathbb R$.

* **Errores de truncación:** cualquier operación matemática en un ordenador se calcula solo aproximadamente, lo que conlleva un error de truncación.

## 2.1.3 Definiendo números reales en un ordenador

* Un ordenador almacena un número real (float) como el producto de:
    
    $$x=(-1)^s\,\times\,a\,\times\,2^{b-E}$$

    donde:    

    * $s$ es el signo: 0/1 para positivo/negativo
    * a es la mantisa: contiene el número de cifras significativas (precisión).   
    
    
* En general todos los números son ${\bf\text{normalizados}}$:  
  aquellos cuya parte entera (sin contar con el exponente) se puede representar como uno, es decir: 

    $$1.a_{n}\cdots a_{0}$$ 
     con $n$ es el número de bits asignados a la mantisa.  
    
    El primer valor es siempre 1, no es necesario guardarlo $\Rightarrow$ hidden bit. 
     
    **¿Entre que valores se encuentra entonces la mantisa?**   
          
    * b es el exponente: contiene el orden de magnitud. 
    * E es el sesgo del exponente: que tiene como valor la mitad del mayor valor de b.    

   
* Los ordenadores suelen trabajar con un formato de 32, 64 o 128 bits o 4, 8 y 16 bytes  (precisión simple, doble o cuádruple).


* Para cada uno de estos formatos los bits asignados (en orden) a la mantisa $(a)$ y al exponente $(b)$ son:


| Formato | Signo | Exponente | Mantisa |
|  :---:    |  :---:  |    :---:    |    :----:  |
| Simple | s | $b_0$ $\cdots$ $b_7$ |$a_0$ $\cdots$ $a_{22}$ |
| Doble | s | $b_0$ $\cdots$ $b_{10}$    |   $a_0$ $\cdots$ $a_{51}$      |
| Cuádruple | s | $b_0$ $\cdots$ $b_{14}$    |   $a_0$ $\cdots$ $a_{111}$      |

# 2.2  El error de redondeo

* Los reales toman un valor continuo de puntos entre $\pm$ int(max_normal),
* El hecho de que la mantisa tenga un número de bits finito implica que para un float los números máquina  $\it A$ sólo toman un número discreto de valores. 
* Cualquier número real tiene que ser aproximado por un número máquina por medio de una aplicación tal que:

$$\forall\;x\;\in\;\mathbb R, \quad x \rightarrow \text{rd}(x),\;\text{con}\;\text{rd(x)}\;\in\;\it A,$$

* cumpliendo la propiedad:

$$\vert x-\text{rd}(x)\vert \le \vert x-a\vert,\; \forall a \in \it A,$$

* Cuando $x$ se encuentra exactamente entre dos números máquinas se redondea al número que tiene como último bit un cero (round-to-nearest-even). 


* Esto implica que, llamando $t$ al número de bits destinados a la mantisa para una precisión dada ($t=52$ en precisión doble), cualquier número real con más de $t$ cifras significativas tendrá que ser aproximado por el número máquina más cercano, que tendrá sólo $t$ cifras significativas. Esta implica que el error de redondeo estará siempre acotado por

    $$\vert\text{rd}(x)-x\vert\le2^{-(t+1)}\times 2^{b-E}=\epsilon_M\times 2^{b-E},$$

    donde $\epsilon_M=2^{-(t+1)}$ es el llamado error de la máquina. 


* De la misma forma, el error de redondeo siempre será menor que la mitad de la distancia entre dos números máquinas. Definiendo como $a\times 2^{b-E}$ y $a'\times 2^{b-E}$ a los números máquinas normalizados mas cercanos a un número real $x$, satisfaciendo 

    $$a\times 2^{b-E}\le x\le a'\times 2^{b-E}$$

    el error de redondeo esta acotado por

    $$\vert\text{rd}(x)-x\vert\le\frac{\vert a'-a\vert}{2}\times 2^{b-E}=2^{-(t+1)}\times 2^{b-E}=\epsilon_M\times 2^{b-E}.$$


- Por tanto, para cada una de las precisiones disponibles, el error de la máquina vendrá dado por:

| Formato | Signo | Exponente | Mantisa | Precisión $\epsilon_M$|
|  :---:    |  :---:  |    :---:    |    :----:  |                       :---:               |
| Simple | s | $b_0$ $\cdots$ $b_7$ |$a_0$ $\cdots$ $a_{22}$ | $2^{-24}$ = $5.96$ $\times$ $10^{-8}$|
| Doble | s | $b_0$ $\cdots$ $b_{10}$    |   $a_0$ $\cdots$ $a_{51}$      |$2^{-53}$ = $1.11$ $\times$ $10^{-16}$|
| Cuádruple | s | $b_0$ $\cdots$ $b_{14}$    |   $a_0$ $\cdots$ $a_{111}$      |$2^{-113}$ = $9.63$ $\times$ $10^{-35}$|

* Lo que implica que para una precisión doble, cualquier número real con más de 16 cifras significativa sufriera un error de redondeo.


* Finalmente, el error relativo asociado a la operación de redondeo está acotado por:

    $$\frac{\vert\text{rd}(x)-x\vert}{x}\le\frac{\vert a'-a\vert\times 2^{b-E}}{2\,a'\times 2^{b-E}}\le \frac{\epsilon_M}{a'}\le \epsilon_M. $$

* Por tanto la operación de redondeo se define como:

$$\text{rd}(x)=x(1+\epsilon),\quad \vert\epsilon\vert\le\epsilon_M.$$

### 2.2.1 Errores en operaciones aritméticas:

#### Suma/resta de floats:

* Consideremos la suma de dos floats:

$$y=x_1+x_2,$$

* En primer lugar $x_1$ y $x_2$ tienen que ser aproximados por números máquina:

\begin{align}
x_1\rightarrow\text{rd}(x_1)=x_1(1+\epsilon_1),\\
x_2\rightarrow\text{rd}(x_2)=x_2(1+\epsilon_2),
\end{align}

* El resultado de la suma, también tiene que ser redondeado, por tanto:

$$y\rightarrow \text{rd}(y)=\left(x_1 (1 + \epsilon_1 ) + x_2 (1 + \epsilon_2 )\right)(1 + \epsilon_y)\sim (x_1+x_2)(1+\epsilon_y)+x_1\,\epsilon_1+x_2\,\epsilon_2\le y\,(1+2\,\epsilon_M).$$

* La suma/resta de N floats representa repetir $N-1$ veces la suma de dos pares de números, y cada uno de estás operaciones conllevará un error de redondeo por lo que uno encuentra que para $y=\sum_{i=1}^N x_i$

    \begin{align}
    y\rightarrow \text{rd}(y)=&\,\left(x_1 (1 + \epsilon_1 ) + x_2 (1 + \epsilon_2 )+\cdots +x_N(1+\epsilon_N)\right)(1 +\epsilon_{y_1}+\epsilon_{y_2}+\cdots+\epsilon_{y_{N-1}})\\
    \sim &\,(x_1+x_2+\cdots+x_N)(1+\epsilon_{y_1}+\cdots+\epsilon_{y_{N-1}})+x_1\,\epsilon_1+x_2\,\epsilon_2+\cdots +x_N\,\epsilon_N,
    \end{align}

   por lo que de forma general:

    $$\text{rd}\left(y=\sum_{i=1}^N x_i\right)\le y\,(1+N\,\epsilon_M).$$

* En particular, el error relativo de redondeo para la suma/resta es:

$$\frac{\text{rd}(y)-y}{y}=\frac{x_1}{y}\epsilon_1+\frac{x_2}{y}\epsilon_2+\epsilon_y,$$

* En una resta, sí $y=x_1-x_2$ es mucho más pequeño que $x_1$ y/o $x_2$ (es decir si $x_1\sim x_2$), el error relativo puede incrementarse dramáticamente.

* Por tanto nunca se debe usar un if para comprobar la igualdad de dos o más floats.

#### Multiplicación/división de floats:

* De la misma forma para la multiplicación y la división de dos floats se tiene:

\begin{align}
y=x_1\times x_2\quad\rightarrow &\quad\text{rd}(y)\sim y\,(1+\epsilon_1+\epsilon_2+\epsilon_y)\le y\,(1+3\epsilon_M),\\
z=\frac{x_1}{x_2}\quad\rightarrow &\quad \text{rd}(z)\sim y\,(1+\epsilon_1-\epsilon_2+\epsilon_z)\le z\,(1+3\epsilon_M)
\end{align}

* por lo que el error relativo
\begin{align}
\frac{\text{rd}(y)-y}{y}\sim\,\epsilon_1+\epsilon_2+\epsilon_y\le 3\epsilon_M,\\
\frac{\text{rd}(z)-z}{z}\sim\,\epsilon_1-\epsilon_2+\epsilon_z \le 3\epsilon_M,
\end{align}

 sólo se suma de forma lineal. 

### 2.2.2 Propagación de errores

  * Si queremos obtener una estimación del error de redondeo más allá de dar una cota,  
   es razonable asumir que error de redondeo esta uniformemente distribuido con desviación típica:
   
   $$\sigma(x) =\epsilon_M x,$$
 
 * Al igual que en las medidas experimentales, esto nos permite propagar errores en las diferentes operaciones matemáticas:
 
 * Para la suma/resta de dos floats: $x=x_1+x_2$:
 
 $$\sigma(x)=\sqrt{\sigma(x_1)^2+\sigma(x_2)^2}=\,\epsilon_M \sqrt{x_1^2+x_2^2},$$
 
 
 * De la misma forma, para el producto/cociente de dos floats: $x=x_1 * x_2$ o $x=x_1/x_2$
 
 $$\sigma(x)=\sqrt 2\epsilon_M x.$$
 
 * En general para la suma de N floats: $x=\sum_{i=1}^N{x_i}$
  
    $$\sigma(x)^2=\sum_{i=1}^N{\sigma(x_i)^2}=\,\epsilon_M^2 \sum_{i=1}^N{x_i^2}=\,\epsilon_M^2\, N\, \overline{x^2},\quad\Rightarrow\quad\sigma(x)=\epsilon_M\,\sqrt N\,\sqrt{\overline{x^2}},$$   

    $\hookrightarrow$ el error de redondeo escala con la raíz del número de floats.
   
* El error relativo por tanto va como:

    $$\frac{\sigma(x)}{x}=\frac{\epsilon_M\,\sqrt N}{N}\,\frac{\sqrt{\overline{x^2}}}{\overline x}=\frac{\epsilon_M}{\sqrt N}\,\frac{\sqrt{\overline{x^2}}}{\overline x},$$

     $\hookrightarrow$ aunque el error relativo decrece al añadir más números hay dos problemas:
   
   
- Si estamos sumando/restando números con tamaños muy diferentes los más pequeños pueden perderse.


- Si estamos restando números casi iguales podemos obtener cancelaciones catastróficas.


- Fundamental para el cálculo de límites y derivadas numéricamente.

### Ejemplo 2.14: Multiplicación de una matriz de $N \times N$.

1. Definir un programa que calcule el producto de dos matrices $N \times N$.

3. Usar el comando de jupyter %timeit para comprobarlo (%%timeit permite evaluar el tiempo necesario para ejecutar una celda completa).

```python
from numpy import empty

def matrix_product(A,B):
    N=len(A)
    C=[]
    for i in range(N):
        for j in range(N):
            for k in range(N):
                C[i,j]+=A[i,k]*B[k,j]
    return C
    
```
