# 9.1 Números aleatorios 

- El primer paso para estudiar fenómenos aleatorios es generar números aleatorios.


- De forma precisa, nunca vamos a ser capaces de generar números completamente aleatorios en un ordenador, sino pseudo-aleatorios: 
<br>

    * Tienen una apariencia aleatoria aunque se generan mediante una fórmula determinista. 
    <br>
        
    * Son los llamados generadores de números aleatorios .

## 9.1.1 Generadores de números aleatorios

- Consideremos la siguiente ecuación:

    $$f(x)=(a x +c)\,\text{mod}\, m,$$

    siendo $a$, $c$ y $m$ constantes enteras, y $x$ una variable entera.


- Vamos a estudiar que números generamos con esa función cuando, partiendo de un valor inicial $x=1$, la iteramos de forma sucesiva. 

```python
from matplotlib.pyplot import plot,show

N=100
a=1664525
c=1013904223
m=4294967296

x=1
xp=[]

for i in range(N):
    x=(a*x+c)%m
    xp.append(x)
    
show(plot(xp,"o"))
```

- Aparentemente, la posición de los números es aleatoria. 


- Es el llamado generador de números aleatorios congruencial lineal. 


- Hay varias cosas que es importante tener en cuenta:
<br>
    
    1.  Realmente no estamos generando números aleatorios, el programa es completamente determinista. Fijados $a$, $c$, $m$ y $x$ podemos predecir los números que se van a obtener. Si ejecutamos el programa dos veces, volveremos a obtener exactamente la misma secuencia.  
    <br>
    
    2. Todos los números generados se encuentran en el intervalo  $[0,m − 1]$. Si queremos un número $r$, tal que $0\le r<1$, sólo es necesario dividir los números de la congruencia por $m$.
    <br>
    
    3. La elección de $a$, $c$, $m$ y $x$ es importante. Para nuestros valores, los números escogidos funcionan razonablemente bien.
    <br>
    
    4.  Para los mismos valores de a, c y m  se puede obtener una secuencia completamente distinta variando el valor de $x$ inicial. Es la llamada semilla.
<br>

- Teniendo en cuenta todos estos aspectos, el generador de números aleatorios congruencial lineal permite obtener secuencias de números pseudoaletorios que son adecuados para muchos problemas en física.  


- Sin embargo, no es un buen generador de números aleatorios. Genera correlaciones entre los diferentes números de la secuencia, algo que no ocurre para números verdaderamente aleatorios (ejercicio de la baraja del Tema 1).


- Por suerte hay muchos otros generadores de números aleatorios que permiten obtener series con pocas o nulas correlaciones. 


- El generador que se usa de forma habitual en física hoy en día es el llamado generador Mersenne Twister, desarrollado en 1997, y basado en una ley de recurrencia lineal para matrices definidas sobre un cuerpo binario en dos dimensiones y cuyo periodo es un número primo de Mersenne (números primos de la forma $2^w−1$  para un $w$  primo dado).


- En python este generador se encuentra dentro del paquete random. 
<br/>

    - La función random(), genera un número aleatorio real uniformemente distribuido en el intervalo $(0,1)$.
    - La función randrange(n), genera un entero  entre 0 y $n-1$. 
    - La función randrange(m,n) genera un entero entre $m$ y $n-1$. 
    - La función randrange(m,n,k) genera un entero en el rango que va de $m$ a $n-1$, con pasos de tamaño $k$.
    
    
- Es importante notar que estos programas van a generar números diferentes cada vez que los llamemos.     

```python
from random import randrange

print(randrange(100))
print(randrange(100))
print(randrange(100))
```

- Por tanto, si queremos generar un número aleatorio y usarlo más de una vez, será necesario guardarlo como una variable. 

### Ejemplo 9.1: simulando un dado

1. Escribir un programa que imprima dos números aleatorios entre 1 y 6 para simular el lanzamiento de dos dados. 


2. Ejecutar el programa un millón de veces y calcular la probabilidad con las que se obtiene un doble seis. 

```python
from random import randrange

N=1000000
it=0

for i in range(N):

    d1=randrange(1,7)
    d2=randrange(1,7)

    if d1 == 6 and d2 == 6:
        it+=1

print("Probabilidad de obtener un 6 doble =", it/N)
print("Probabilidad exact:", 1/36)
```

## 9.1.2 Semillas de números aleatorios

- Tal y como hemos visto en el generador de números aleatorios congruencial lineal, la semilla de un generador especifica el comienzo de la secuencia de números aleatorios, y por tanto la secuencia entera. 


- En el caso del generador congruencial lineal, la semilla es el primer número de la secuencia, pero en muchos otros casos esto no es así; puede ser que la semilla se use indirectamente dentro de una formula matemática para especificar el primer número de la serie. 


- Esto es precisamente lo que ocurre pare el generador Mersenne twister; la semilla sólo se usa indirectamente para obtener el primer número de la secuencia. 


- De esa forma, fijando la semilla, podemos fijar exactamente la secuencia entera de números aleatorios que vamos a obtener. 


- En python, la semilla se fija usando la función seed(n), donde $n$ es el valor de la semilla que queremos fijar. 

```python
from random import randrange,seed

seed(17)

for i in range(5):
    print(randrange(10))
```

* Si volvemos a ejecutar la misma línea de código. 

```python
seed(17)

for i in range(5):
    print(randrange(10))
```

* Obtenemos exactamente la misma secuencia. 


* Sin embargo, si no hubiéramos especificado la semilla

```python
for i in range(5):
    print(randrange(10))
```

* La secuencia habría sido diferente. 


- La cuestión es preguntarse por qué podría ser útil fijar la semilla y garantizar que vamos a obtener la misma serie de números pseudoaleatorios. 


- La respuesta es que es útil para escribir el código de un programa por primera vez. 


- En ocasiones encontraremos que nuestro programa falla (da un error) para una determinada secuencia inicial, pero que al volverlo a ejecutar parezca funcionar bien.


- Esto puede pasar porque el programa no hace exactamente lo mismo para secuencias de números diferentes, lo que hace aún más complicado encontrar los problemas y solucionarlos.


- Al fijar la semilla evitamos este problema, y garantizamos que las secuencias serán la misma al ejecutar el código. 


- Una vez el código funcione correctamente se puede eliminar la semilla para garantizar una aleatoriedad mayor del programa.

## 9.1.3 Probabilidad y números aleatorios

-  Como ya hemos comentado hay muchos fenómenos en física que suceden de forma aleatoria con una probabilidad $p$ determinada.


- Este tipo de comportamiento puede ser descrito fácilmente usando números aleatorios.


- Sólo es necesario generar un número aleatorio entre 0 y 1 y asociarlo una probabilidad, de forma que por sí sólo determine si un determinado fenómeno aleatorio está teniendo lugar. 


- Por ejemplo, la forma de describir el resultado de tirar una moneda al aire podría venir dada por el código:

```python
from random import random

for i in range(10):
    if random()<0.5:
        print("C")
    else:
        print("X")
```

### Ejemplo 9.2: la desintegración de un isótopo radiactivo


El isótopo radiactivo $^{208}\text{Tl}$ (Talio 208) se desintegra en $^{208}\text{Pb}$ (plomo 208), un elemento estable, con una vida media de 3053 minutos. La vida media de un elemento radiactivo es el tiempo medio necesario para que dada una muestra inicial de átomo de $^{208}\text{Tl}$  $N(0)$  esta se reduzca a la mitad. 


Esto nos permite obtener la ecuación de la desintegración radiactiva

$$N(t)=N(0)\,2^{-t/\tau},$$

siendo el número de elementos radiactivos por unidad de tiempo y $\tau$ la vida media del elemento. 


1. Usa la formula de la desintegración radiactiva para definir la probabilidad con la que un determinado átomo de $^{208}\text{Tl}$ se desintegrará en un tiempo $t$.


2. Supongamos que partimos de una muestra con 1000 átomos de talio. Describir como evoluciona el número de átomos de $^{208}\text{Tl}$ y $^{208}\text{Pb}$ en un periodo de $10^{3}$ segundos.

1. Después de un periddo $t$ el número de átomos de Talio que permanecen en la muestra será

$$\frac{N(t)}{N(0)}=2^{-t/\tau},$$

$\quad$ por tanto la fracción de átomos que se han desintegrado, es decir, la probabilidad de que un átomo se desintegre después de un tiempo $t$ será

$$p(t)=1-2^{-t/\tau}.$$


```python
# 2. Vamos a describir la evolución de átomos en la muestra. 

from random import random
from numpy import arange
from matplotlib.pyplot import plot,show,xlabel,ylabel,legend

# Constantes

NTl=1000          # Número de átomos de talio
NPb=0             # Número de átomos de plomo
tau=3.053*60       # vida media del talio en segundos
tmax=1000          # tiempo total de la simulación
h=1                # tamaño de los pasos en el intervalo temporal 
p=1-2**(-h/tau)    # Probabilidad de decaimiento en un paso
 

# Listas de puntos para la gráfica

tp=arange(0,tmax,h)
Tlp=[]
Pbp=[]

# bucle principal

for t in tp:
    Tlp.append(NTl)
    Pbp.append(NPb)
    
    # Calculamos el número de átomos que decaen
    
    decay = 0    
    for i in range(NTl):
        if random()<p:
            decay+=1
    NTl-=decay
    NPb+=decay

# Hacemos el gráfico
plot(tp,Tlp,label="Átomos de Talio")
plot(tp,Pbp,label="Átomos de Plomo")
xlabel("Tiempo")
ylabel("Número de átomos")
legend()
show()
```

## 9.1.4 Números aleatorios con distribución no-uniforme

- Todos los números aleatorios que hemos generados hasta ahora están distribuidos uniformemente. Es decir todos los valores posibles que pueden tomar esos números tienen la misma probabilidad de aparecer. 


- Sin embargo, en física hay muchas procesos que suceden con probabilidad no-uniforme. Un ejemplo es la situación anterior del decaimiento de un número $N$ de átomos de un isótopos radiactivos.


- Ya hemos visto que la probabilidad de que un átomo decaiga en un intervalo de tiempo $\text{d}t$ viene dado por

    $$p(\text{d}t)=1-2^{-\text{d}t/\tau}=1-\exp\left(-\frac{\text{d}t}{\tau}\log2\right)\simeq \frac{\log2}{\tau}\text{d}t,$$

    donde hemos despreciado términos ${\cal O}\left(\text{d}t^2\right)$.


- Ahora podemos por tanto preguntarnos cuál es la probabilidad de que un isótopo radiactivo se desintegre en el intervalo $t+\text{d}t$.


- Para que un átomo se desintegre en el intervalo $(t,t+\text{d}t)$ tiene en primer lugar que haber sobrevivido hasta el tiempo $t$, lo cual ocurre con la probabilidad $2^{-t/\tau}$, y luego desintegrase en el intervalo, durante un tiempo $\text{d}t$. Por tanto, la probabilidad de que se desintegre entre los tiempos $t$ y $t+\text{d}t$ es

    $$p(t)\text{d}t=2^{-t/\tau}\frac{\log2}{\tau}\text{d}t,$$

    que es el ejemplo de una distribución no uniforme pues los tiempos de desintegración $t$ se distribuyen con una proporción $2^{-t/\tau}$, es decir, es más probable que se desintegre en tiempo anteriores que en tiempos posteriores. 


- Por tanto, una forma mucho más eficiente de calcular la desintegración de $N$ isótopos radiactivos habría sido generar $N$ números aleatorios generados con la distribución derivada arriba para representar el tiempo en el que cada uno de los átomos se desintegrarán. 


- Con ello, la curva que describe la evolución del número de átomos del isótopo se podría obtener contando simplemente el número de átomos que han decaído en un tiempo dado. 
  
  
- El problema consiste en obtener un generador de números aleatorios con una distribución dada.  Para hacerlo usaremos el llamado método de la transformación. 

### 9.1.4.1 El método de la transformación

- Supongamos que tenemos un generador de números aleatorios en coma flotante asociado a una densidad de probabilidad $q(z)$, que implica que la probabilidad de generar un número en el intervalo $z+\text{d}z$ es $q(z)\text{d}z$.


- Supongamos a continuación que tenemos una función $x=x(z)$. Por tanto, para un número aleatorio $z$ generado con la densidad de probabilidad $q(z)$, $x(z)$ será otro número aleatorio que en general estará asociado a otra distribución de probabilidad $p(x)$. 


- Nuestro objetivo no es otro que encontrar la función $x(z)$ que nos permite generar un número $x$ con la probabilidad que queremos $p(x)$.


- Por definición, la probabilidad de generar un número $x$ entre $x+\text{d}x$ será igual que la probabilidad de generar un número $z$  en el intervalo $z+\text{d}z$

    $$p(x)\text{d}x=q(z)\text{d}z,$$

    donde ya hemos especificado que $x=x(z)$.


- En general, tendremos un generador de números aleatorios uniforme en el intervalo $[0,1)$, tal como el que obtenemos usando la función random. 


- En ese caso, la probabilidad será $q(z)=1$ en el intervalo $(0,1)$ y 0 fuera de él. 


- Por tanto la probabilidad de obtener un número menor de $z$ será

    $$\int_{-\infty}^{x(z)}p(x')\,\text{d}x^\prime=\int_{0}^{z}\text{d}z=z.$$


- Si pudiéramos hacer la integral de la izquierda, obtendríamos por tanto el valor que ha de cumplir nuestra función $x(z)$.


- No esta claro que siempre podamos resolver la integral, y que si lo hacemos, entonces podamos resolver la ecuación resultante; sin embargo, si lo hacemos, habremos obtenido exactamente la función que queremos.


- A modo de ejemplo, supongamos que tenemos la distribución exponencial 

    $$p(x)=\mu\text{e}^{-\mu x},\quad \text{con}\quad x\in[0,\infty)$$

    donde el factor $\mu$ es necesario para cumplir que 

    $$\int_{0}^{\infty} p(x)\text{dx}=\mu\text{e}^{-\mu x}=-\text{e}^{-\mu x}\,\big\vert_{0}^{\infty}=1,$$
    
    efectivamente corresponde a una probabilidad. 


- Aplicando la formula anterior, la probabilidad de obtener un número menor de $z$ será

    $$\int_{0}^{x(z)}\mu\text{e}^{-\mu x^\prime}\,\text{d}x^\prime=1-\text{e}^{-\mu x(z)}=z,$$

    que implica que

    $$x(z)=-\frac{1}{\mu}\log(1-z).$$


- Por tanto para obtener nuestros números aleatorios con distribución exponencial todo lo que tenemos que hacer es generar números aleatorios con distribución uniforme en el intervalo $[0,1)$ e introducirlos en la ecuación anterior.

## 9.1.5 Números aleatorios gaussianos

- La distribución gaussiana o normal juega un papel fundamental tanto en física como en matemáticas gracias al teorema central del límite que establece que, bajo ciertas condiciones generales, la suma de $N$ variables aleatorias tiene aproximadamente una distribución normal. 


- El teorema del límite central implica por tanto que la distribución normal puede usarse como aproximación de muchas otras distribuciones.


- La distribución gaussiana esta definida por la densidad de probabilidad

    $$p(x)=\displaystyle\frac{1}{\sqrt{2\pi}\,\sigma}\text{e}^{-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\,\right)^2},\quad\text{con}\quad x\in(-\infty,\infty),$$

    donde $\mu$ es la media y $\sigma$ la desviación típica o anchura de la distribución.


- Siguiendo el método de la transformación tenemos por tanto la igualdad

    $$\frac{1}{\sqrt{2\pi}\,\sigma}\int_{-\infty}^{x(z)}\text{e}^{-\frac{1}{2}\left(\frac{x'-\mu}{\sigma}\,\right)^2}\,\text{d}x'=z.$$


- Sin embargo la integral no se conoce de forma analítica, por lo que el método de la transformación falla en este caso. 


- Por suerte, hay un truco que podemos emplear para generar números aleatorios distribuidos normalmente a partir de números aleatorios con distribución uniforme.  


- Imaginemos dos números aleatorios $x$ e $y$ extraídos de una distribución gaussiana con la misma media $\mu$ y desviación típica $\sigma$.  
  La probabilidad de que esos números se encuentren en el elemento de superficie $(x+\text{d}x)(y+\text{d}y)$ del plano $xy$ es
  
    $$p(x)\text{d}x\times p(y)\text{d}y=\frac{1}{2\pi\,\sigma^2}\exp\left({-\frac{\left(x-\mu\;\right)^2\;+\,\left(y-\mu\;\right)^2}{2\sigma^2}}\right),$$

    que también se puede expresar en coordenadas polares. Haciendo

    $$x'=x-\mu,\;y'=y-\mu,\;\quad x'^2+y'^2=r^2\quad\text{y}\quad \text{d}x\text{d}y=r\,\text{d}r\,\text{d}\theta,$$

    podemos expresar nuestra probabilidad como

    $$p(x)\text{d}x\times p(y)\text{d}y=p(r,\theta)\text{d}r\,\text{d}\theta=\frac{1}{2\pi\,\sigma^2}\exp{\left(-\frac{r^2}{2\sigma^2}\right)}\;r\,\text{d}r\,\text{d}\theta=\frac{r}{\sigma^2}\exp{\left(-\frac{r^2}{2\sigma^2}\right)}\;\text{d}r\,\frac{\text{d}\theta}{2\pi}=p(r)\text{d}r\times p(\theta)\text{d}\theta,$$

    donde las densidades de probabilidad $p(r)$ y $p(\theta)$ estás normalizadas correctamente. 


- Es decir, sí generamos números aleatorios $r$ y $\theta$ con las distribuciones de probabilidad $p(r)$ y $p(\theta)$ respectivamente, al hacer el cambio a coordenadas cartesianas tendremos dos variables aleatorias con una densidad de probabilidad normal.


- Generar un número con distribución $p(\theta)$ es trivial, no es más que una distribución uniforme en el intervalo $[0,2\pi)$.


- Para la variable $r$ podemos usar de nuevo el método de la transformación

    $$p(r)\text{d}r=\frac{1}{\sigma^2}\int_0^{r(z)}\exp\left(-\frac{r'^2}{2\sigma^2}\right)r'\text{d}r'=1-\exp\left(-\frac{r(z)^2}{2\sigma^2}\right)=z,$$

    que implica que

    $$r(z)=\sqrt{-2\sigma^2\log(1-z)}.$$


- Una vez obtenidos dos variables aleatorias $r$ y $\theta$ podemos convertirlas en cartesianas usando que

$$x=\mu+r\cos\theta, \quad y=\mu+r\sin\theta.$$

### Ejemplo 9.6: dispersión de Coulomb o Rutherford

A principios del S. XX, Ernest Rutherford y sus colaboradores mostraron que cuando una partícula $\alpha$, un núcleo de helio formado por dos protones y dos neutrones, es lanzada cerca de un átomo, es dispersada principalmente debido a la repulsión eléctrica asociada a dos cargas con el mismo signo. 


Esto permitió descartar el modelo atómico de Thomson, los átomos deben tener una estructura con carga positiva diferenciada, que a partir de ese momento se llamaron núcleos atómicos.


A través del experimento se comprobó que el ángulo dispersado por la partícula alpha venía dado por la expresión

$$\tan\frac{1}{2}\theta=\frac{Ze^2}{2\pi\epsilon_0 E\,b}$$

```python
from IPython.display import Image,display

display(Image("rutherford.png",width=500))
```

siendo $E$ la energía cinética de la partícula $\alpha$, $Z$ el número atómico del material sobre el que la partícula $\alpha$ es lanzada, $e$ la carga eléctrica, y $b$ el parámetro de impacto, la distancia perpendicular entre la trayectoria de la partícula $\alpha$ y el átomo. 


Consideremos un haz de un miĺlón partículas $\alpha$ con energía cinética $7.7$ eV, que tienen un perfil Gaussiano en el plano perpendicular a la dirección de incidencia, tanto en el eje $x$ como el eje $y$ con una desviación estándar $\sigma = a_0/100$ donde $a_0$ es el radio de Bohr, que se lanza sobre un átomo de oro. 


Calcular la fracción de partículas que rebotan hacia atrás, es decir cuyo ángulo de dispersión sea mayor de $90^\circ$.

- En primer lugar si $\theta >\pi/2$ las partículas rebotarán hacia atrás. Como la tangente incrementa con el ángulo en el intervalo $[0,\pi/2]$  tenemos que la partículas rebotaran si

$$\tan\frac{\theta}{2}=\frac{Ze^2}{2\pi\epsilon_0 Eb}>\tan\frac{\pi}{4}=1,$$

$\quad$ o lo que es lo mismo, si el parámetro de impacto cumple

$$b<\frac{Ze^2}{2\pi\epsilon_0 E}.$$

- Simulemos a continuación el proceso de dispersión

```python
from numpy import sqrt,log,cos,sin,pi
from random import random

# Constantes

Z=79                  # número atómico del oro
e=1.602e-19           # carga eléctrica
E=7.7e6*e             # energía cinética de haz en eV
epsilon0=8.854e-12    # permitividad del vació
a0=5.292e-11          # radio de bohr
sigma=a0/100          # desviación típica de la distribución gaussiana incidente.
N=1000000             # número de partículas del haz


# Función para generar dos números aleatorios con una distribución Gaussiana

def gaussian():
    r=sqrt(-2*sigma**2*log(1-random()))
    theta=2*pi*random()
    x=r*cos(theta)
    y=r*sin(theta)
    
    return x,y

# Programa principal

it = 0

for i in range(N):
    
    x,y=gaussian()
    b=sqrt(x*x + y*y)
    
    if b<Z*e*e/(2*pi*epsilon0*E):
        it+=1

print("Habrá", it,"partículas reflejadas de", N)
print("Que suponen el", it/N*100, "%")
```

## 9.2.1 Cálculo de integrales por el método de Monte Carlo

- Para entender en más profundidad como funciona el método, consideremos la integral 

    $$I=\int_{0}^2\sin^2\left[\frac{1}{x(2-x)}\right]\text{d}x,$$

    cuyo cálculo no implica más que calcular el área bajo la curva

```python
from numpy import sin,linspace
from matplotlib.pyplot import plot,show,fill_between

x=linspace(0.0001,1.9999,1000000)
y=sin(1/x/(2-x))**2
plot(x,y)
fill_between(x,y,step="pre",alpha=0.4)
plot(x,y,"0.")
show()
```

- El comportamiento de la función es perfectamente suave fuera de los extremos, pero, cerca de ellos, la función empieza a oscilar infinitesimalmente rápido, lo que hace que calcular su valor estudiando los métodos que hemos visto hasta ahora sea bastante complicado.


- Sin embargo, la integral está perfectamente definida y acotada, pues está contenida en un rectángulo de área 2. 


- Por tanto  una forma de replantear el problema es asociar la integral con la probabilidad de que un punto generado aleatoriamente dentro del rectángulo de área A que contiene a la gráfica, caiga en la zona en azul, es decir

    $$p=\frac{I}{A},$$

    donde $A$ es el área del rectángulo total e $I$ es nuestra integral.


- Por tanto, para estimar la integral sólo tenemos que calcular dicha probabilidad lo cual podemos hacer de la siguiente forma

    1.  Generamos un número $N$ de puntos dentro del área $A$ que contiene nuestra curva.
    
    2. Contamos el número de puntos que caen en la región azul, digamos que son por ejemplo $k$.
    
    3. Estimamos la probabilidad a la fracción de puntos que cae debajo de la curva, es decir con $k/N$.
    
    4. Estimamos la integral a partir de la expresión
    
    $$I=\frac{k}{N}A.$$
    
    
- Este es el llamado de hit or miss (acertar o fallar).

### Ejemplo 9.7: cálculo de una integral usando el método de Monte Carlo

Calcular la integral 

$$I=\int_{0}^2\sin^2\left[\frac{1}{x(2-x)}\right]\text{d}x,$$

usando números aleatorios. 

```python
from numpy import sin
from random import random

def f(x):
    return sin(1/x/(2-x))**2

N=1000000
it=0
A=2

for i in range(N):
    x=2*random()
    y=random()    
    if y<f(x):
        it+=1
I=A*it/N
print(I)    
```

Un valor correcto hasta dos decimales.

## 9.2.2  Error de las integrales obtenidas por el método de Monte Carlo

- El problema de los métodos de Monte Carlo para resolver integrales es que no son muy precisos. 

- Su error, por supuesto, está asociado a nuestra limitación para estimar una probabilidad usando números aleatorios. 

- Vamos a calcular cual es la precisión que podemos esperar para dicha estimación.

    1. La probabilidad de que un sólo número aleatorio se encuentre por debajo de la curva, es tal y como hemos dicho
    
        $$p=\frac{I}{A}.$$
    
        por lo que 1-p será la probabilidad de que se encuentre sobre la curva.     

    2. La probabilidad de que sólo $k$ de entre $N$ números aleatorios se encuentren bajo la curva será por tanto 
    
        $$p^k(1-p)^{N-k}$$
    
       Sin embargo eso puede ocurrir en ${N\choose k}$ formas diferentes, es decir, hay ${N\choose k}$  posibilidades de escoger $k$ números de entre $N$. 
       
       Por tanto, la probabilidad de tener exactamente k puntos bajo la curva es
       
        $$P(k)={N\choose k}p^k(1-p)^{N-k},$$

        que no es más que la probabilidad de la distribución binomial.

  3. Por consiguiente, para calcular el error de nuestra probabilidad, sólo tenemos que calcular la variaza de la distribución binomial.
      
       - En primer lugar, el valor esperado después de haber lanzado N números aleatorios será:
          
        \begin{align}
         \langle k\rangle=&\sum_{k=0}^N kP(k)=\sum_{k=1}^N{k{N\choose k}p^k(1-p)^{N-k}}=\sum_{k=1}^N{N{N-1\choose k-1}p^k(1-p)^{N-k}}\\
         \\
        =&Np\sum_{k=1}^N{{N-1\choose k-1}p^{k-1}(1-p)^{N-k}}=Np\sum_{k=0}^{N-1}{{N-1\choose k}p^{k}(1-p)^{N-1-k}}=Np.\end{align}

       - Mientras que el valor esperado al cuadrado vendrá dado por
       
        \begin{align}\langle k^2\rangle=&Np\sum_{k=1}^{N}{k{N-1\choose k-1}p^{k-1}(1-p)^{N-k}}=Np\sum_{k=0}^{N}{(k+1){N-1\choose k}p^{k}(1-p)^{N-1-k}}\\
        \\
        =&Np+Np\sum_{k=0}^{N}{k{N-1\choose k}p^{k}(1-p)^{N-1-k}}=Np+N(N-1)p^2.\end{align}

       - La varianza de la distribución binomial es por tanto
       
       $$\text{var}(k)=\langle k^2\rangle-\langle k\rangle^2=Np(1-p)=N\frac{I}{A}\left(1-\frac{I}{A}\right).$$

       - Por tanto teniendo en cuenta que estamos estimando nuestra integral como $I=k A/N$, el error de nuestra estimación es
       
       $$\sigma=\sqrt{\text{var}(k)}\frac{A}{N}=\frac{\sqrt{I(A-I)}}{\sqrt N},$$
       
       es decir, nuestro error escala como $N^{-1/2}$.
     
- Por tanto el error de nuestra integral estimada usando el método de MonteCarlo varia como $N^{-1/2}$.

- Si calculamos una integral por el método de MonteCarlo con 100 puntos, es de esperar una precisión del 10%; por supuesto el valor exacto dependerá de $I$ y $A$.


- Éste es un resultado mucho más modesto de lo que hemos obtenido con el método del trapecio, que escala con $N^{-2}$, con el de Simpson, $N^{-4}$, o con el de la cuadratura gaussiana, $N^{-N}$.


- Sin embargo los métodos de Monte Carlo funcionan incluso para la integrales de funciones patológicas con grandes variaciones, o, como veremos, para integrales multidimensionales. 

## 9.2.3 El método del valor medio

- El método de hit or miss, no es el mejor para estimar una integral usando números aleatorios.


- El método más común para hacerlo se llama el método del valor medio. 


- Sea una vez más la integral 

    $$I=\int_{a}^b f(x)\text{d}x.$$


- La media de la función en el intervalo viene dada por

    $$\langle f\rangle=\frac{1}{b-a}\int_{a}^b f(x)\,\text{d}x=\frac{I}{b-a},$$

    por lo que conociendo la media podemos obtener la integral

    $$I=(b-a)\langle f\rangle.$$


- Pero la media se puede estimar fácilmente eligiendo $N$ puntos aleatorios en el intervalo $[a,b]$ $x_i$ con $i=1,\cdots,N$, por lo que

    $$I=(b-a)\langle f\rangle\simeq\frac{(b-a)}{N}\sum_{i=1}^Nf(x_i),$$

    que es el llamado método del valor medio. 


- El error del método se puede estimar fácilmente teniendo en cuenta que 

    $$\text{var}(f)=\langle f^2\rangle-\langle f\rangle^2,$$

    y que la varianza de una suma es la suma de las varianza. 


- Por tanto nuestro error será

    $$\sigma(I)=\frac{(b-a)}{N}\sqrt{\sum_{i=1}^N\text{var}(f)}=(b-a)\frac{\sqrt{\text{var}(f)}}{\sqrt N},$$

    es decir, el error sigue yendo como $N^{-1/2}$.


- Sin embargo el factor de proporcionalidad es siempre menor o igual que el método hit or miss, que implica que el método del valor medio siempre va a ser mejor o igual. 

### 9.2.3.1 Demostración de que el método del valor medio es más preciso que el método hit or miss

- Para demostrar que este es el caso asumamos que nuestra función $f(x)>0$ en todo el dominio de integración y que área total de probabilidad, el área que engloba la función en su intervalo, es 

    $$A=(b-a)H,$$ 
  
  siendo $a$ y $b$ los puntos inicial y final del intervalo en el que se está integrando la función y $H$ la altura. 


- Definamos ahora la variable $s$, definida tal que para cada una de los $i$ "lanzamientos" de MonteCarlo, $s_i=1$ si es un hit y 0 si es un miss. 


- De esta forma, la estimación de la integral usando el método hit or miss, vendrá dada por 

    $$I\simeq \frac{k A}{N}=(b-a)H\langle s\rangle,$$

  siendo $\langle s\rangle=\frac{k}{N}=\frac{1}{N}\sum_{i=1}^{N}s_i.$
  
  
- De esta forma, el error de nuestra estimación usando el método hit or miss será

    $$\sigma(I)=\sqrt{\text{var}(k)}\frac{A}{N}=\sqrt{N\text{var}(s)}\frac{(b-a)H}{N}=\sqrt{\text{var}(s)}\frac{(b-a)H}{\sqrt N},$$
    
  donde una vez más, hemos tenido en cuenta que la varianza de una suma es la suma de las varianzas.   
  
  
- Comparando este resultado con el error correspondiente al método del valor medio tendremos que el segundo es más o igual de preciso si

    $$(b-a)\frac{\sqrt{\text{var}(f)}}{\sqrt N}\le\sqrt{\text{var}(s)}\frac{(b-a)H}{\sqrt N},$$
  
  o equivalentemente si
  
    $$\text{var}(f)\le\text{var}(s)H^2.$$
    
    
- Por otro lado, la varianza de la variable $s$, viene dada por

    $$\text{var}(s)=\langle s^2\rangle-\langle s\rangle^2,$$ 
    
    siendo $\langle s\rangle=\frac{k}{N}=p$, la probabilidad media de un acierto, y 
    
    $$\langle s^2\rangle=p\times 1^2+(1-p)\times 0^2=p,$$ 
    
    la media del cuadrado. Por tanto 

   $$\text{var}(s)=\langle s^2\rangle-\langle s\rangle^2=p(1-p).$$
   
   
- Por otro lado tenemos que 

    $$\langle f\rangle=\frac{I}{(b-a)}\;\Rightarrow\;\frac{\langle f\rangle}{H}=\frac{I}{A}=p,$$

  por lo que sustituyendo en $\text{var}(s)$ tenemos  
  
   $$H^2\text{var}(s)=H^2\,p(1-p)=\langle f\rangle\left(H-\langle f\rangle\right).$$


- Por tanto, el método del valor medio será más preciso que el método hit or miss si

    $$\text{var}(f)=\langle f^2\rangle-\langle f\rangle^2\le\langle f\rangle\left(H-\langle f\rangle\right),$$
    
  o lo que es lo mismo si  
  
   $$\langle f^2\rangle\le H\langle f\rangle,$$
      
  condición que se satisface trivialmente porque $f$ y $H$ cumplen $0<f<H$,  en el dominio de integración.   

### Ejemplo 9.8: cálculo de una integral usando el método del valor medio

Volver a evaluar la integral 

$$I=\int_{0}^2\sin^2\left[\frac{1}{x(2-x)}\right]\text{d}x,$$

y su error usando un millón de puntos aleatorios para

1. El método hit or miss. 
2. El método del valor medio. 

Comprobar que el error es menor en el segundo caso. 

```python
from numpy import sin,sqrt
from random import random

def f(x):
    return(sin(1/(x*(2-x))))**2

a=0
b=2
N=1000000


A=2
it=0

for i in range(N):
    x=2*random()
    y=random()
    
    if y<f(x):
        it+=1
        
I=A*it/N
sigma=sqrt(I*(A-I)/N)

print(I,sigma)
```

```python
# Usamos el método del valor medio

sum=sumsq=0

for i in range(N):
    
    fx=f(2*random())
    sum+=fx
    sumsq+=fx*fx

media=sum/N
var=sumsq/N-media*media
I=(b-a)*media
sigma=(b-a)*sqrt(var/N)

print(I,sigma)
```

## 9.2.4 Integrales multidimensionales

- Aparte de para integrales con patologías, los métodos de MonteCarlo son útiles en integrales multidimensionales. 


- Tal y como vimos en el Tema 4, la cuadratura gaussiana en más de una dimensión tiene el problema de que no hay una forma clara de elegir los puntos y los pesos. 


- Por otro lado, métodos más simples, como el del trapecio, exigen muestrear el integrando en cada una de las variables, lo que hace que se vuelvan lentos cuando el número de dimensiones aumentan. 


- Por ejemplo, usar el método del trapecio o de Simpson en tres dimensiones con 100 puntos en cada una de ellas implica evaluar el integrando en $100^3=10^6$, lo que para funciones complicadas puede ser lento. 


- Para cuatro dimensiones ya tendríamos $10^8$ evaluaciones lo que puede no ser una opción desde el punto de vista del tiempo computacional. 


- Los métodos de MonteCarlo, se pueden generalizar trivialmente a dimensiones arbitrarias, y su precisión sigue escalando con $N^{-1/2}$, es decir, siguen dando respuestas razonables usando un número reducido de evaluaciones.


- La generalización del método del valor medio para integrales multidimensionales con volumen $V$, viene dada por

    $$I\simeq \frac{V}{N}\sum_{i=1}^Nf(r_i),$$

  donde $r_i$ representa los puntos escogidos aleatoriamente del volumen $V$. 

### Ejemplo 9.9: volumen de una hiperesfera

Vamos a estimar el volumen de una esfera de radio unidad en 10 dimensiones usando el método de Monte Carlo. 

Para plantear el problema consideremos el ejemplo equivalente en dos dimensiones

```python
from IPython.display import Image,display
display(Image("circle.png",width=500))
```

El área del círculo viene por tanto dado por la integral

$$I=\iint_{-1}^1f(x,y)\,\text{d}x\,\text{d}y,$$

donde la función viene definida como

$$f(x,y)=\left\{\begin{array}{cc}1&\text{if}\;x^2+y^2\le1,\\0&\text{de otra forma},\end{array}\right.$$

por lo que la estimación de la integral  usando el método del valor medio será

$$I\simeq\frac{4}{N}\sum_{i=1}^Nf(x_i,y_i).$$

Generalizar este caso para calcular el volumen de la hiperesfera en diez dimensiones.

```python
from numpy import empty,dot
from random import random

d=10       # dimensiones de nuestra esfera
N=1000000  # número de puntos

# integrando

def f(r):
    if dot(r,r)<=1:
        return 1
    else:
        return 0
    
s=0      # suma de nuestra integral

for i in range(N):
    
    r=empty(d,float)
    for k in range(d):
        r[k]=2*random()-1
        
    s+=f(r)
    
print(2**d*s/N)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import gaussxw
N=10000

#Integrando
def f(x,y,z,w):
    return np.cos(x)*np.sin(y)*np.exp(z)*np.tanh(w)
#Valor medio
s=0
for i in range(N):
    x=random()*np.pi/2
    y=random()*np.pi
    z=random()
    w=random()
    s+=f(x,y,z,w)
V=np.pi**2/2
print('El metodo del valor medio da',V/N*s)
#Cuadratura Gaussiana
def gaussxwab(N,a,b):
    x,w = gaussxw.gaussxw(N)
    return 0.5*(b-a)*x+0.5*(b+a),0.5*(b-a)*w
x,w=gaussxwab(10,0,np.pi/2)
y,v=gaussxwab(10,0,np.pi)
z,u=gaussxwab(10,0,1)
p,t=gaussxwab(10,0,1)
I=0
for i in range(10):
    for j in range(10):
        for k in range(10):
            for l in range(10):
                I+=w[i]*v[j]*u[k]*t[l]*f(x[i],y[j],z[k],p[l])
print('El valor por cuadratura gaussiana es', I)
print('El valor real es', 2*(np.e-1)*np.log((np.e+np.exp(-1))/2))
print()
print('La cuadratura gaussiana resulta más precisa para este caso porque no se trata de una función problemática')
```

## 9.2.5 Muestreo por importancia (promedios con peso)

- Tal y como hemos visto los métodos de Monte Carlo son adecuados para calcular integrales de funciones con patológicas, con rápidas variaciones. 


- Sin embargo hay un tipo de funciones para las que la integración por Monte Carlo no funciona. Consideremos por ejemplo la integral

    $$I=\int_0^1\frac{x^{-1/2}}{\text{e}^x+1}\,\text{d}x,$$
    
  que aparece en los gases de Fermi.
  
  
- Aunque el integrando diverge para x=0, la integral tiene un valor finito. 


- Sin embargo, al aplicar el método del valor medio 

    $$I=\frac{(b-a)}{N}\sum_{i=1}^Nf(x_i)$$

  podemos encontrar problemas pues $f(x_i)\to\infty$ cuando $x_i\to 0$, por lo que podemos obtener resultados muy diferentes a la suma dependiendo de donde caigan nuestros puntos aleatorios.  


- Esto obviamente implica que la incertidumbre de nuestra estimación va a ser mucho más grande, pudiendo llegar incluso a infinito si estamos evaluando la función en $x=0$.


- Una forma de solucionar este problema es escoger nuestros puntos de forma no uniforme en el intervalo de integración, es decir, calcular un valor medio con un cierto peso para cada valor de la variable aleatoria.


- Es lo que llamamos el muestreo de importancia, que funciona como sigue. 



- Dada una función arbitraria $g(x)$ definimos su media ponderada con el peso $\omega$ sobre el intervalo $[a,b]$ como

    $$\langle g(x) \rangle_\omega=\frac{\int_a^b\omega(x)g(x)\,\text{d}x}{\int_a^b\omega(x)\,\text{d}x},$$

  donde $\omega(x)$ es cualquier función a elegir. 
  
  
- Volviendo de nuevo a nuestra integral

    $$I=\int_{a}^bf(x)\,\text{d}x.$$
  
  y, haciendo $g(x)=f(x)/\omega(x)$, tenemos que
  
  $$\left\langle \frac{f(x)}{\omega(x)} \right\rangle_\omega=\frac{\int_a^b\omega(x)f(x)/\omega(x)\,\text{d}x}{\int_a^b\omega(x)\,\text{d}x}=\frac{I}{\int_a^b\omega(x)\,\text{d}x},$$
  
  o lo que es lo mismo
  
  $$I=\left\langle \frac{f(x)}{\omega(x)} \right\rangle_\omega\,\int_a^b\omega(x)\,\text{d}x,$$
  
  que nos permite calcular el valor de nuestra integral a partir de cualquier media ponderada, en vez de a partir de una media uniforme. 

### 9.2.5.1 Calculando la media ponderada

- La cuestión ahora es calcular la media ponderada. 


- Para ello, definamos la densidad de probabilidad

  $$p(x)=\frac{\omega(x)}{\int_{a}^b\omega(x)\,\text{d}x},$$
  
  que es como la función peso $\omega(x)$ pero normalizada a 1. 


- Dados $N$ números $x_i$ en el intervalo [a,b] generados con esa distribución aleatoria, $p(x)\,\text{d}x$ será la probabilidad de que uno caiga en el intervalo $[x,x+\text{d}x]$, de forma que el promedio o valor esperado de números aleatorios que encontraremos en ese elemento diferencial será $N p(x)\,\text{d}x$.


- Usando este resultado, el promedio de evaluar una función $g(x)$ sobre los puntos $x_i$ generados con la distribución $p(x)$ en el intervalo $[a,b]$ será

    $$\frac{1}{N}\sum_{i=1}^N g(x_i)\simeq\frac{1}{N}\int_a^b\, N\,p(x)\,g(x)\,\text{d}x=\int_a^b\, p(x)\,g(x)\,\text{d}x=\frac{\int_a^b\omega(x)g(x)\,\text{d}x}{\int_a^b\omega(x)\,\text{d}x},$$
    
  donde estamos aproximando la suma por una integral ponderada, y por tanto, donde la aproximación será mejor cuanto mayor sea el número de elementos de la suma. 


- Con este resultado es trivial calcular la media ponderada pues

   $$\langle g(x) \rangle_\omega=\frac{\int_a^b\omega(x)g(x)\,\text{d}x}{\int_a^b\omega(x)\,\text{d}x}=\int_a^b\,p(x)g(x)\,\text{d}x\simeq\frac{1}{N}\sum_{i=1}^Ng(x_i),$$
   
   donde los puntos $x_i$ están generados con la distribución no uniforme $p(x)$. 


- De esta forma, nuestra integral $I$ se puede expresar como
  
  $$I\simeq \frac{1}{N}\sum_{i=1}^N\frac{f(x_i)}{\omega(x_i)}\int_{a}^b\omega(x)\,\text{d}x,$$
  
  que es la formula fundamental del muestreo por importancia. 
  
  
- Por supuesto, escogiendo $\omega(x)=1$, recuperamos la expresión para el método del valor medio.


- El muestreo por importancia nos permite calcular una estimación de la integral $I$ evaluando no la suma $\sum_i f(x_i)$ sino $\sum_i f(x_i)/\omega(x_i)$, donde $\omega(x)$ puede ser cualquier función a elegir. 


- Esta flexibilidad a la hora de elegir $\omega(x)$ es muy útil ya que podemos escogerla de forma que podamos evitar las posibles patologías de la función. 
  Por ejemplo, dada una singularidad, podemos escoger $\omega(x)$ para cancelarla, de forma que la suma $\sum_i f(x_i)/\omega(x_i)$ este perfectamente definida. 


- Finalmente el error de nuestra estimación se puede calcular de forma totalmente equivalente a la del método del valor medio, viniendo dado por

   $$\sigma(f_\omega)=\frac{\sqrt{\text{var}_\omega\left( f/\omega \right)}}{\sqrt N}\int_a^b\omega(x)\,\text{d}x,$$
   
  siendo una vez más $\text{var}_\omega(f/\omega)=\langle g^2\rangle_\omega -\langle g\rangle_\omega^2$.

### Ejemplo 9.10: Calculando integrales con el muestreo por importancia

Dada la integral

$$I=\int_0^1\frac{x^{-1/2}}{\text{e}^x+1}\,\text{d}x,$$

calcular su valor estimado usando el muestreo por importancia. 

- Tal y como hemos visto, la función a integrar $f(x)=x^{-1/2}/(\exp(x)+1)$ tiene  una divergencia en x=0, consecuencia del término $x^{-1/2}$.


- Para cancelar su efecto en la suma ponderada, escogemos por tanto $\omega(x)=x^{-1/2}$, de forma que

    $$\frac{f(x)}{\omega(x)}=\frac{1}{\text{e}^{x}+1}.$$


- Por tanto, para evaluar nuestra integral tenemos que generar números aleatorios distribuidos con la densidad de probabilidad

    $$p(x)=\frac{\omega(x)}{\int_0^1 \omega(x)}=\frac{x^{-1/2}}{\int_0^1 x^{-1/2}\,\text{d}x}=\frac{1}{2\sqrt x},$$
    
  que podemos obtener a partir de números uniformemente distribuidos usando el método de la transformación,
  
  $$\frac{1}{2}\int_0^x\,(x')^{-1/2}\,\text{d}x'=\sqrt x=z,\quad  x= z^2. $$


- Por tanto, nuestra integral será

  $$I\simeq \frac{1}{N}\sum_{i=1}^N\frac{f(x_i)}{\omega(x_i)}\int_{a}^b\omega(x)\,\text{d}x=\frac{2}{N}\sum_i^N\frac{1}{\exp{(z_i^2)}+1},$$
  
  donde tal y como hemos visto los puntos $z_i$ son números aleatorios generados en el intervalo $[0,1]$.

```python
from numpy import exp
from random import random

N=1000000
s=0.0

for i in range(N):
    s+=1/(exp(random()**2) + 1)

print("La estimación de la integral es: %0.2f" %(2*s/N))
```

### 9.2.5.2 El muestreo por importancia e integrales sobre rangos infinitos

- El muestreo por importancia también puede usarse para evaluar integrales definidas en dominios infinitos. 


- Este caso no se puede resolver con el método del valor medio, pues tendríamos que obtener números aleatorios distribuidos uniformemente en el intervalo $[0,\infty]$, lo cual no es posible, porque es una distribución obviamente no integrable. 


- Sin embargo, es algo que podemos hacer usando el muestreo por importancia, pues nos permite calcular la integral a partir de distribuciones no uniformes, que aun extendiéndose a infinito sean integrables.


- Es por ejemplo el caso de la distribución exponencial. Escogiendo $\omega(x)=\text{e}^{-x}$ tenemos

    $$I\simeq \frac{1}{N}\sum_{i=1}^N\text{e}^{x_i}f(x_i)\int_0^\infty e^{-x}\,\text{d}x=\frac{1}{N}\sum_{i=1}^N\text{e}^{x_i}f(x_i).$$


- Es decir podemos usar una muestra de puntos generada con una distribución no uniforme, sólo si lo compensamos  pesando más en la suma aquellos puntos que se encuentran en la región donde generamos menos. 

# 9.3 Simulaciones de Monte Carlo: cadenas de Markov y el algoritmo de Metrópolis

- Llamamos simulaciones de MonteCarlo a aquellos programas de ordenador que utilizan números aleatorios para simular el comportamiento aleatorio de variables físicas. 


- Ejemplo de tales simulaciones son la desintegración radiactiva o el scattering de Rutherford.


- Aunque las simulaciones de Monte Carlo se usan en muchas ramas de la física, juegan un papel central en la mecánica estadística, ya que la aleatoriedad es la parte esencial de este campo. 


- Por ello, en este sección nos vamos a centrar en estudiar las simulaciones de MonteCarlo en Física Estadística. 

## 9.3.1 El muestreo por importancia y la mecánica estadística

- Uno de los problemas esenciales en física estadística es el de calcular el valor esperado o medio de una cantidad de interés en un sistema físico en equilibrio térmico a temperatura $T$.


- Aunque no sabemos el estado exacto del sistema, sí sabemos que, a una temperatura $T$ dada, el sistema pasa por una sucesión de estados tal que la probabilidad de que en un momento dado el sistema se encuentre en el estado $E_i$ viene dada por

    $$P(E_i,\beta)=\frac{\text{e}^{-\beta\,E_i}}{Z(\beta)},\quad Z(\beta)=\sum_i\text{e}^{-\beta E_i},$$

  siendo $\beta=1/(k_B\,T)$ con $k_B$ la constante de Boltzmann y $Z(\beta)$ la función de partición de la colectividad canónica. 


- Dada la probabilidad, el valor esperado de la cantidad X que toma el valor $X_i$ en el estado $E_i$ viene dada por

    $$\langle X\rangle=\sum_i X_iP(E_i).$$


- Aunque en algunos casos podremos realizar la suma de forma analítica, la mayoría de las veces sólo se podrá hacer de forma numérica. 


- Sin embargo, en muchas ocasiones no podemos simplemente evaluar la suma porque el número de estados que intervienen es demasiado grande. 


- Por ejemplo, hay $6.023\times 10^{23}$ moléculas en un mol de un gas, por lo que, si incluso asumimos que cada molécula sólo tiene dos estados posibles (de hecho tienen decenas de ellos), el número total de estados sería $2^{10^{23}}$, que no se puede evaluar con ningún ordenador. 



- La idea por tanto es enfocar el problema de la misma forma que hemos planteado la integración por MonteCarlo. 


- El método del valor medio nos permite evaluar una integral sólo con sumar un conjunto de puntos escogidos aleatoriamente en el dominio de integración.


- La idea ahora es hacer lo mismo con la suma, escoger un número aleatorio de términos y sumarlos para obtener el resultado final. Por ejemplo dados $N$ estados escogidos aleatoriamente denotados con $k=1,\cdots,N$ tendremos

   $$\langle X\rangle=\frac{\sum_{k=1}^N X_kP(E_k)}{\sum_{k=1}^N P(E_k)},$$

  donde el denominador es esencial para normalizar la media ponderada, pues no estamos cogiendo todos los estados que garantizan que 
  
  $$\sum_{i} P(E_i)=1.$$


- Desafortunadamente la ecuación de arriba no funciona en nuestro caso. La probabilidad de Boltzmann esta exponencialmente suprimida para aquellos estados cuya energía $E_i\gg k_B T$, lo cual ocurre para la mayoría de estados. 


- Esto implica que la mayoría de los estados que escogeremos contribuirán muy poco a la suma que queremos obtener, o lo que es lo mismo, hay muy pocos estados que contribuyen de forma significativa a la suma. 


- La forma de resolver este problema es clara: usar una distribución de números aleatorios no uniforme;  es decir, el muestreo por importancia. 


- Hasta ahora hemos definido el muestreo por importancia como una forma de evitar patologías en funciones.    
  La consecuencia de ello es que uno debe escoger una muestra de números aleatorios con una distribución no uniforme. 


- En nuestro caso el objetivo es de hecho escoger una muestra no uniforme, de forma que podamos seleccionar correctamente los términos que debemos incluir en la suma para evaluar el promedio de nuestro observable. 


- Una vez más, dada cualquier función $g_i$ que depende del estado $i$, definimos la media ponderada como

  $$\langle g\rangle_\omega=\frac{\sum_i\omega_i g_i}{\sum_i \omega_i},$$
  
  siendo $\omega_i$ cualquier conjunto de pesos que nosotros elijamos. 


- Escogiendo $g_i=X_i P(E_i)/\omega_i$, es decir, escogiendo cada elemento de nuestra suma pero dividido por el peso $\omega_i$, tenemos

  $$\left\langle \frac{X_iP(E_i)}{\omega_i}\right\rangle_\omega=\frac{\sum_i\omega_i X_i P(E_i)/\omega_i}{\sum_i \omega_i}=\frac{\sum_i X_i P(E_i)}{\sum_i \omega_i}=\frac{\langle X\rangle}{\sum_i \omega_i},$$
  
  o lo que es lo mismo, el promedio de X vendrá dado por
  
  $$\langle X\rangle=\left\langle \frac{X_iP(E_i)}{\omega_i}\right\rangle_\omega \sum_k \omega_k.$$
  
  
- Una vez más, podemos evaluar la media ponderada escogiendo $N$ estados aleatorios con probabilidad

    $$p_i=\frac{\omega_i}{\sum_j \omega_j},$$
        
  de forma que nuestra estimación para la media ponderada sea
  
  $$\langle g\rangle_\omega\simeq\frac{1}{N}\sum_{k=1}^Ng_k,$$
  
  siendo $g_k$ el valor de g para el estado $E_k$.
  
  
  
- Aplicando este resultado para el valor esperado de $X$, obtenemos

  $$\langle X\rangle\simeq \frac{1}{N}\sum_{i=1}^N \frac{X_i P(E_i)}{\omega_i}\sum_k \omega_k,$$
  
  donde es fundamental darse cuenta que mientras que la primera suma sólo se realiza sobre los $N$ estados de nuestra muestra aleatoria, la segunda suma es sobre todos los estados posibles. Por tanto, si es posible, es mejor calcular esta segunda suma analíticamente.  
  
  
- La cuestión ahora es elegir los pesos $\omega_k$. 


- Debemos elegirlos de forma que la mayoría de los estados escogidos aleatoriamente caigan en la zona donde $P(E_i)$ es grande, y a la vez, de forma que la suma $\sum_k\omega_k$ se puede hacer de forma analítica. 


- Ambos requisitos los podemos satisfacer trivialmente si hacemos $\omega_i=P(E_i)$. 


- En ese caso tendremos que que $\sum_i\omega_i=\sum_iP(E_i)=1$ por lo que

  $$\langle X\rangle\simeq \frac{1}{N}\sum_{k=1}^N X_i.$$


- Es decir, si escogemos los $N$ estados aleatoriamente según las distribución de Boltzmann, entonces sólo tenemos que tomar la media de $X$ sobre todos ellos y obtener nuestro promedio. 


- En otras palabras dejamos que el sistema vaya pasando por los $N$ estados con probabilidad $P(E_i)$, de forma que el valor medio de $X$ no sea más que el promedio de los valores que ha tomado en cada uno de ellos.

## 9.3.2 Cadenas de Markov

- Hay un pequeño problema con respecto al algoritmo anterior. 


- Para obtener la muestra aleatoria de estados necesitamos calcular $P(E_i)$ que venía dada por

  $$P(E_i,\beta)=\frac{\text{e}^{-\beta\,E_i}}{Z(\beta)},\quad Z(\beta)=\sum_i\text{e}^{-\beta E_i},$$

  y para ello tenemos que calcular la función de partición $Z$, que es una suma sobre todos los estados del sistema.   
  Es el problema que de hecho queríamos evitar en primer lugar. 
  
  
- Por suerte hay una forma de calcular la probabilidad de Boltzmann sin conocer la función de partición.  Es gracias a las llamadas cadenas de Markov.


- La idea esencial es que podemos generar estados uno detrás de otro, obteniendo por tanto una sucesión que se denomina cadena de Markov.


- Para describir el proceso, asumamos que partimos de un estado inicial $i$, y que, en vez de elegir de forma totalmente aleatoria el siguiente estado, lo obtenemos perturbando ligeramente el estado $i$.


- Por ejemplo, en un gas, sería equivalente a cambiar una sólo de sus moléculas a un nuevo nivel de energía. 


- Para determinar cuán probable es ese nuevo estado, definimos un conjunto de transiciones de probabilidad $T_{ij}$, que definen la probabilidad de pasar del estado $i$ al estado $j$.


- Si somos capaces de determinar de forma correcta este conjunto de probabilidades de transición, es decir, si conseguimos que la probabilidad de llegar en cualquier paso de nuestro cadena al estado $k$ sea la probabilidad de Boltzmann $P(E_k)$, entonces, al movernos sobre la cadena de Markov, estaremos obteniendo exactamente la muestra de estados generados a partir de la distribución de Boltzmann que estábamos buscando.



- El problema reside por tanto en elegir correctamente los valores de $T_{ij}$.


- Puesto que dado un estado $i$ siempre tenemos que acabar en cualquier otro estado, se cumple que

$$\sum_j T_{ij}=1.$$


- Junto a este requisito, el punto clave para obtener $T_{ij}$ es imponer que

    $$\frac{T_{ij}}{T_{ji}}=\frac{P(E_j)}{P(E_i)}=\frac{\text{e}^{-\beta E_j}/Z}{\text{e}^{-\beta E_i}/Z}=\text{e}^{-\beta (E_j-E_i)},$$

  es decir, que el cociente entre la probabilidad de ir de $i\to j$ y de $j\to i$, sea exactamente el cociente de las probabilidades de Boltzmann que determinan cuán probable es que el sistema se encuentre en el estado $j$ e $i$.


- Es importante darse cuenta de que el valor de la función de partición se cancela en el cociente, por lo que no la necesitamos para determinar el cociente. 


- Supongamos ahora que conocemos las probabilidades de transición $T_{ij}$ que satisfacen las dos condiciones anteriores, y asumamos ahora que, en un paso dado de la cadena de Markov, la probabilidad de encontrarse en un estado $i$ viene dada por la probabilidad de Boltzmann. 


- Con todo ello, teniendo en cuenta las condiciones que ha de satisfacer $T_{ij}$, la probabilidad de acabar en el estado $j$ sera

    $$\sum_i T_{ij} P(E_i)=\sum_i T_{ji} P(E_j)=P(E_j)\sum_i T_{ji} =P(E_j).$$


- Esto implica que si en un momento dado la probabilidad de estar en cualquier estado es la probabilidad de Boltzmann, también tendremos la probabilidad de Boltzmann en el siguiente paso, o lo que es lo mismo, la distribución de Boltzmann es un punto fijo de la cadena de Markov. 


- Quedan sólo dos hipótesis por satisfacer:

    1. Que podemos llegar a un paso de la cadena de Markov donde la distribución de probabilidad de acabar en cualquier estado es la de Boltzmann. 

    2. Que podemos encontrar un conjunto de probabilidades de transición que satisfacen las condiciones impuestas.
    
    
    
- Vamos a explicar cómo podemos garantizar cada uno de estos dos puntos. 

### 9.3.2.1 Convergencia de una cadena de Markov hacía la distribución de Boltzmann

- Supongamos que partimos de un estado inicial de nuestro sistema $i$, y de forma iterativa nos movemos a cualquier otro estado con probabilidad $T_{ij}$, donde, tal y como hemos especificado, la probabilidad de transición satisface
  
  $$\sum_j T_{ij}=1,\quad \text{y}\quad\frac{T_{ij}}{T_{ji}}=\frac{P(E_j)}{P(E_i)},$$
  
  donde $P(E_i)$ es la probabilidad de Boltzmann de encontrarse en el estado $E_i$.
  
  
- Definamos ahora como $p_i(t)$ la probabilidad de que la cadena de Markov se encuentre en el estado $i$ en el paso $t$. La probabilidad de transitar al estado $j$ en el siguiente paso $t+1$ será
    
    $$p_j(t+1)=\sum_i T_{ij}p_i(t),$$
    
  que en notación matricial implica que
  
  $${\bf p}(t+1)={\bf T}\cdot{\bf p}(t)$$
  
  donde ${\bf p}$ es el vector con elementos $p_i$ y ${\bf T}$ la matriz cuyo elemento $ij$ es $T_{ji}$.
  
  
- Como $T_{ij}$ es constante en el tiempo, el vector de probabilidad esta multiplicado en cada paso por ella, por lo que tras $t$ pasos tenemos que

  $${\bf p}(t)={\bf T}^t\cdot {\bf p}(0).$$
  
  
- Descompongamos ahora el vector ${\bf p}(0)$ como una combinación lineal de los autovalores por la derecha ${\bf v}_k$ de ${\bf T^t}$

   $${\bf p}(0)=\sum_k c_k {\bf v}_k,$$
    
  e introduciendo esta descomposición en la probabilidad para el paso $t$, tenemos
  
   $${\bf p}(t)={\bf T}^t\sum_k c_k {\bf v}_k=\sum_k c_k \lambda^t_k{\bf v}_k=\lambda_1^t\sum_k c_k \left(\frac{\lambda_k}{\lambda_1}\right)^t{\bf v}_k,$$
   
  donde $\lambda_k$ es el autovalor de ${\bf v}_k$ y $\lambda_1$ es el autovalor principal, es decir, el de mayor magnitud. 
  
  
- La cuestión ahora es estudiar qué ocurre cuando dejamos evolucionar el tiempo, es decir cuando $t\to\infty$.   
  En ese caso, 
  
  $$\lim_{t\to\infty}\frac{{\bf p}(t)}{\lambda_1^t}=c_1 {\bf v}_1,$$
  
  que implica que si dejamos evolucionar nuestro sistema el tiempo suficiente, acabará en el estado correspondiente al autovalor principal.
  
  
- Sólo nos queda por tanto probar que la distribución de Boltzmann es un autovector de la matriz de transición y que además corresponde al autovalor principal.


- La primera parte es fácil de probar, pues sabemos que

    $$\sum_jT_{ji}P(E_j)=\sum_jT_{ij}P(E_i)=P(E_i)\sum_jT_{ij}=P(E_i),$$
    
  por tanto, llamando $p_i=P(E_i)$ tenemos en forma matricial que
  
   $$\bf T\cdot p=p,$$
   
  es decir, la distribución de Boltzmann es un autovector de la matriz de transición con autovalor 1. 


- Sólo nos queda por probar que es el autovector con el autovalor principal.


- Para ello, en primer lugar es conveniente notar que todas las componentes del autovector de Boltzmann son mayores de cero $P(E_i)>0$. 


- Por otro lado, debido a la propiedad

    $$\sum_j T_{ij}=1,$$
    
  podemos ver que el vector ${\bf 1}=(1,1,1,\cdots)$  es un autovector por la izquierda de ${\bf T}$ con autovalor 1, ${\bf 1}^T\cdot {\bf T}={\bf 1}^T$. 
  
  
- Por otro lado, todos los autovectores por la derecha ${\bf \omega}$ con un autovalor $\mu$ diferente de uno  tienen que ser ortogonales al vector identidad, pues
<br>

    \begin{align}
        {\bf 1}^T\cdot{\bf T}\cdot {\bf \omega}=&{\bf 1}^T\cdot(T\cdot {\bf \omega})=\mu\,{\bf 1}^T\cdot{\bf \omega}\\
        {\bf 1}^T\cdot{\bf T}\cdot {\bf \omega}=&({\bf 1}^T\cdot T)\cdot {\bf \omega}=\;\;\,{\bf 1}^T\cdot{\bf \omega}
    \end{align}

    que implica que ${\bf 1}^T\cdot{\bf \omega}=0$.



- Eso implica que todos los autovectores por la derecha con autovalor $\neq 1$ tienen que ser ortogonales al vector $\bf 1$, lo cual sólo es posible si algunas de sus componentes son negativas. 


- Por tanto, si partimos de un punto inicial donde las componentes del vector de probabilidades $\bf p$ son no negativas, al ser $T_{ij}>0$, siempre tenderemos al autovector con el autovalor principal cuyas componentes deben ser todas positivas.


- Sin embargo, el único autovector cuyas componentes son todas positivas es la distribución de Boltzmann (cuyo autovalor es también 1), que implica por tanto que empezando con cualquier probabilidad, nuestro sistema siempre convergerá hacía la distribución de Boltzmann. 


- Lo único que faltaría por probar es que sólo hay un autoestado por la izquierda y derecha asociado al autovalor 1. Lo cual se consigue si todos los estados del sistema son accesibles (si el sistema es ergódico).

## 9.3.3 Obteniendo la matriz de transición: el algoritmo de Metrópolis

- Para finalmente garantizar que podemos calcular el valor medio de nuestro estado $X$, sólo nos queda encontrar encontrar el valor de $T_{ij}$, las cuales deben satisfacer


  $$\sum_j T_{ij}=1,\quad \text{y}\quad\frac{T_{ij}}{T_{ji}}=\frac{P(E_j)}{P(E_i)}.$$
  
- Estas dos restricciones dejan muchas posibilidades para definir $T_{ij}$, pero de forma mayoritaria se usa el algoritmo de Metrópolis.   



- Para explicar cómo funciona, es importante hacer notar en primer lugar, que el sistema puede pasar varias veces por el mismo estado, y que de hecho, puede hacerlo en dos pasos consecutivos. Es decir, $T_{ii}$ puede ser no nulo. 



- Teniendo esto en cuenta, el algoritmo funciona de la siguiente forma. 


- El punto de partida es un estado inicial cualquiera $i$, sobre el que aplicaremos una pequeña modificación aleatoria, generando por tanto el estado $j$. 


- Los cambios posibles son lo que llamamos *move set*.


- Volviendo al ejemplo del gas, una posibilidad sería elegir aleatoriamente qué molécula alterar y cómo variar su estado de energía, aumentándolo o disminuyéndolo. 


- Por último, aceptaremos o rechazaremos la transición a este estado definiendo la probabilidad

    $$P_a=\left\{\begin{array}{ll}1&\text{si}\;E_j\le E_i,\\\text{e}^{-\beta(E_j-E_i)}&\text{si}\;E_j>E_i.\end{array}\right.$$
    
    
- Si la transición se rechaza, entonces el sistema pertenece en el estado $i$ durante un paso más.  
  Si la transición se acepta, el sistema pasará al nuevo estado $j$.


- Dicho de otra forma, si la energía del nuevo estado $j$ es menor que la del estado $i$ la transición siempre será aceptada.   Por otro lado si el nuevo estado tiene más energía sólo permitiremos la transición con la probabilidad de arriba. 


- Veamos ahora que con este algoritmo nuestra probabilidad de transición satisface las condiciones requeridas. 


- En primer lugar la condición 

    $$\sum_j T_{ij}=1,$$
    
  se satisface trivialmente puesto que dado un estado $i$ siempre pasaremos a cualquier otro estado, que puede ser el mismo estado $i$.  


- Siguiendo este esquema, la probabilidad de pasar de un estado $i$ a un estado $j$, será en primer lugar la probabilidad de escoger aleatoriamente el estado $j$, la cual es $1/M$ con $M$ el número total de cambios posibles, multiplicada por la probabilidad de aceptar la transición si $E_j>E_i$.


- Por ejemplo, si $E_j>E_i$ tendremos

   $$T_{ij}=\frac{1}{M}\times \text{e}^{-\beta(E_j-E_i)},\quad T_{ji}=\frac{1}{M}\times 1,$$
   
  que satisface que 
  
  $$\frac{T_{ij}}{T_{ji}}=\frac{\text{e}^{-\beta(E_j-E_i)}/M}{1/M}=\text{e}^{-\beta(E_j-E_i)}.$$
  
  
- Mientras que en el caso contrario, es decir, si $E_j<E_i$  

  $$T_{ij}=\frac{1}{M}\times 1,\quad T_{ji}=\frac{1}{M}\times \text{e}^{-\beta(E_i-E_j)},$$

  y su cociente 
  
  $$\frac{T_{ij}}{T_{ji}}=\frac{1/M}{\text{e}^{-\beta(E_i-E_j)}/M}=\text{e}^{-\beta(E_j-E_i)}.$$



- Así definidas, las probabilidades de transición satisfacen las condiciones necesarias. 



- De esta forma el algoritmo de Metrópolis implica los siguientes pasos

    1.  Elegimos un estado de partida aleatorio.
    2.  Elegimos un movimiento aleatorio de forma uniforme entre un conjunto de estados posibles (tales como cambiar una molécula de estado).
    3.  Calculamos el valor de la probabilidad de aceptación $P_a$.
    4. Aceptamos el movimiento con probabilidad $P_a$, de tal manera que el estado cambiará a otro estado o permanecerá en el estado inicial (en cualquier caso se cuenta en el cálculo del promedio).
    5.  Medir/calcular el valor de la cantidad $X$ que nos interesa en el estado actual e incluirla en la suma de los elementos de la cadena de Markov.
    6. Repetir desde el paso 2. 


- Una vez hemos dado muchos pasos de Markov, dividimos el valor de la suma por el número de pasos de la cadena y obtenemos $\langle X\rangle$.


- La forma más eficiente de implementar la probabilidad de aceptación es calcular el cambio de energía entre los estados $i$ y $j$, $E_j-E_i$, generar un número aleatorio $z$ con probabilidad uniforme entre 0 y 1, y aceptar la transición si $z<\text{e}^{-\beta(E_j-E_i)}$.   
  Si $E_j\le E_i$ la exponencial es mayor o igual que 1 y la transición siempre será aceptada. 
  
  
- Es importante tener en cuenta la siguientes sutilezas

    - Los pasos en los que rechazamos la transición deben contar como estados, incluso cuando no hemos hecho ningún cambio. 
    
    - Para que la transición esté bien definida, es esencial que el número $M$ de transiciones posibles para ir de $i\to j$ sea igual al de $j\to i$.
    - Es esencial asegurar que cualquier estado posible es accesible. Es lo mismo que decir que el estado debe ser ergódico. Es una condición necesaria para asegurar que el sistema converge a la distribución de Boltzmann.
    
    - Aunque hemos probado que el sistema converge a la distribución de Boltzmann, no hemos calculado cuánto tarda en converger a ella, o lo que es lo mismo cuánto tarda en equilibrarse. Aunque no hay una regla universal para calcular este tiempo, en general es bastante fácil determinarlo. 
    
    - Lo mejor es verlo con un ejemplo.

### Ejemplo 9.11: simulación de Monte Carlo de un gas ideal

- Los estados cuánticos de una partícula de masa $m$ en una caja cúbica de longitud $L$ vienen caracterizados por los enteros $n_x,\, n_y,\, n_z=1,\cdots,\infty$, siendo su energía

    $$E(n_x,n_y,n_z)=\frac{\pi^2\hbar^2}{2m^2L^2}\left(n_x^2+n_y^2+n_z^2\right).$$
    
    
- Sea un gas ideal formado por $N$ partículas definidas por sus energías cuánticas que no interaccionan. La energía del sistema será por tanto la suma de las energías de cada una de las partículas
    
  $$E=\sum_i E(n_x^{(i)},n_y^{(i)},n_z^{(i)}),$$
  
  siendo $n_{x,y,z}^{(i)}$ el número cuántico correspondiente de la partícula $i$.
  
 
- Estimar la energía promedio de un gas ideal con $N=1000$ partículas para $k_B T=10$ usando las unidades $L=m=\hbar=1$.

- En primer lugar el conjunto de estados posibles serán los que se obtienen de cambiar los números cuánticos $n_x$,  $n_y$ o $n_z$ de un único átomo en $\pm 1$.


- En cada paso de la cadena de Markov escogeremos aleatoriamente una partícula, escogeremos aleatoriamente un número cuántico $n_x$,  $n_y$ o $n_z$ y aleatoriamente elegiremos un cambio, +1 o −1 para el número cuántico.


- El cambio de la energía será, si pasamos de $n_x$ a $n_x + 1$ será

    $$\Delta E=\frac{\pi^2\hbar^2}{2m^2L^2}(2n_x+1),$$    
    
  y si pasamos de    $n_x$ a $n_x - 1$

   $$\Delta E=\frac{\pi^2\hbar^2}{2m^2L^2}(-2n_x+1),$$
      
  y analogamente para los otros números cuánticos.  
  
  
- Partamos del estado inicial de menor energía, es decir $n_x^{(i)}= n_x^{(i)}=n_x^{(i)}=1$, $\forall$ $i \in [0,N]$ y realicemos 250000 pasos de Monte Carlo. 

```python
from random import random,randrange
from numpy import exp,pi
from numpy import ones
from matplotlib.pyplot import plot,xlabel,ylabel,show

# Constantes

T=20.0           # tempeatura
N=1000           # número de partículas
pasos=250000     # número de pasos


# Creamos un array 2-dimensional para almacenar los números cuánticos
n=ones([N,3],int)

# Bucle principal

eplot = []
E=3*N*pi*pi/2  # Energía total del estado inicial del gas

for k in range(pasos):

    # Escogemos la partícula y el movimiento (número cuántico y +1 o -1)

    i=randrange(N)    # escogemos la partícula
    j=randrange(3)    # escogemos el número cuántico

    # escogemos si aumentamos o disminuimos el nivel de enerǵia    
        
    if random()<0.5:
        dn=1                        # Incremento del número cuantico
        dE = (2*n[i,j]+1)*pi*pi/2   # incremento de energía
    
    else:
    
        dn = -1                          # Decremento del número cuantico
        dE = (-2*n[i,j]+1)*pi*pi/2       # Decremento de energía

    # Decidimos si se acepta el movimiento
    
    if n[i,j]>1 or dn==1:          # Si n es 1 dn sólo puede ser 1, pues 1 es el menor nivel de energía.
        
        if random()<exp(-dE/T):
            n[i,j]+=dn
            E+=dE

    eplot.append(E)
```

```python
# Hacemos la gráfica

plot(eplot)
ylabel("Energía")
xlabel("pasos")
show()
```

### Ejercicio 9.4: El modelo de Ising


El modelo de Ising proporciona una descripción teórica del ferromagnetismo. La magnetización de un material magnético es consecuencia de la combinación de muchos dipolos magnéticos a lo largo de todo el material. 


Si cada uno de estos dipolos apuntan en direcciones aleatorias, entonces la magnetización total del sistema será prácticamente nula. 
Sin embargo si la mayoría de ellos apuntan en una única dirección, entonces el sistema adquiere una magnetización macroscópica. 


El modelo de Ising describe este fenómeno en el caso de que cada uno de los dipolos del material sea consecuencia del spin de cada uno de sus átomos, los cuales se ordenan en un retículo y apuntan sólo en dos direcciones, hacía arriba o hacía abajo. Por ejemplo en un retículo de dos dimensiones

```python
from IPython.display import Image,display
display(Image("spins.png",width=400))
```

Matemáticamente los espines pueden representarse con la variable $s_i=\pm 1$ para cada punto del retículo, donde +1 representa que el espín apunta hacía arriba y -1 cuando apunta hacía abajo. 

El otro aspecto fundamental es que los dipolos interaccionan entre ellos de forma que puede ser energéticamente favorable que se alineen en la misma dirección. La energía potencial magnética debida a la interacción de dos dipolos $i$ y $j$ en el modelo de Ising viene dada por el producto $s_i s_j$, siendo la energía de la interacción $-Js_i s_j$, con $J$ una constante positiva.

El signo negativo asegura que la energía del sistema disminuye cuando los dipolos están alineados, y es lo que define a los materiales ferromagnéticos. Los materiales ferromagnéticos adquieren por tanto una magnetización si tienen la oportunidad para ello. 
Aquellos materiales donde el signo de la interacción es positivo se llaman antiferromagnéticos, porque son sistemas que tenderán a tener una magnetización total nula. 

En el modelo de Ising se asume que los espines sólo interaccionan con sus vecinos más próximos, es decir con aquellos que se encuentran en elementos adyacentes del retículo. De esta forma, la energía total del sistema es

$$E=-J\sum_{\langle ij\rangle}s_is_j,$$

# 9.4 Máximos y mínimos: simulated annealing (templado simulado)

- Uno de los mayores usos de los métodos de Monte Carlo en los últimos años es en la optimización numérica. 


- En el tema 5 vimos diferentes métodos que nos permiten encontrar mínimos o máximos de una función. Sin embargo, estos métodos no nos dicen si ese mínimo es global o local. 



- Sin embargo, en muchos casos esto no es suficiente. Por ejemplo si buscamos el estado fundamental de un estado físico estamos buscando el mínimo global del sistema. Es el mismo caso si buscamos diseñar los parámetros de un automóvil para minimizar su consumo. 


- Cuando queremos encontrar el mínimo global de una función $f$ (el máximo sería equivalente a buscar el mínimo de $-f$), los métodos del tema 5, al menos por sí solos, no funcionan.  


- La optimización global es uno de los problemas computacionales más complicados y en los que más recursos se invierte.


- Uno de los métodos más prometedores para este problema es el llamado simulated annealing (templado simulado), que fue propuesto en 1985 por Scott Kirkpatrick. 


- Para explicarlo volvamos a la Mecánica Estadística e imaginemos un sistema en equilibrio a la temperatura $T$. 


- La probabilidad de que en algún momento nuestro sistema se encuentre en el estado $i$ viene dada por la probabilidad de Boltzmann

    $$P(E_i)=\frac{\text{e}^{-\beta E_i}}{Z},\quad\text{con}\quad Z=\sum_i\text{e}^{-\beta E_i},$$
    
  donde una vez más $\beta=1/k_B T$.
  
  
  
- Asumamos ahora que el sistema tiene un sólo estado fundamental y elijamos el sistema de referencia de forma que $E_i=0$ para el estado fundamental y $E_i>0$ para el resto. 



- Supongamos ahora que enfriamos nuestro sistema al cero absoluto. En el límite $T\to 0$, tenemos que $\beta\to\infty$, por lo que $\exp\left(-\beta E_i\right)\to 0$ excepto para el estado fundamental donde siempre se cumplirá que $\exp\left(-\beta E_i\right)=1$.



- Por tanto en este límite Z=1 y la probabilidad de Boltzmann es

    $$P(E_i)=\left\{\begin{array}{ll}1&\text{para}\;E_i=0,\\ 0&\text{para}\;E_i>0.\end{array}\right.$$
    
    
- En otras palabras, implica que el sistema tiene que estar en el estado fundamental.    


- Por consiguiente, si conseguimos enfriar el sistema hasta el cero absoluto sólo tenemos que observar el estado en el que ha terminado.


- Esto sugiere una posible estrategia para encontrar el estado fundamental: 

    1. Simulamos nuestro sistema a una temperatura $T$ usando el método de las cadenas de Markov. 
    2. Reducimos la temperatura paso a paso, mientras vamos dejando evolucionar el sistema. 
    3. Llegamos a cero y observamos el estado en el que nuestro sistema ha termalizado.
    
    
- El aspecto interesante del templado simulado es que el mismo método se puede aplicar a cualquier otra función, no sólo a un sistema cuyo estado fundamental queremos encontrar.  


- De forma general, podemos coger cualquier función matemática $f(x,y,z,\cdots)$, y tratar las variables como aquellas que nos permiten definir un "estado" del "sistema", y la función como su energía. 


- Aplicando el mismo proceso que en física estadística y reduciendo la temperatura a cero, podemos volver a obtener el "estado fundamental" del sistema, los valores de las variables que minimizan el valor de la función.


- Para que el método funcione es necesario bajar la temperatura del sistema lentamente. Si este es el caso, se puede probar que el sistema siempre encontrará el estado fundamental, o lo que es lo mismo, el mínimo de la función. 


- Por regla general, la temperatura inicial del sistema se debe tomar mayor que la variación típica en un paso de MonteCarlo, es decir $\beta (E_i-E_j)\le 1$, de forma que $P_a\simeq 1$ para casi todos los movimientos posibles, y el sistema se vuelva totalmente aleatorio sin importar cual era su estado inicial. 


- Tras ello uno empieza a enfriar el sistema, lo cual se suele hacer eligiendo la función de enfriamiento

    $$T=T_0\text{e}^{-t/\tau},$$
    
  siendo $T_0$ la temperatura inicial del sistema y $\tau$ una constante, que en general se suele coger para optimizar el cociente precisión ($\tau$ grande) vs rapidez ($\tau$ pequeña).

## Ejemplo 9.12: el problema del viajante

Para entender el templado simulado vamos a resolver uno de los problemas de optimización canónicos, el problema del viajante, que involucra encontrar el camino más corto que recorre toda una serie de puntos en un mapa. 

```python
from IPython.display import Image,display
display(Image("salesman_problem.png",width=400))
```

Es un problema importante porque pertenece a la clase de problemas NP, problemas cuyo tiempo de resolución escala como un polinomio. 

El problema se formula de la siguiente manera. Un vendedor tiene que recorrer $N$ ciudades, pudiendo viajar siempre en la linea recta que une cualquier par de ciudades. Dada la localización de las ciudades, el problema reside en encontrar el camino más corto (en términos de la distancia recorrida), que le permite recorrer todas las ciudades empezando y acabando en la misma ciudad. 

Para simplificar el problema asumamos que el vendedor se mueve en un espacio bidimensional, y escojamos la posición de $N=25$ ciudades de forma aleatoria  en un espacio de longitud 1 por cada lado.
Numeremos las ciudades en el orden en las que el viajero las va a visitar, y denotemos la posición de la ciudad  $i$ con el vector $r_i=(x_i,y_i)$, siendo $r_0=r_N$, pues el viaje tiene que acabar en el punto de partida.   

De esta forma, la distancia recorrida por el viajante es

$$D=\sum_{i=0}^{N-1}\vert r_{i+1}-r_i\vert.$$

Queremos minimizar esta cantidad sobre todo el posible orden de las ciudades usando el templado simulado. Para hacerlo, tenemos que definir en primer lugar los movimientos posibles dentro de la cadena de Markov, que definimos como el intercambio del orden de dos ciudades. 

Si el cambio disminuye la distancia total, entonces siempre lo aceptamos. Si no, lo aceptamos con la probabilidad del algoritmo de Metrópolis. 
En cada paso hay que actualizar la temperatura siguiendo la función de enfriamiento exponencial. 

Vamos a generar una animación que describa el proceso de la animación.

```python
from numpy import sqrt,exp,empty
from numpy.linalg import norm
from random import random,randrange

# constantes del problema

N=25         # número de las particulas

Tmax=10      # temperatura inicial
Tmin=1e-3    # temperatura final 
tau=1e4      # escala de enfriamiento


# función para calcular la distancia total del viaje

def distancia():
    s=0
    for i in range(N):
        s+=norm(r[i+1]-r[i])
    return s

# escogemos la posicion de las N ciudades de forma aleatoria

r=empty([N+1,2],float)

for i in range(N):
    r[i,0]=random()
    r[i,1]=random()
        
r[N]=r[0]

# distancia inicial

D=distancia()
```

```python
# preparamos la animación

from vpython import curve,rate,canvas,vector,sphere
scene = canvas(width=500, height=500,
     center=vector(0.5,0.5,0))

R=0.02    # describimos cada ciudad por medio de una efera de radio R

listp=[]

for i in range(N):
    sphere(pos=vector(r[i,0],r[i,1],0),radius=R)
    listp.append(vector(r[i,0],r[i,1],0))

listp.append(vector(r[N,0],r[N,1],0))    
l=curve(pos=listp,radius=R/4)
```

```python
# loop principal

t=0        # tiempo de enfriamiento
T=Tmax     # temperatura inicial

while T>Tmin:
    t+=1
    T=Tmax*exp(-t/tau)    # enfriamos
        
    if t%100==0:          # cada 100 pasos actualizamos la animación y cambios el orden de las ciudades
        listp=[]
        for i in range(N+1):
             l.modify(i,pos=vector(r[i,0],r[i,1],0))
        rate(25)

    # escogemos dos ciudades para intercambiar. Nos aseguramos que son diferentes. 
    
    i,j=randrange(1,N),randrange(1,N)
    while i==j:
        i,j=randrange(1,N),randrange(1,N)

    # intercambiamos las ciudades
    
    Dp=D                  # distancia antes de cambiar
    r[i,0],r[j,0]=r[j,0],r[i,0]
    r[i,1],r[j,1]=r[j,1],r[i,1]
    
    D=distancia()
    deltaD=D-Dp
    
    # si el movimiento es recahzado las cambiamos otra vez
    
    if random()>exp(-deltaD/T):
        r[i,0],r[j,0]=r[j,0],r[i,0]
        r[i,1],r[j,1]=r[j,1],r[i,1]
        D=Dp
```

## Ejercicio 9.8: obteniendo el número pi con una cadena de Markov

El objetivo de este problema es ilustrar cómo se puede calcular $\pi$ mediante un muestreo aleatorio del disco unitario utilizando una cadena de Markov. 

La idea es inicializar nuestra cadena en el centro de un cuadrado de lado 2, y en cada paso, alterar su posición mediante un paso aleatorio que nunca permita salir del cuadrado. De esta forma, el número $\pi$ se podrá estimar contando el número de pasos en los que la cadena se ha encontrado dentro de una circunferencia de radio unidad.  

Para crear la cadena de Markov hay que seleccionar un tamaño de paso en el intervalo $[-p,p]$. El valor de $p$ influye mucho en el algoritmo. Si $p$ es demasiado pequeño, convergerá lentamente.
Si $p$ es demasiado grande, se rechazarán muchos movimientos. Lo ideal es que el $50\,\%$ de los movimientos sean aceptados. Para comprobarlo, mida la probabilidad de que se produzca un movimiento, es decir, 

prob = (#jugadas_aceptadas/#total_intentos). Los valores de $p$ entre 0.2 y 1.0 suelen ser los óptimos.


En la simulación ejecutar el código varias veces para $N= 10^i$, $i = 1,\,2,\,3,\cdots$ números aleatorios, observar cómo mejora la estimación de $\pi$ al aumentar N y calcular la desviación del resultado exacto.

```python
#imports
import numpy as np
import matplotlib.pyplot as plt
from random import random
#Ejercicio
L=1
#Definimos una función para el paso de tamaño p en la posicion (x,y)
def step(x,y,p): 
    dx=p*(random()*2-1)
    dy=p*(random()*2-1)
    if -L<(x+dx)<L and -L<(y+dy)<L: #La funció se asegura de que el paso se da dentro del cuadrado
        return x+dx,y+dy,1
    else:
        return x,y,0
#Definimos una función que compruebe si el paso se ha dado en el disco unidad o no
def inout(x,y):
    if np.sqrt(x**2+y**2)<=1:
        return 1
    else:
        return 0
x=0 
y=0
#Comenzamos el bucle:
pi=[]
X=[]
sigma=[]
for i in range(2,7):
    N=10**i
    X.append(N)
    pasos=0
    pasosadd=0
    s=0
    for j in range(N):
        x,y,add=step(x,y,1.15)
        pasos+=1 
        pasosadd+=add
        s+=inout(x,y)
    prob=pasosadd/pasos
    p=4*s/N
    sig=np.sqrt(p*(4-p)/N)
    sigma.append(sig)
    pi.append(p)
    print(f'Para {N} pasos, la probabilidad de que se de un paso dentro del disco unidad es {prob}')
plt.plot(X,pi,'r.')
plt.axhline(y=np.pi,color='green',linestyle='--',label='pi')
plt.xscale('log')
plt.legend()
plt.title('Valor de pi para N pasos')
plt.grid()
plt.show()
plt.plot(X,sigma,'r.')
plt.xscale('log')
plt.yscale('log')
plt.title('Desviación del resultado de pi para cada N pasos')
plt.grid()
plt.show()
```

## Ejercicio 9.11: estudiando el fenómeno de la percolación

Los fenómenos de percolación son aquellos en los que se filtra una sustancia a través de un determinado espacio o medio poroso. 

Para estudiar este fenómeno, vamos tratar de calcular la probabilidad de que, dentro de un conjunto de elementos distribuidos de forma aleatoria,  exista un camino que conecte dos extremos del conjunto. Este sistema puede servir para estudiar gran variedad de fenómenos, como por ejemplo la propagación de incendios forestales o epidemias, la filtración de líquidos en materiales porosos o la extracción de petroleo.


Para estudiar la percolación considérese una red bidimensional de $N\times N$ sitios. Cada sitio tiene una probabilidad $p$ de estar ocupado y, por tanto, una probabilidad $(1-p)$ de estar vacío. Dos sitios están conectados entre si sólo si ambos están llenos (o vacíos). 
Diremos que el sistema percola si existe un cluster que conecta ambos extremos. 

Si $p=1$, es evidente que todos los sitios están ocupados y por tanto el sistema percola. 
Por el contrario, si $p = 0$ todos los sitios están vacíos y el sistema no percolará nunca. Así pues, podemos definir una probabilidad de percolación, la cual aumentará conforme $p$ lo haga. Además, existirá un umbral crítico de percolación asociado a una probabilidad crı́tica de ocupación $p_c$, de modo que si $p>p_c$ existe para un determinado sistema, habrá al menos un camino que conecte los extremos del sistema.

La percolación está caracterizada por el valor de la probabilidad crítica $p_c$ y el exponente crítico que esta relacionado con el tamaño medio de los clusters. Esta última rige el proceso de cambio de fase geométrico que estamos estudiando.

```python
import numpy as np
import matplotlib.pyplot as plt
from random import random
#1
N=3
p=0.5 #Probabilidad de que el elemento sea 1
def matriz(N,p):
    Matriz=np.random.choice([0,1],size=(N,N),p=[1-p,p])
    return Matriz
#2
#Comprobar si percola:
def foco(Matriz):
    N=len(Matriz)
    for i in range(N):
        if Matriz[i,0]==1:
            Matriz[i,0]=-10
            
    cambiando=True
    while cambiando:
        cambiando=False
        for i in range(N):
            for j in range(N):
                if Matriz[i,j]==-10:
                    for dx,dy in [(1,0),(-1,0),(0,1),(0,-1)]:
                        ni, nj= i+dx,j+dy
                        if 0<=ni<N and 0<=nj<N and Matriz[ni,nj]==1:
                            Matriz[ni,nj]=-10
                            cambiando=True
    return Matriz
def percola(Matriz):
    return np.any(Matriz[:, -1] == -10)
def main(N,p):
    Matriz=matriz(N,p)
    Foco=foco(Matriz)
    resultado=percola(Foco)
    return resultado
#3 Calculamos la probabilidad por Monte Carlo:
M=1000
N=20
p=0.6
s=0
for _ in range(M):
    resultado=main(N,p)
    if resultado:
        s+=1
P=s/M
sigma=np.sqrt(P*(1-P)/M)
print(f'La probabilidad de percolación para dimensión {N} y probabilidad de ocupación {p}, es P={P}, con una desviación estándar {sigma}')
#4
P=np.arange(0.58,0.62,0.001)
N=[20,30,40,50]
lprob=[]
lsigma=[]
tal = np.array(['red','blue','green','yellow'],dtype=str)
aux=0
for n in N:
    for p in P:
        s=0
        for _ in range(M):
            resultado=main(n,p)
            if resultado:
                s+=1
        prob=s/M
        sigma=np.sqrt(P*(1-P)/M)
        lprob.append(prob)
        lsigma.append(sigma)
    plt.plot(P,lprob,label=f'P(N={n})',color=tal[aux])
    lprob=[]
    lsigma=[]
    aux+=1
plt.xlim(0.58,0.60)
plt.ylim(0,1.1)
plt.xlabel('Valores de p')
plt.ylabel('P(p)')
plt.legend()
plt.show()
```

## Ejercicio 9.12: planificación empresarial mediante una simulación de Montecarlo


Imagínese que usted es el director de marketing de una empresa que está planeando introducir un nuevo producto. Necesita estimar el beneficio neto del primer año de este producto, que en general dependerá de

- El precio por unidad
- Coste unitario
- Costes fijos


El beneficio neto se calculará como:

$$\text{Beneficio neto}=\text{Volumen de ventas}*(\text{Precio de venta} - \text{Coste unitario})-\text{Costes fijos}.$$  

Se sabe que los costes fijos (para gastos generales, publicidad, etc.) son de 120.000 euros. Pero los demás factores implican cierta incertidumbre. **Variables imprevistas**

Para construir un modelo de análisis de riesgos, primero debemos identificar las variables imprevistas, también llamadas variables aleatorias.  Aunque hay cierta incertidumbre en casi todas las variables de un modelo de negocio, queremos centrarnos en las variables cuyo rango de valores es significativo.

**Ventas y precio**

Basándose en su investigación de mercado, usted cree que hay las mismas posibilidades de que el mercado sea "lento", "neutro" o "caliente".

- En el escenario de "mercado lento", espera vender 50.000 unidades a un precio medio de venta de 11 euros por unidad.

- En el escenario de "mercado neutro", usted espera vender 75.000 unidades, pero probablemente tendrá que fijar un precio medio de venta más bajo, de 10 euros por unidad.

- En el escenario de "mercado caliente", espera vender 100.000 unidades, pero esto hará que los competidores bajen el precio medio de venta a 8 euros por unidad.


Como resultado, usted espera vender 75.000 unidades (es decir, (50.000+75.000+100.000)/3 = 75.000) a un precio medio de venta de 9,67 euros por unidad (es decir, (11+10+8)/3 = 9,67 euros).


**Coste unitario**

Otra variable imprevista es el coste unitario.  El director de producción de su empresa le informa de que los costes unitarios pueden oscilar entre 5,50 y 7,50 dólares, con un coste más probable de 6,50 dólares. En este caso, el coste más probable es también el coste medio.


**Funciones imprevistas:**

**Beneficio neto**

Nuestro siguiente paso es identificar las funciones imprevistas, también llamadas funciones de una variable aleatoria.  En principio el volumen de ventas, el precio de venta y el coste unitario son variables inciertas, por lo que el beneficio neto es una función incierta.


**La paradoja de Simpson**

Antes de explorar cómo realizar una simulación de Montecarlo para analizar este problema, considere el resultado de tomar la media esperada tanto en las ventas como en los costes:
    
$$\text{Beneficio neto}=\text{Volumen de ventas}*(\text{Precio de venta} - \text{Coste unitario})-\text{Costes fijos}=75000*(9.67-6.50)-120000=117500\,\text{euro}$$  
   
La intuición podría sugerir que si introducimos el valor medio de nuestros factores inciertos (volumen de ventas, precio de venta y coste unitario) en nuestro modelo deberíamos obtener el valor medio del resultado (beneficio neto).  

Sin embargo, como veremos en un momento, la cifra de beneficio neto de 117.750 dólares calculada por este modelo, basada en los valores medios de los factores inciertos, es bastante engañosa. El verdadero beneficio neto medio está más cerca de los 93.000 dólares.  Como advierte el Dr. Sam Savage, "los planes basados en supuestos medios serán erróneos por término medio".


**Introduciendo la aleatoriedad:**

**Ventas y precio**

Dado que existen las mismas posibilidades de que el mercado sea "lento", "neutro" o "caliente", queremos crear una variable imprevista que seleccione entre estas tres posibilidades generando un número aleatorio, digamos 1, 2 o 3, con la misma probabilidad. 

Podríamos entonces asociar el número 1 con el estado "Mercado caliente", el número 2 con el estado "Mercado neutro", y el número 3 con el estado "Mercado lento". 
De esta forma, basaremos el Volumen de Ventas y el Precio de Venta en el valor de esta variable incierta.

A continuación, utilizaremos ese valor aleatorio para obtener los valores adecuados de Volumen de ventas y Precio de venta  asociados a cada estado del mercado.


**Coste unitario**

 Ahora nos ocuparemos del Coste Unitario.  Tenemos no sólo tres, sino muchos valores posibles para esta variable: Puede estar entre 5,50 y 7,50 dólares, con un coste más probable de 6,50 dólares.  Una forma burda pero eficaz de modelar esta probabilidad es usar una distribución triangular (https://es.wikipedia.org/wiki/Distribuci%C3%B3n_triangular). Para ello es necesario usar el método de la transformación para obtener un número aleatorio distribuido según la distribución triangular. 


**Simulación de Montecarlo**

Ya tenemos todas las herramientas para realizar nuestra simulación de MonteCarlo asociado a la incertidumbre del mercado. Repetir este procedimiento un número $N$ de veces, y a partir de él estimar el Beneficio neto esperado y su desviación típica. 

```python
import numpy as np
import matplotlib.pyplot as plt
import random
def main():
    N=1000000
    CostesFijos=120000
    lBN=[]
    for _ in range(N):
        x=random.randint(1,3)
        if x==1:
            VolumenVentas=50000
            PrecioVentas=11
        elif x==2:
            VolumenVentas=75000
            PrecioVentas=10
        else:
            VolumenVentas=100000
            PrecioVentas=8
        CosteUnitario=np.random.triangular(5.5,6.5,7.5)
        BeneficioNeto=VolumenVentas*(PrecioVentas-CosteUnitario)-CostesFijos
        lBN.append(BeneficioNeto)
    media=np.mean(lBN)   
    BNesperado=round(media,2)
    sigma=np.std(lBN,ddof=1)
    print('El beneficio neto esperado es de ',BNesperado,'.La desviaciçon típica es de',sigma,'dólares')
    plt.grid()
    plt.title('Beneficio neto')
    plt.hist(lBN, bins=100)
   
main()
```

## Ejercicio 9.13: el juego de serpientes y escaleras


El juego de Serpientes y Escaleras (https://es.wikipedia.org/wiki/Serpientes_y_escaleras) es un buen candidato para el análisis con una Cadena de Markov debido a su falta de memoria: en un punto dado del juego, la progresión del jugador desde la casilla actual es independiente de cómo llegó a esa casilla.

En la teoría de las cadenas de Markov, la probabilidad de un movimiento de la casilla i a la j viene dada por una matriz de transición, T. Consideremos primero un tablero con 100 casillas y sin serpientes ni escaleras. 

El jugador comienza el tablero en una casilla que numeramos como 0, por lo que la matriz de transición tiene unas dimensiones de 101×101, donde etiquetamos las filas i=0,1,2,⋯100 como casillas desde las que moverse y las columnas j=0,1,2,⋯100 como casillas a las que moverse. 

La primera fila de la matriz de transición representa, por tanto, las probabilidades de moverse a cada casilla desde la casilla 0, la inicial; la segunda fila representa las probabilidades de moverse a cada casilla desde la casilla 1, y así sucesivamente. Los movimientos se deciden mediante la tirada de un dado justo de seis caras, por lo que las primeras filas de la matriz de transición en este caso tienen el siguiente aspecto:

$$\left(\begin{array}{cccccccccccc}
0&1/6&1/6&1/6&1/6&1/6&1/6&0&0&0&\cdots&0\\
0&0&1/6&1/6&1/6&1/6&1/6&1/6&0&0&\cdots&0\\
\vdots&&\dots&\dots&&\vdots&&\dots&&&\vdots\\
\end{array}\right)
$$


Es decir, (leyendo la primera fila) empezando en la casilla 0, hay una probabilidad cero de permanecer allí, y una probabilidad 1/6 de caer en cada una de las casillas numeradas del 1 al 6. Al leer la segunda fila se obtienen las probabilidades de progresión desde la casilla 1 (1/6 para cada una de las casillas de destino 2-7), y así sucesivamente.

Algunas personas juegan a Serpientes y Escaleras con el requisito de que, para ganar, el jugador debe caer exactamente en la casilla 100, pero para evitar prolongar el juego más de lo estrictamente necesario permitamos que cualquiera con una tirada que caiga o pase de 100 gane. Esto cambiará las probabilidades en la matriz de transición para las casillas 94-99. Por ejemplo, a partir de la casilla 97, una tirada de 3,4,5 o 6 es suficiente para ganar. La fila inferior de la matriz de transición estará llena de ceros porque no hay ningún lugar al que ir desde la casilla 100.


El juego puede analizarse con un vector de filas, v con 101 componentes, que representa las probabilidades de que el jugador esté en cada una de las casillas. Inicialmente, v(0) es (1,0,0,⋯,0): el jugador está ciertamente en la casilla 0 antes de que el juego haya comenzado. Posteriormente, v evoluciona según la relación

$${\bf v^{(k+1)}}={\bf v^{(k)}}\cdot {\bf T}$$

Es decir, las probabilidades para el siguiente movimiento, k+1, vienen dadas por el producto punto del vector de estado actual, v(k) y la matriz de transición, T.

|15|	37|	18|	13|
|22|	42|	28|	12|
|25|	64|	36|	34|

**Ayuda**: La probabilidad de llegar a las casillas "De" anteriores puede ponerse a cero y añadirse a la casilla "A" correspondiente.





```python
import numpy as np
import matplotlib.pyplot as plt
import random
def main():
    N=101
    k=100
    T=np.zeros((N,N),float)
    for i in range(N-1):
        for j in range(1,7):
            destino=i+j
            if destino >= N-1:
                T[i,N-1]+=1/6
            else:
                T[i,destino]=1/6
    T[N-1,N-1]=1
    v0=np.zeros((1,N),int)
    v0[0,0]=1
    v=[v0.copy()]
    for _ in range(k):
        v0=v0 @ T
        v.append(v0.copy())
        
    # Graficar la evolución para ver cómo se mueve la densidad
    plt.figure(figsize=(10, 6))
    for p, color in zip([1, 5, 10, 20, 50, 100],
                        ['b', 'g', 'r', 'c', 'm', 'k']):
        plt.plot(v[p].ravel(), label=f"{p} tiradas")

    plt.title("Evolución de la probabilidad en Serpientes y Escaleras (sin serpientes)")
    plt.xlabel("Casilla")
    plt.ylabel("Probabilidad")
    plt.legend()
    plt.grid(True)
    plt.show()
main()
```
