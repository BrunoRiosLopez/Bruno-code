- Las ecuaciones diferenciales en derivadas parciales aparecen en algunos de los problemas físicos más relevantes:
<br/>

    - La ecuación de ondas.

    - La ecuación de difusión.

    - La ecuación del calor.

    - La ecuación de Laplace y Poisson (electrostática o dinámica de fluidos)

    - Las ecuaciones de Maxwell.

    - La ecuación de Schrödinger.


- Son ecuaciones para una o varias variables dependientes de varias variables independientes.


- En general las ecuaciones en derivadas parciales que aparecen en física son ecuaciones diferenciales semilineales de orden 2.   
  (Una ecuación en derivadas parciales de orden $n$ se llama semilineal si puede escribirse como suma de una función lineal de la derivada de orden $n$ más una función cualquiera del resto de derivadas).
  
  
- Es decir, para el caso de dos dimensiones, son ecuaciones del tipo  
  
  $${\displaystyle A\phi_{xx}+2B\phi_{xy}+C\phi_{yy}+D=0,}$$
  
    donde $D$ es una función que no involucra las segundas derivas de la función $\phi(x,y)$. 


- Las ecuaciones semi-lineales de segundo orden se clasifican en ecuaciones en:
<br/>
    
    - Elípticas (Laplace, Poisson), $AC-B^2>0$.
    - Hiperbólicas (ecuación de ondas), $AC-B^2<0$.
    - Parabólicas (ecuación del calor o difusión) $AC-B^2=0$.
<br/>
      
      
- Al igual que la ecuaciones diferenciales ordinarias, las ecuaciones en derivadas parciales se suelen clasificar en
<br/>

    - Ecuaciones con condiciones de contorno, en general asociados a las ecuaciones elípticas  (Laplace, Poisson).
    - Ecuaciones con condiciones iniciales, en general asociados a las ecuaciones hiperbólicas y parabólicas. 


# 8.1: Ecuaciones con condiciones de contorno

- Son aquellos sistemas de ecuaciones cuya solución se conoce en la frontera o en un determinado contorno.


- La principal técnica para resolverla es el método de las diferencias finitas que hemos visto en el Tema 7. 

## 8.1.1 Método de Jacobi o de diferencias finitas

- Es la generalización a derivadas parciales del método que hemos visto para ecuaciones diferenciales ordinarias.


- Vamos a estudiar el método para el caso más sencillo, la ecuación de Laplace en dos dimensiones, que, por ejemplo, aparece en problemas de electrostática. 


- El potencial electrestático $\phi$ está relacionado con el vector del campo eléctrico ${\bf E}$ a través de la ecuación

    $${\bf E}=-{\bf \nabla}\phi,$$

    de forma que la primera ecuación de Maxwell en el vació, la ecuación diferencial de gauss para el campo eléctrico, implica

    $${\bf \nabla}\cdot {\bf E}={\bf \nabla}\cdot {\bf \nabla}\phi=\nabla^2\phi=0,$$

    que es la ecuación de Laplace.


- En coordenadas cartesianas se puede expresar como

    $$\frac{\partial^2\phi}{\partial x^2}+\frac{\partial^2\phi}{\partial y^2}+\frac{\partial^2\phi}{\partial z^2}=0.$$


- Para resolver el problema asumamos una caja en el vacío cuyas paredes son conductoras. Todas las paredes están conectadas a la tierra excepto la pared superior que tiene un potencial $V$. 

```python
from IPython.display import Image,display
display(Image("electrostatic.png",width=400))
```

- La dificultad del problema consiste en resolver la ecuación de Laplace en dos dimensiones

    $$\frac{\partial^2\phi}{\partial x^2}+\frac{\partial^2\phi}{\partial y^2}=0,$$

    con las condiciones de contorno $\phi=V$ en la parte superior de la caja y $\phi=0$ para el resto de paredes. 


- Como ya hemos visto en las ecuaciones diferenciales ordinarias, el método de las diferencias finitas implica discretizar nuestro espacio en una red de puntos (lattice) equiespaciados en las dos direcciones cartesianas por una distancia $a$, y expresar las derivadas segundas usando diferencias finitas

    \begin{align}
    \frac{\partial^2\phi(x,y)}{\partial x^2}=&\frac{\phi(x+a,y)-2\phi(x,y)+\phi(x-a,y)}{a^2},\nonumber\\
    \nonumber\\
    \frac{\partial^2\phi(x,y)}{\partial y^2}=&\frac{\phi(x,y+a)-2\phi(x,y)+\phi(x,y-a)}{a^2},\nonumber\\
    \end{align}

    de forma que en cada punto de nuestro retículo la ecuación de Laplace es

    $$\frac{\partial^2\phi}{\partial x^2}+\frac{\partial^2\phi}{\partial y^2}=\frac{\phi(x+a,y)+\phi(x-a,y)+\phi(x,y+a)+\phi(x,y-a)-4\phi(x,y)}{a^2},$$

```python
from IPython.display import Image,display
display(Image("electrostatic2.png",width=400))
```

- Por tanto, tenemos una ecuación para cada punto interior de la malla, donde los puntos exteriores estarán fijados por las condiciones de contorno. 


- La solución de el conjunto de todas las ecuaciones nos da el valor de $\phi(x,y)$ en cada uno de los puntos de la malla, por lo que hemos transformado el problema una ecuación diferencial lineal en derivadas parciales en un sistema lineal de ecuaciones simultaneas.


- Este sistema se puede resolver por el método de Gauss o por métodos iterativos.


- Sin embargo, ya que las ecuaciones son particularmente simples el método más rápido para resolver el sistema es el del punto fijo o relajación. 


- Para ello escribimos nuestra ecuación como

    $$\phi(x,y)=\frac{1}{4}\left(\phi(x+a,y)+\phi(x-a,y)+\phi(x,y+a)+\phi(x,y-a)\right),$$

    y partimos de una estimación inicial, por ejemplo que el potencial es cero en todos los puntos interiores.


- Con la estimación inicial, podemos obtener una nueva estimación de nuestro potencial $\phi'(x,y)$ iterando nuestra ecuación, 

    $$\phi'(x,y)=\frac{1}{4}\left(\phi(x+a,y)+\phi(x-a,y)+\phi(x,y+a)+\phi(x,y-a)\right).$$


- Este método, resolver la ecuación de Laplace usando la técnica del punto fijo, es conocido como método de Jacobi.


- Como vimos en el Tema 5, el método del punto fijo no es siempre estable, es decir, no siempre converge hacía la solución, sin embargo, puede probarse que siempre converge para la ecuación de Laplace. 


- Por tanto, sólo es necesario definir una tolerancia que nos permite definir cuando la diferencia entre dos iteraciones sucesivas es suficientemente pequeña como para parar el algoritmo. 

### 8.1.1.1 Algoritmo del método de Jacobi

1. Fijamos los valores de $\phi(x,y)$ en la frontera.


2. Tomamos una hipótesis de partida para los valores de $\phi(x,y)$ para los puntos en el interior de la región que queremos estudiar (la hipótesis no tiene porque ser muy buena).


3. Obtenemos una nueva estimación de los puntos del interior aplicando la ecuación de Laplace discretizada

    $$\phi'(x,y)=\frac{1}{4}\left(\phi(x+a,y)+\phi(x-a,y)+\phi(x,y+a)+\phi(x,y-a)\right).$$


4. Repetimos el punto 3 hasta que $\phi(x,y)$ no cambie de una iteración a la siguiente dentro de una tolerancia.

### Ejemplo 8.1: solución de la ecuación de Laplace

Calcular la solución del problema del potencial electrostático que hemos descrito para una caja de L=1 m de lado con $V= 1$ voltio y $a = 1$ cm. 


```python
import numpy as np
import matplotlib.pyplot as plt
L=1
V=1
a=0.01
N=L/a
N=int(N)
phi=np.zeros([N+1,N+1],float)
phi[0,:]=V
eps=1e-6
err=1
phip=np.empty([N+1,N+1],float)
phip[0,:]=V
phip[N,:]=0
phip[:,0]=0
phip[:,N]=0
while err>eps:
    phip[1:N,1:N]=(phi[2:N+1,1:N]+phi[0:N-1,1:N]+phi[1:N,2:N+1]+phi[1:N,0:N-1])/4
    err=np.max(abs(phip-phi))
    phi=np.copy(phip)
plt.imshow(phi,origin='lower')
```

## 8.1.2 Haciendo el método de Jacobi más rápido

- El método de jacobi proporciona soluciones correctas, pero tiene el problema de ser muy lento. 


- Hay dos pequeñas modificaciones que permiten hacer el método de Jacobi mucho más rápido. 


- La primera es aplicar el método de la sobrerelajación, que ya hemos aplicado para resolver ecuaciones lineales, no lineales y ecuaciones diferenciales ordinarias. 


- Una vez más, la idea del método de sobrerelajación es alargar el paso entre dos estimaciones para converger a la solución más rápidamente.

### 8.1.2.1 El método SOR o de sobrerelajación

- Dada una estimación de nuestra solución $\phi(x,y)$, la siguiente estimación del método de jacobi vendrá dada por

    $$\phi'(x,y)=\phi(x,y)+\Delta \phi(x,y).$$

    siendo $\Delta\phi(x,y)=\phi'(x,y)-\phi(x,y)$.


- Sin embargo, aplicando el método de la sobrerelajación el nuevo valor será

    $$\phi_\omega(x,y)=\phi(x,y)+(1+\omega)\Delta \phi(x,y),\quad \omega>0,$$

    que nos permite cambiar $\phi(x,y)$ una cantidad un poco mayor de la que dictamina el método de Jacobi.


- Teniendo en cuenta que $\Delta\phi(x,y)=\phi'(x,y)-\phi(x,y)$ y usando el método de Jacobi para la ecuación de Laplace, obtenemos

    $$\phi_\omega(x,y)=\phi(x,y)+(1+\omega)\left[\phi'(x,y)-\phi(x,y)\right]=\frac{1+\omega}{4}\left[\phi(x+a,y)+\phi(x-a,y)+\phi(x,y+a)+\phi(x,y-a)\right]-\omega\,\phi(x,y).$$


- Al igual que para la solución de ecuaciones lineales, no lineales o diferenciales ordinarias, el valor de $\omega$ ha de ser cogido cuidadosamente para garantizar la convergencia del método. 

### 8.1.2.2 El método de Gauss-Seidel

- Es el mismo método que utilizamos para resolver sistemas de ecuaciones lineales. 


- Al resolver la ecuación de Laplace vamos calculando los nuevos valores de nuestra solución $\phi(x,y)$ uno a uno, a lo largo de cada una de las filas y columnas de nuestra array,


```python
from IPython.display import Image,display
display(Image("grid-GS.png",width=300))
```

$\quad$ por lo que, en cada nuevo punto, ya hemos obtenido la nueva estimación para dos de sus vecinos.

- La idea detrás del método de Gauss-Seidel es actualizar directamente en una misma estimación los valores que ya hemos calculado:

    $$\phi'(x,y)\leftarrow\frac{1}{4}\left(\phi(x+a,y)+\phi(x-a,y)+\phi(x,y+a)+\phi(x,y-a)\right),$$

    para el método de jacobi o 

    $$\phi_\omega(x,y)\leftarrow\frac{1+\omega}{4}\left[\phi(x+a,y)+\phi(x-a,y)+\phi(x,y+a)+\phi(x,y-a)\right]-\omega\,\phi(x,y),$$

    para el método de Jacobi sobrerelajado.


- La flecha significa que calculamos la parte de la derecha y lo remplazamos directamente el valor de la izquierda, es decir no tenemos que definir una nueva array para calcular la nueva iteración. 


- La primera ventaja del método de Gauss-Seidel no es que aumente su velocidad (el método de sobrerelajación la aumenta mucho más), sino que nos permite ahorrar también memoria, sólo necesitamos una array en vez de dos. 


- Además, la combinación del método de Jacobi sobrerelajado con Gauss-Seidel garantiza que el método sea estable, cosa que en general no es cierta para el método sobrerelajado. 


- Una vez más, el valor óptimo de $\omega$ dependerá de cada caso, pero en general es estable para $\omega <1$ y para el caso de la ecuación de Laplace $\omega\sim 0.9$.

```python
L=1
V=1
a=0.01
N=L/a
N=int(N)
phi=np.zeros([N+1,N+1],float)
phi[0,:]=V
eps=1e-6
err=1
phip=np.empty([N+1,N+1],float)
phip[0,:]=V
phip[N,:]=0
phip[:,0]=0
phip[:,N]=0
w=0.9
while err>eps:
    for i in range(1,N):
        for j in range(1,N):
            phip=np.copy(phi)
            phi[i,j]=0.25*(1+w)*(phi[i-1,j]+phi[i+1,j]+phi[i,j-1]+phi[i,j+1])-w*phi[i,j]
    err=np.max(abs(phip-phi))
    
plt.imshow(phi)
```

# 8.2: Ecuaciones con condiciones iniciales

- Aunque los problemas con condiciones de contorno son habituales en física, es común que cuando estudiamos un problema donde aparece el tiempo, sólo conozcamos el valor de nuestras variables en un momento $t_0$ y queremos estudiar como evolucionará nuestro sistema para $t>t_0$.


- Estos tipos de problemas están asociado por tanto a condiciones iniciales, y en general, aparecen en ecuaciones diferenciales semilineales parabólicas o hiperbólicas. 


- Un buen ejemplo de este caso es la ecuación de la difusión o del calor que en una dimensión viene dada por

    $$\frac{\partial \phi(x,t)}{\partial t}=D \frac{\partial^2\phi(x,t)}{\partial x^2},$$

    siendo $D$ el coeficiente de difusión, que determina el movimiento de difusión de gases y líquidos, o el flujo de calor en conductores térmicos.


- Si bien es posible que conozcamos condiciones de contorno para la variable $x$, por ejemplo $\phi(0,t)=\phi(L,t)=0$, en general sólo conoceremos las condiciones iniciales de nuestra solución en $t$, y por tanto no podemos usar el método de relajación.

## 8.2.1 Método FTCS (Forward Time Centered-Space)

- Una vez más, comenzamos discretizando nuestras coordenadas espaciales en un conjunto de puntos, y aproximamos la derivada segunda por medio del método de las diferencias finitas

    $$\frac{\partial^2\phi(x,t)}{\partial x^2}=\frac{\phi(x+a,t)-2\phi(x,t)+\phi(x-a,t)}{a^2}$$

    con lo que la ecuación de la difusión viene dada por

    $$\frac{\partial \phi(x,t)}{\partial t}=\frac{D}{a^2}\left[\phi(x+a,t)+\phi(x-a,t)-2\phi(x,t)\right].$$


- Si consideramos los valores de $\phi$ en los diferentes puntos de la malla como variable independientes, tenemos un sistemas de ecuaciones diferenciales ordinarias que podemos resolver por los métodos numéricos que hemos visto en el Tema 7.


- El problema es que dependiendo de como fina sea la partición, o del número de dimensiones de nuestro problema, nos podemos encontrar con un número muy elevado de ecuaciones diferenciales ordinarias.


- Como la aproximación que hemos hecho al usar diferencias finitas en la parte espacial no es muy precisa, no merece la pena utilizar métodos muy precisos para integrar la ecuaciones diferenciales ordinarias: utilizamos por tanto el método de Euler.


- Es decir, definiendo nuestra ecuación discretizada de la difusión como

    $$\frac{\partial \phi(x,t)}{\partial t}=f(\phi(x,t),t),$$

    nuestra solución vendrá dada por

    $$\phi(x,t+h)=\phi(x,t)+h\frac{\partial \phi(x,t)}{\partial t}=\phi(x,t)+f\left(\phi(x,t),t\right)=\phi(x,t)+h\frac{D}{a^2}\left[\phi(x+a,t)+\phi(x-a,t)-2\phi(x,t)\right].$$


- Si conocemos el valor de $\phi(x,t_0)$ en algún tiempo inicial $t=t_0$, entonces esta ecuación nos permite obtener nuestra solución en un tiempo $t=t_0+h$, que, aplicándolo de forma iterativa, nos permite obtener la solución del problema en todo nuestro intervalo temporal.

## 8.2.2 Estabilidad de Von Neumann para ecuaciones lineales

- Para entender porque el método FTCS esta fallando con la ecuación de ondas volvamos primero a la ecuación de difusión. 


- Ya hemos visto que la solución usando el método FTCS viene dada por la ecuación

    $$\phi(x,t+h)=\phi(x,t)+h\frac{D}{a^2}\left[\phi(x+a,t)+\phi(x-a,t)-2\phi(x,t)\right].$$


- En un instante de tiempo determinado, una onda es una superposición de ondas periódicas, y como se puede comprobar en el Tema 6, una función periódica siempre puede escribirse como una serie de Fourier. Es decir, a través de la expansión

    $$\phi(x,t)=\sum_{k}c_k(t)\text{e}^{ikx}.$$


- Por tanto, estudiar la evolución de nuestra solución es equivalente a estudiar su desarrollo de Fourier. 


- Sin embargo, como la ecuación es lineal, podemos estudiarlo analizando la evolución de un único modo de Fourier, es decir, cogiendo el término $\phi_k(x,t)=c_k(t)\exp{ikx}$. 


- Introduciendo nuestro modo el la ecuación FTCS obtenemos que

    \begin{align}
    \phi_k(x,t+h)&=c_k(t) \text{e}^{ikx}+h\frac{D}{a^2}c_k(t)\left[\text{e}^{ik(x+a)}+\text{e}^{ik(x-a)}-2\text{e}^{ikx}\right]\nonumber\\
    &=\left[1+h\frac{D}{a^2}\left(\text{e}^{ik a}+\text{e}^{-ik a}-2\right)\right]c_k(t)\text{e}^{ikx}\nonumber\\
    &=\left[1-h\frac{4D}{a^2}\sin^2\left(\frac{1}{2}ka\right)\right]c_k(t)\text{e}^{ikx}.\nonumber
    \end{align}

    que implica que cada modo de Fourier en nuestra solución evoluciona de forma independiente de los otros, o lo que es lo mismo

    $$c_k(t+h)=\left[1-h\frac{4D}{a^2}\sin^2\left(\frac{1}{2}ka\right)\right]c_k(t).$$


- El factor entre corchetes nos explica por tanto como va a evolucionar nuestra solución con el tiempo dado un valor inicial. En particular si

    $$\left\vert\,1-h\frac{4D}{a^2}\sin^2\left(\frac{1}{2}ka\right)\,\right\vert>1,$$

    nuestra solución ira aumentando iterativamente con el tiempo y eventualmente divergirá. 


- En nuestro caso, el factor es siempre menor que uno, sin embargo, sólo será mayor que menos uno si

    $$h\frac{4D}{a^2}\sin^2\left(\frac{1}{2}ka\right)<2\,\Longrightarrow\,h\le h_0=\frac{a^2}{2D}.$$


- Si $h>h_0$, nuestras soluciones divergirán, lo cual es un comportamiento no físico (ningún objeto evoluciona libremente para acabar teniendo temperatura infinita).


- Por el contrario si $h\le h_0$, todos los coeficientes de Fourier decrecerán de forma exponencial excepto aquel con $k=0$ que valdrá $c_0(t+h)=c_0(t)$, que es exactamente el comportamiento que esperamos encontrar en el fenómeno de la difusión.

### 8.2.2.1 Estabilidad de Von Neumann para la ecuación de ondas

- Apliquemos ahora el análisis de Von Neumann a la ecuación de ondas. 


- En este caso, al ser una ecuación de orden dos para la variable temporal, tendremos un desarrollo de Fourier para la función de ondas $\phi(x,t)$ y su derivada $\psi(x,t)$, por lo que, cogiendo un único modo

$$\left(\begin{array}{c}\phi_k(x,t)\\\psi_k(x,t)\end{array}\right)=\left(\begin{array}{c}c_k^\phi(t)\\c_k^{\psi}(t)\end{array}\right)\text{e}^{ikx}.$$


- Introduciendo ahora su contribución en la ecuación FTCS se obtiene el sistema

    \begin{align}
    c_k^\phi(t+h)=&c_k^\phi(t)+hc_k^\psi(t),\nonumber\\
    c_k^\psi(t+h)=&c_k^\psi(t)-hc_k^\phi(t)\frac{4v^2}{a^2}\sin^2\left(\frac{1}{2}ka\right),\nonumber
    \end{align}

    o en forma vectorial

    $${\bf c_k}(t+h)={\bf A}{\bf c_k}(t),$$

    siendo ${\bf c_k}=(c_k^\phi,c_k^\psi)$ y 

    $${\bf A}=\left(\begin{array}{cc}1&h\\-hr^2&1\end{array}\right),\quad\text{con}\quad r=\frac{2v}{a}\sin\left(\frac{1}{2}ka\right).$$


- Expresando ahora ${\bf c_k}$ en función de los autovectores de la matriz A, $\bf v_i$, nuestro sistema se puede escribir como

    $${\bf c_k}(t+h)={\bf A}\left({\alpha_1\bf v_1}(t)+{\alpha_2\bf v_2}(t)\right)={\alpha_1\lambda_1\bf v_1}(t)+{\alpha_2\lambda_2\bf v_2}(t),$$

    siendo $\lambda_1$ y $\lambda_2$ los autovalores del sistema. 


- Por tanto, en el segundo paso temporal tendremos

    $${\bf c_k}(t+2h)={\alpha_1\lambda_1^2\bf v_1}(t)+{\alpha_2\lambda_2^2\bf v_2}(t),$$

    y después de $m$ pasos 

    $${\bf c_k}(t+mh)={\alpha_1\lambda_1^m\bf v_1}(t)+{\alpha_2\lambda_2^m\bf v_2}(t).$$


- Si cualquiera de los autovalores cumple $|\lambda_i>1|$, la solución será divergente. 


- En nuestros casos, los autovalores vienen dado de obtener la solución 

    $$\det({\bf A}-\lambda_i\mathbb 1)=0\,\Rightarrow\,\left(1-\lambda_i\right)^2+h^2r^2=0\,\Rightarrow\, \lambda_i=1\pm ihr$$


- Los autovalores son complejos y su modulo es en ambos casos

    $$\vert \lambda_i\vert=\sqrt{1+\frac{4h^2v^2}{a^2}\sin^2\left(\frac{1}{2}ka\right)}>1.$$


- El módulo de los autovalores es siempre mayor que 1 y por tanto, el método FTCS nunca va a proporcionar una solución estable para la ecuación de ondas (excepto para intervalos de tiempo muy cortos).


## 8.2.3 El método de Euler implícito

- Una forma de intentar remediar este problema es aplicar el método FTCS pero con un paso negativo $h\to-h$.


- En ese caso, las ecuaciones del método FTCS para las ecuaciones de onda son:

    \begin{align}
    \phi(x,t-h)=&\phi(x,t)-h\psi(x,t),\nonumber\\
    \psi(x,t-h)=&\psi(x,t)-h\frac{v^2}{a^2}\left[\phi(x+a,t)+\phi(x-a,t)-2\phi(x,t)\right].
    \end{align}

    o, en un nuevo tiempo $t\to t+h$

    \begin{align}
    \phi(x,t+h)-h\psi(x,t+h)=&\,\phi(x,t),\nonumber\\
    \psi(x,t+h)-h\frac{v^2}{a^2}\left[\phi(x+a,t+h)+\phi(x-a,t+h)-2\phi(x,t+h)\right]=&\,\psi(x,t).
    \end{align}


- Estas ecuaciones determinan $\phi$ y $\psi$ en $t+h$ de forma implícita. 


- Por tanto, para obtener la solución hay que resolver sistema de ecuaciones lineales, lo cual es conocido como **el método de Euler implícito**. 


- De forma vectorial podemos expresar las ecuaciones como

    $${\bf B\,c}(t+h)={\bf c}(t),$$

    siendo $B$ la matriz 

    $${\bf B}=\left(\begin{array}{cc}1&-h\\hr^2&1\end{array}\right),\quad\text{con}\quad r=\frac{2v}{a}\sin\left(\frac{1}{2}ka\right).$$


- Por tanto nuestra nueva solución vendrá dada por la inversa de ${\bf B}$

    $${\bf c}(t+h)={\bf B^{-1}\,c}(t),$$

    donde

    $${\bf B^{-1}}=\frac{1}{1+h^2r^2}\left(\begin{array}{cc}1&h\\-hr^2&1\end{array}\right).$$


- Los autovalores serán ahora por tanto 

    $$\lambda_i=\frac{1\pm ihr}{1+h^2r^2},$$

    cuyo módulo  viene dado ahora por

    $$\vert\lambda_i\vert=\frac{1}{\sqrt{1+h^2r^2}}<1,$$

    y que por tanto garantiza que el método es numéricamente estable para cualquier valor de $h$,  **incondicionalmente estable**.


- Sin embargo la solución tampoco es físicamente aceptable.   
  Todos los coeficientes de Fourier excepto  el modo $k=0$ que es constante tenderán a cero exponencialmente, lo que implica que con el método implícito cualquier onda se extinguirá al cambo de un determinado tiempo. 

## 8.2.4 El método de Crank-Nicolson

- Mientras que el método FTCS produce un crecimiento exponencial para la solución de la ecuación de ondas, el método implícito da un comportamiento de decaimiento exponencial.  


-  El método de Crank-Nicolson combina ambos métodos tomando el promedio de sus soluciones. Es decir, escribiendo una vez más la solución para la ecuación de ondas del método FTCS

    \begin{align}
    \phi(x,t+h)=&\phi(x,t)+h\psi(x,t),\nonumber\\
    \psi(x,t+h)=&\psi(x,t)+h\frac{v^2}{a^2}\left[\phi(x+a,t)+\phi(x-a,t)-2\phi(x,t)\right].
    \end{align}

    y la del método implícito

    \begin{align}
    \phi(x,t+h)-h\psi(x,t+h)=&\,\phi(x,t),\nonumber\\
    \psi(x,t+h)-h\frac{v^2}{a^2}\left[\phi(x+a,t+h)+\phi(x-a,t+h)-2\phi(x,t+h)\right]=&\,\psi(x,t).
    \end{align}

    cogiendo el promedio de ambas soluciones para la primera y segunda ecuación se obtiene

    \begin{align}
    \phi(x,t+h)-\frac{h}{2}\psi(x,t+h)=&\,\phi(x,t)+\frac{1}{2}h\psi(x,t),\nonumber\\
        \psi(x,t+h)-h\frac{v^2}{2a^2}\left[\phi(x+a,t+h)+\phi(x-a,t+h)-2\phi(x,t+h)\right]=&\,\psi(x,t-h)+h\frac{v^2}{2a^2}\left[\phi(x+a,t)+\phi(x-a,t)-2\phi(x,t)\right],
    \end{align}

    que también son ecuaciones implícitas y de forma vectorial se pueden escribir como

    $${\bf B\,c}(t+h)={\bf A c}(t),$$

    o tomando la inversa de $\bf B$

    $${\bf c}(t+h)={\bf B^{-1}A\,c}(t),$$

    donde

    $${\bf B^{-1}A}=\frac{1}{1+h^2r^2}\left(\begin{array}{cc}1&h\\-hr^2&1\end{array}\right)\left(\begin{array}{cc}1&h\\-hr^2&1\end{array}\right)=\frac{1}{1+h^2r^2}\left(\begin{array}{cc}1-h^2r^2&2h\\-2hr^2&1-h^2r^2\end{array}\right),$$

    cuyos autovalores son

    $$\lambda_i=\frac{1-h^2r^2\pm 2ihr}{1+h^2r^2},$$

    con módulo

    $$\vert\lambda\vert=\frac{\sqrt{\left(1-h^2r^2+ 2ihr\right)\left(1-h^2r^2- 2ihr\right)}}{1+h^2r^2}=1.$$


- El método de Crank-Nicolson se encuentra por tanto en la frontera entre el método implícito y el método FTCS, y por tanto sus soluciones ni crecen exponencialmente ni se apagan inevitablemente, permite reproducir el comportamiento de las ondas. 

### 8.2.4.1 El método de Crank-Nicolson y la ecuación de Schrödinger

- La ecuación de Schrödinger es posiblemente una de las ecuación en derivadas parciales más importantes en Física. 


- Vamos a intentar resolverla en una sóla dimensión puesto que el problema en dos o tres dimensiones es esencialmente igual pero conlleva mucho más tiempo de ejecución. 


- La propagación libre de una partícula microscópica de masa $m$ viene dada por

    $$-\frac{\hbar^2}{2m}\frac{\partial^2\psi(x,t)}{\partial x^2}=i\hbar\frac{\partial \psi(x,t)}{\partial t}.$$


- Para simplificar el problema asumamos que la partícula se encuentra confinada en una caja con paredes de potencial infinito en $x=0$ y $x=L$.


- Sustituyendo la segunda derivada espacial por su aproximación usando diferencias finitas, y usando el método de Euler para la derivada temporal obtenemos las ecuaciones FTCS para la ecuación de Schrödinger

    $$\psi(x,t+h)=\psi(x,t)+h\frac{i\hbar}{2ma^2}\left[\,\psi(x+a,t)+\psi(x-a,t)-2\psi(x,t)\,\right],$$

    siendo $a$ el espaciado en el retículo espacial y $h$ en el temporal.


- Aplicando el mismo método pero con un paso inverso -h obtenemos por otro lado las ecuaciones del método implícito

    $$\psi(x,t)=\psi(x,t+h)-h\frac{i\hbar}{2ma^2}\left[\,\psi(x+a,t+h)+\psi(x-a,t+h)-2\psi(x,t+h)\,\right].$$


- Finalmente, tomando la media de ambas estimaciones obtenemos las ecuaciones del método de Crank-Nicolson

    $$\psi(x,t+h)-\frac{h}{2}\frac{i\hbar}{2ma^2}\left[\,\psi(x+a,t+h)+\psi(x-a,t+h)-2\psi(x,t+h)\,\right]=\psi(x,t)+\frac{h}{2}\frac{i\hbar}{2ma^2}\left[\,\psi(x+a,t)+\psi(x-a,t)-2\psi(x,t)\,\right],$$

    que nos proporciona un conjunto de ecuaciones simultaneas para cada punto de nuestro retículo. 


- Las condiciones de contorno para la variable espacial implican que $\psi(0,t)=\psi(L,t)=0$, por lo que tenemos que resolver nuestras ecuaciones para cada punto interior $\psi(na,t)$ con $0<n<L/a$.


- Representemos todo ese conjunto de puntos espaciales en un vector

    $$\psi(t)=\left(\begin{array}{c}\psi(a,t)\\\psi(2a,t)\\\vdots\\\psi(L/a-1,t)\end{array}\right),$$

    de forma que podamos escribir la ecuación de Schrödinger vectorialmente como

    $${\bf A \psi}(t+h)={\bf B\psi}(t),$$

    donde las matrices ${\bf A}$ y ${\bf B}$ vienen dadas por 

    $${\bf A}=\left(\begin{array}{ccccc}a_1&a_2&&&\\a_2&a_1&a_2&&\\&a_2&a_1&a_2&\\&&a_2&a_1&\\&&&&\ddots\end{array}\right),\qquad{\bf B}=\left(\begin{array}{ccccc}b_1&b_2&&&\\b_2&b_1&b_2&&\\&b_2&b_1&b_2&\\&&b_2&b_1&\\&&&&\ddots\end{array}\right),$$

    que son simétricas y tridiagonales y cuyos coeficientes son

    $$a_1=1+h\frac{i\hbar}{2ma^2},\quad a_2=-h\frac{i\hbar}{4ma^2},\quad b_1=1-h\frac{i\hbar}{2ma^2},\quad b_2=h\frac{i\hbar}{4ma^2}.$$

##  8.2.5 El método Espectral o de la Transforma de Fourier

- Consideremos otra vez la ecuación de ondas para una cuerda vibrante de longitud $L$ 

    $$\frac{\partial^2\phi(x,t)}{\partial x^2}-\frac{1}{v^2}\frac{\partial^2\phi(x,t)}{\partial t^2}=0.$$

    con los extremos fijos, es decir $\phi(0,t)=\phi(L,t)=0$.


- Sabemos que un ansatz de la solución es

    $$\phi_k(x,t)=\Re\left[\sin\left(\frac{\pi k x}{L}\right)\text{e}^{i\omega t}\right].$$


- Efectivamente, introduciéndolo en la ecuación de ondas se obtiene

    $$\left[-\frac{\pi^2k^2}{L^2}+\frac{\omega^2}{v^2}\right]\phi_k(x,t)=0,$$
    
    por lo que nuestro ansatz satisfará la ecuación si 

    $$\omega=\frac{\pi k v}{L}.$$


- Para resolver nuestra ecuación, discreticemos una vez nuestra variable espacial en $N$ intervalos iguales, que en el intervalo $[0,L]$ definen los puntos

    $$x_n=\frac{n}{N}L$$

    por lo que el valor de nuestro ansatz en dichos puntos vendrá dada por

    $$\phi_k(x_n,t)=\Re\left[\sin\left(\frac{\pi k n}{N}\right)\exp\left(i\frac{\pi v k  t}{L}\right)\right].$$


- Ahora bien, como nuestra ecuación es lineal en $\phi(x,t)$ cualquier combinación lineal de nuestro ansatz para diferentes valores de $k$ también será solución de la ecuación de ondas. 


- Dada la periodicidad de la función seno, eso implica valores de $k\in[1, N-1]$ que permiten obtener la solución no trivial, por lo que, de forma general la combinación

    $$\phi(x_n,t)=\frac{1}{N}\Re\left[\sum_{k=1}^{N-1}b_k\sin\left(\frac{\pi k n}{N}\right)\exp\left(i\frac{\pi v k  t}{L}\right)\right],$$

    será una solución de nuestra ecuación para una combinación dada de los coeficientes 

    $$b_k=\alpha_k+i\eta_k,$$

    que en general pueden ser complejos. 


- En particular para $t=0$ nuestra solución se reduce a 

    $$\phi(x_n,0)=\frac{1}{N}\Re\left[\sum_{k=1}^{N-1}b_k\sin\left(\frac{\pi k n}{N}\right)\right]=\frac{1}{N}\sum_{k=1}^{N-1}\alpha_k\sin\left(\frac{\pi k n}{N}\right),$$

    que no es más que una expansión de Fourier en senos. 


- Por tanto, los coeficientes $\alpha_k$ de nuestra solución vienen completamente fijados por la serie de Fourier seno discreta, es decir, por los valores de $\phi(x_n,0)$, que es nuestra condición inicial.


- Como la ecuación de ondas es una ecuación diferencial de orden 2 en la variable temporal, su solución también requiere fijar una condición inicial para su derivada. 


- Tomando la derivada de nuestra solución y evaluándola en el instante inicial encontramos que

    $$\left.\frac{\phi(x_n,t)}{\partial t}\;\right\vert_{t=0}=\frac{1}{N}\frac{\pi v}{L}\Re\left[\sum_{k=1}^{N-1}i k b_k\sin\left(\frac{\pi k n}{N}\right)\right]=-\frac{1}{N}\frac{\pi v}{L}\sum_{k=1}^{N-1}k\eta_k\sin\left(\frac{\pi k n}{N}\right),$$

    que nos permite de nuevo fijar los valores $\eta_k$ haciendo la transformada de Fourier seno discreta a la derivada de la solución en el instante inicial. 


- Por tanto, con las condiciones iniciales, podemos obtener exactamente nuestra solución en cualquier momento. 

    $$\phi(x_n,t)=\frac{1}{N}\sum_{k=1}^{N-1}\left[\alpha_k\cos\left(\frac{\pi v k  t}{L}\right)-\eta_k \sin\left(\frac{\pi v k  t}{L}\right)\right]\sin\left(\frac{\pi k n}{N}\right),$$

    que no es más que otra serie de senos. 


- Por tanto, teniendo en cuenta que en cada instante conocemos el valor de los términos entre corchetes para cada elemento de la suma, podemos obtener nuestra solución en cada instante de tiempo realizando la transformada de Fourier inversa discreta seno. Es el llamado método espectral. 


- Una diferencia con el método de Crank-Nicloson es que el método espectral nos da la solución que queremos en cada instante de tiempo, es decir, no hace falta iterar el método de Euler paso a paso en la variable temporal para saber cuanto vale solución en un determinado momento. 


- Para un único paso, el método espectral es más lento que el método de Crank-Nicolson (la transformada de Fourier inversa requiere $N\log N$ operaciones mientras que el método FTCS sólo requiere hacer N), sin embargo, se vuelve más rápido cuando queremos evolucionar el sistema a grandes escalas. 


- Una limitación adicional es que el método espectral sólo funciona para ecuaciones lineales con condiciones de contorno sencillas.

### 8.2.5.1 El método espectral y la ecuación de Schrödinger

- Vamos a volver a estudiar la solución de la ecuación de Schrödinger usando el método espectral.


- Una vez más, vamos a considerar una partícula de masa $m$ en una caja de longitud $L$ con paredes impenetrables, cuya función de onda viene dada por la solución de

    $$-\frac{\hbar^2}{2m}\frac{\partial^2\psi(x,t)}{\partial x^2}=i\hbar \frac{\partial \psi(x,t)}{\partial t}.$$


- La función de onda tiene que ir a cero en las paredes de la caja, por lo que una posible solución de la ecuación es

    $$\phi_k(x,t)=\sin\left(\frac{\pi k x}{L}\right)\text{e}^{i E t/\hbar},$$

    donde el valor de la energía se puede determinar introduciendo la solución en la ecuación de Schrödinger que impone la condición

    $$E=\frac{\pi^2\hbar^2 k^2}{2mL^2}.$$


- La solución general, por tanto, se podrá escribir como una combinación de estas soluciones individuales, por lo que, para un retículo de $x_n=nL/N$ puntos, tendremos

    $$\phi(x_n,t)=\frac{1}{N}\sum_{k=1}^{N-1}b_k\sin\left(\frac{\pi k n}{N}\right)\exp\left(i\frac{\pi^2\hbar k^2}{2mL^2}t\right).$$


- Al ser una ecuación de primer orden con respecto al tiempo, sólo necesitamos una condición inicial para resolver la ecuación diferencial, que implica

    $$\phi(x_n,0)=\frac{1}{N}\sum_{k=1}^{N-1}b_k\sin\left(\frac{\pi k n}{N}\right),$$

    por lo que la parte real e imaginaria de los coeficientes $b_k=\alpha_k+i\eta_k$ se pueden determinar aplicando la transformada de Fourier discreta seno a la  
$\quad$ parte real e imaginaria de la condición inicial para la función de onda.


- Determinados de esta forma los coeficientes, la solución de la ecuación de Schrödinger para cualquier instante de tiempo vendrá dada por

    $$\Re\phi(x_n,t)=\frac{1}{N}\sum_{k=1}^{N-1}\left[\alpha_k\cos\left(\frac{\pi^2\hbar k^2}{2mL^2}t\right)-\eta_k\sin\left(\frac{\pi^2\hbar k^2}{2mL^2}t\right)\right]\sin\left(\frac{\pi k n}{N}\right),$$

    y

    $$\Im\phi(x_n,t)=\frac{1}{N}\sum_{k=1}^{N-1}\left[\alpha_k\sin\left(\frac{\pi^2\hbar k^2}{2mL^2}t\right)+\eta_k\cos\left(\frac{\pi^2\hbar k^2}{2mL^2}t\right)\right]\sin\left(\frac{\pi k n}{N}\right),$$

    por lo que la solución de nuestra función de onda en cada instante se puede obtener aplicando la transformada de Fourier inversa seno.

## Ejercicio 8.4: La ecuación de Gross–Pitaevskii y los condensados de Bose-Einstein (Examen de Enero de 2023)

Un condensado de **Bose-Einstein** es un estado de la materia que se forma típicamente cuando un gas de bosones a densidades muy bajas se enfría (la llamada **trampa de potencial**) a temperaturas muy cercanas al cero absoluto.

En estas condiciones, y por debajo de una **temperatura crítica $\text{T}_c$**, una gran fracción de estos bosones se encuentran en su estado fundamental (**condensan**), haciendo que fenómenos cuánticos microscopios, tal como la interferencia de su función de onda, se conviertan en **fenómenos macroscópicos**.  

La ecuación de **Gross-Pitaevskii (GP)** es una **ecuación de Schrödinger no lineal** que describe la evolución de funciones de onda macroscópicas, y por tanto, nos permite describir las propiedades de un condensado de Bose-Einstein a temperaturas $T$ muy por debajo de la temperatura crítica de condensación $T_c$.


La ecuación GP incluye: 
<br>

* Un término para la **trampa de potencial** $V(r)$.
* La interacción de **campo medio entre los átomos** en el gas, que se manifiesta como un **término no lineal**. 
* **Interacciones** entre los bosones que se tienen en cuenta mediante una **constante $g$** que puede ser positiva (focalización) o negativa (desfocalización).

    Es decir:

    $$i\hbar\,\frac{\partial\psi({\bf r},t)}{\partial t}=\left(-\frac{\hbar^2}{2m}\nabla^2+V({\bf r})+g\,\vert\psi({\bf r},t)\vert^2\right)\psi({\bf r},t),$$    
    
    donde $\psi({\bf r},t)$ es la **función de onda macroscópica del condensado**.

    $${\bf A(t+h)\, \psi}(t+h)={\bf B(t)\,\psi}(t),$$

Remarcar que los indices $i, j, k, \dots$ hacen referencias a cada una de las componentes del vector $\vec{x}$ una vez discretizado el espacio, por ejemplo, en 2 dimensiones: $\vec{x}=(x,y)$, de forma que un punto del espacio (después de discretizar) puede identificarse como $\vec{x}_{i,j}=(x_i,y_j)$. ## Ejercicio 8.10: ecuaciones de Euler para un fluido no viscoso


Las ecuaciones de Euler gobiernan la dinámica del movimiento de un fluido adiabático no viscoso. En el límite de Froude (sin fuerzas externas), en una dimensión, con densidad $\rho$, velocidad $u$, energía total $E$ y presión $p$, están dadas en forma adimensional como

\begin{align}
\frac{\partial \rho}{\partial t}+\frac{\partial\,(\rho\,u)}{\partial x}=&\;0,\\
\frac{\partial\,(\rho\,u)}{\partial t}+\frac{\partial\,(\rho\,u^2+p)}{\partial x}=&\;0,\\
\frac{\partial\,(\rho\,E)}{\partial t}+\frac{\partial\,(u\cdot (E+p))}{\partial x}=&\;0,
\end{align}

que describen la conservación de masa, momento y energía del fluido. 

Para un gas ideal la energía del fluido sigue la ecuación:

$$E=\frac{1}{2}\rho\,u^2+\frac{p}{\gamma-1},$$

donde $\gamma$ representa el calor especifico del fluido. En nuestro caso, asumiendo que corresponde al aire, $\gamma=1.4$. 

Usando la relación entre la  presión del fluido como función de su energía $E$, velocidad $u$ y densidad $\rho$, nuestro sistema hiperbólico de EDPs de primer orden puede escribirse como un conjunto de leyes de conservación:

$$\partial_t Q+\partial_x F(Q)=0,\quad Q=\left[\begin{array}{c} 
\rho\\
\rho\,u \\
E\end{array}\right]\quad F(Q)=\left[\begin{array}{c} 
\rho\,u\\
\rho\,u^2+p\\
\rho\,u(E+p) \\
\end{array}\right],$$

siendo $F$ el llamado flujo del vector conservado $Q$. 


En este problema nos vamos a limitar a estudiar la dinámica de nuestro fluido en una sóla dimensión espacial $x$. 


1. Considerando el dominio espacial definido por el intervalo $[x_L,x_R]$ y dos puntos temporales $t_2>t_1$, demostrar que integrando las ecuaciones de Euler en forma diferencial en el espacio y el tiempo, estás adquieren la forma integral,

    $$\int_{x_L}^{x_R}Q(x,t_2)\text{d}x=\int_{x_L}^{x_R}Q(x,t_1)\text{d}x+\int_{t_1}^{t_2}F(Q(x_R,t))\text{d}t-\int_{t_1}^{t_2}F(Q(x_L,t))\text{d}t$$

    Esta relación va a ser la base de nuestras discretizaciones espaciales y temporales. Para simplificar el problema, tomamos como dominio de cálculo el intervalo $[0,1]$ que dividiremos en $N$ celdas iguales de anchura $Δx=1/N$.

    El problema de estudiar un fluido radica en sus condiciones de frontera que a priori no son conocidas. Por ello, para estudiar el sistema vamos a añadir en cada extremo de nuestro intervalo espacial una intervalo extra, las llamadas celdas fantasma, de forma que, considerando $N$ puntos, la discretización del espacio viene dada por el código:
    
        from numpy import linspace
        a=0
        b=1
        N=100
        dx=(b-a)/N

        x=linspace(a-0.5*dx,b+0.5*dx,N+2)
        
Sin embargo, al incluir estás dos celdas fantasmas, será necesario definir el valor del vector de estado $Q$ y su flujo $F$, en el intervalo definido entre dos valor de $x$. 

2. Definiendo $Q(x_i,t_n)=Q^n_i$, demostrar que la media espacial en el intervalo $[x_{i-1/2},x_{i+1/2}]$ en el tiempo $t_n$, viene dado por

    $$Q_i^n=\frac{1}{\Delta_x}\int_{x_{i-1/2}}^{x_{i+1/2}}\,Q(x,t_n)\,\text{d}x,$$
    
   y que, del mismo modo, la media temporal del flujo a través del borde de la celda en el punto $x_{i+1/2}$, definido como $F^n_{i+1/2}$, viene dado por:
   
   $$F_{i+1/2}^n=\frac{1}{\Delta t_n}\int_{t_{n}}^{t_{n+1}}\,F\left(Q(x_{i+1/2},t)\right)\,\text{d}t,$$
   
   donde $\Delta t_n$ es el tamaño temporal del paso $n$.
   
   
3. Probar que insertando estas discretizaciones en la forma integral de las ecuaciones de Euler, obtenemos la ecuación

    $$Q_i^{n+1}=Q_i^n+\frac{\Delta t_n}{\Delta x}\left(F^n_{i+1/2}-F^n_{i-1/2}\right),$$

    y explicar sus similaridades y diferencias con el método FTCS.

Un aspecto fundamental para describir la dinámica de un fluido es elegir de forma correcta el paso temporal, de forma que, dependiendo de la velocidad del fluido, el algoritmo siempre sea estable. Este requerimiento viene dado por la condición de Courant–Friedrichs–Lewy,

$$\Delta t_n= \frac{c\,\Delta x}{\max(\vert u_i^n\vert+a_i^n)},$$

donde $a_i^n$, la velocidad del sonido de un gas ideal, viene dada por

$$a=\frac{\gamma\,p}{\rho},$$

y $c$, el número de Courant–Friedrichs–Lewy (CFL), toma cualquier valor $c<1$. 

El problema ahora, radica en aproximar el valor del flujo en las fronteras entre dos intervalos $F_{i±1/2}$. Para ello, aplicaremos un esquema relativamente sencillo llamado FIrst-ORder CEntred (FORCE); dado el estado de dos intervalos vecinos a la izquierda y derecha, el flujo FORCE en su frontera se calcula como

$$F^n_{\frac{L+R}{2}}=\frac{1}{2}\left(F_0^n+\frac{1}{2}(F^n_{L}+F^{n}_{R})\right)+\frac{1}{4}\frac{\Delta x}{\Delta t_n}\left(Q_{L}-Q_{R}\right),$$

donde $F^n_k=F(Q^n_k)$ y 

$$Q_0^n=\frac{1}{2}\left(Q_L+Q_R\right)+\frac{1}{2}\frac{\Delta t_n}{\Delta x}\left(F_L-F_R\right).$$


Llegados a este punto, se hace evidente la razón de las celdas fantasma introducidas anteriormente. Dado que el interfaz de cada celda requiere información a ambos lados, la primera y la $N$ celda de nuestro dominio carecen de información a la izquierda y la derecha, respectivamente. Al inicializar una celda fantasma en cada lado y transmitir la información desde el interior del dominio hacia el exterior en cada paso temporal, estas celdas ficticias proporcionan la información necesaria para realizar nuestros cálculos.

Dada una condición inicial $Q(x,0)$, queremos evolucionar el sistema en el tiempo para predecir el estado en algún momento futuro $t=T$.
