- Las ecuaciones diferenciales son el problema en física para el cual el uso de ordenadores es más común.


- En este tema nos centraremos en las ecuaciones diferenciales ordinarias. 


- Las ecuaciones en derivadas parciales serán analizadas en el Tema 8. 

# 7.1 Ecuaciones diferenciales de primer orden en una variable

- Son aquellas que sólo tienen una variable independiente, que de forma general asignaremos al tiempo $t$.


- Por ejemplo la ecuación:

    $$\frac{dx}{dt}=\frac{2x}{t}.$$

    es una ecuación diferencial de primer orden.


- Sin embargo, mientras que esta ecuación se puede resolver de forma analítica porque es separable, la ecuación:

    $$\frac{dx}{dt}=\frac{2x}{t}+\frac{x^2}{t^3},$$

    es no separable y además no lineal, lo que de forma general hace que este tipo de ecuaciones no sean resolubles analíticamente.


- Estas son por tanto el tipo de ecuaciones diferenciales que sólo podemos resolver numéricamente en un ordenador, que de forma general expresamos como:

    $$\frac{dx}{dt}=f(x,t),$$

    donde $f(x,t)$, la función dada, dependerá de forma general de $x$, la variable dependiente, y de $t$, la variable independiente.


- La solución de las ecuaciones diferenciales requiere por supuesto fijar las condiciones iniciales/de contorno, que por lo general asumiremos dadas.

## 7.1.1 El método de Euler 

- Es el método más sencillo e intuitivo, pero se usa muy raramente pues el método de Runge-Kutta es casi igual de sencillo y rápido, pero más preciso.


- La idea es, una vez más, usar una serie de Taylor; si queremos saber el valor de $x$ en un tiempo $t+h$ hacemos:

$$x(t+h)=x(t)+h\frac{dx}{dt}+\frac{1}{2}h^2\frac{d^2x}{dt^2}+\cdots=x(t)+h f(x,t)+{\cal O}(h)^2.$$


- Sabiendo el valor de $x$ para un tiempo $t$ esta ecuación nos permite calcular el valor de x un tiempo $x+h$ después. 


- Partiendo por tanto de nuestra condición inicial o de frontera, podemos iterar esta ecuación para obtener $x$ en un conjunto de puntos equiespaciados por una distancia $h$.


- Si el valor de nuestro paso, $h$, es suficientemente pequeño, esto nos permitirá obtener una buena descripción de nuestra solución. 

### Ejemplo 7.1: aplicando el método de Euler

Aplicar el método de Euler para resolver la ecuación:

$$\frac{\text{d}x}{\text{d}t}=-x^3+\sin t,$$

en el intervalo $[0,10]$ usando 1000 puntos equiespaciados con la condición inicial $x(t=0)=0$. 

```python
import numpy as np
import matplotlib.pyplot as plt
#Euler
a=0
b=10
N=1000
def f(x,t):
    return -x**3+np.sin(t)
def euler(a,b,N,f,x0):
    tp=np.linspace(a,b,N)
    h=(b-a)/N
    x=x0
    xp=[]
    for t in tp:
        xp.append(x)
        x+=h*f(x,t)
    return tp,xp
tp,xp=euler(a,b,N,f,0)
plt.plot(tp,xp)
```

### 7.1.1.1 Error del método de Euler

- El error en el método de Euler queda patente a partir del desarrollo en serie de Taylor que hemos utilizado a la hora de construir la solución.


- El error introducido en el paso inicial $t+h$ es:

$$\Delta x(t+h)=\frac{1}{2}h^2\frac{\text{d}^2 x}{\text{d}\,t^2}.$$


- A primera vista podría parecer que es un error de segundo orden es decir, proporcional a $h^2$, 
  sin embargo, es sólo el error asociado a un paso.


- Partiendo de un punto inicial $t=a$ el error en el punto final del intervalo t=b, después de haber recorrido $N=(b-a)/h$ pasos de longitud $h$.  
  el error acumulado en nuestra solución vendrá dado por la suma de los errores individuales de cada paso:

    $$\sum_{k=0}^{N-1}\frac{1}{2}h^2\left(\frac{\text{d}^2x}{\text{d}t^2}\right)_{t=t_k}=\frac{1}{2}h\sum_{k=0}^{N-1}h\left(\frac{\text{d} f(x,t)}{\text{d}t}\right)_{t=t_k}\simeq \frac{1}{2}h\int\limits_{a}^b \frac{\text{d}f(x,t)}{\text{d}t}\text{d}t=\frac{1}{2}h\left(f(x(b),b)-f(x(a),a)\right),$$

    donde hemos aproximado la suma por una integral, lo cual es una buena aproximación si $h$ es pequeño.


- Aunque el error de cada paso es ${\cal O}(h^2)$, el error final es lineal en $h$: para reducir el error por dos tenemos que duplicar el número de pasos.

## 7.1.2 Métodos de Runge-Kutta

- Dada la incertidumbre del método de Euler, uno se puede plantear varias estrategias para aumentar su precisión.


- La primera de ellas es incluir más términos en el desarrollo de Taylor de la solución, lo cual requiere evaluar la derivada de la función. Esto es problemático porque:

    1. En muchas ocasiones no conocemos la derivada de la función. 
    2. Incluso si lo hacemos, trabajar con ella vuelve el método menos eficiente y en muchos casos menos preciso.
   
   
- La segunda alternativa esta asociada a los métodos que vimos para aumentar la precisión de las derivadas numéricas:  
  escoger los puntos en los que evaluamos la solución para que al combinarlos, los términos superiores del desarrollo de Taylor se cancelen.
  
  
- Esta segunda alternativa es la base de los métodos de Runge-Kutta. 

### 7.1.2.1  Método de Runge-Kutta de segundo orden o método del punto medio

- A primer orden, el método de Runge-Kutta es el método de Euler. La derivada de la solución en la serie de Taylor en el punto $t+h$ se evalúa en $t$.


- A segundo orden, el método de Runge-Kutta evalúa la función en el punto que garantiza que el término ${\cal O}(h^2)$ de la serie de Taylor se cancela.


- Es algo que se puede representar mediante la figura:  

```python
from IPython.display import display,Image
display(Image("Runge-kutta.png",width=600))
```

- Mientras que en el método de Euler la derivada de la solución se evalúa en $t$, introduciendo un error proporcional a su segunda derivada,   
  este error ${\cal O}(h^2)$ se puede cancelar si se evalúa en el punto medio.
  
  
- Para verlo repitamos el desarrollo de Taylor en el instante $t+h$ en torno a $t+h/2$:

    $$x\left(t+h\right)=x\left(t+\frac{1}{2}h\right)+\frac{1}{2}h\left(\frac{\text{d}x}{\text{d}t}\right)_{t+h/2}+\frac{1}{8}h^2\left(\frac{\text{d}^2x}{\text{d}t^2}\right)_{t+h/2} +{\cal O}\left(h^3\right),$$

    aplicándoselo de nuevo a la condición inicial:

    $$x\left(t\right)=x\left(t+\frac{1}{2}h\right)-\frac{1}{2}h\left(\frac{\text{d}x}{\text{d}t}\right)_{t+h/2}+\frac{1}{8}h^2\left(\frac{\text{d}^2x}{\text{d}t^2}\right)_{t+h/2} +{\cal O}\left(h^3\right),$$

    por lo que al tomar su diferencia obtenemos:

    $$x\left(t+h\right)=x\left(t\right)+h\left(\frac{\text{d}x}{\text{d}t}\right)_{t+h/2}+{\cal O}\left(h^3\right)=x\left(t\right)+hf\left(x\left(t+h/2\right),t+h/2\right)+{\cal O}\left(h^3\right),$$

    que proporciona una estimación de la solución en el tiempo $t+h$ con un error $\cal O\left(h^3\right)$.


- Sin embargo, esta ecuación requiere evaluar la función en $x\left(t+h/2\right)$, un valor que no disponemos. Para estimarla usamos el método de Euler:

    $$x\left(t+\frac{1}{2}h\right) = x(t)+\frac{1}{2}h\,f(x,t)+ \cal O\left(h^2\right),$$


- De esta forma, el conjunto de ecuaciones necesarias para evaluar el método son:

    \begin{align}
    k_1=&h\,f\left(x,t\right),\nonumber\\
    k_2=&h\,f\left(x+\frac{1}{2}k_1,t+\frac{1}{2}h\right),\nonumber\\
    x(t+h)=&x(t)+k_2.
    \end{align}

    que aplicadas secuencialmente e iterativamente a partir de una condición inicial, nos permiten obtener nuestra solución en todo el intervalo de interés. 


- Sin embargo para obtener nuestro algoritmo hemos aproximado la solución en $t+h/2$ con el método de Euler.   
  Para ver el efecto de tal aproximación, expandamos en serie de Taylor la función en el punto $x'(t)=x(t)+k_1/2$ en torno a a $x(t+h/2)$ en un tiempo $t+h/2$:

    $$f\left(x'(t),t+h/2\right)=f\left(x\left(t+h/2\right),t+h/2\right)+\left[x'(t)-x\left(t+h/2\right)\right]\left(\frac{\partial f}{\partial x}\right)_{\left(x\left(t+h/2\right),t+h/2\right)}+{\cal O}\left(\left[x'(t)-x\left(t+h/2\right)\right]^2\right).$$


- Por otro lado, tenemos que:

    $$x\left(t+h/2\right)=x(t)+h/2 f(x,t)+{\cal O}(h^2)=x(t)+k_1/2+{\cal O}(h^2)=x'(t)+{\cal O}(h^2),$$

    que implica por tanto que 

    $$x'(t)-x\left(t+h/2\right)=\cal O(h^2)$$

    y 

    $$f\left(x'(t),t+h/2\right)=f\left(x\left(t+h/2\right),t+h/2\right)+\cal O(h^2)$$

    que una vez introducida en la solución para $x(t+h)$ proporciona un nuevo término ${\cal O} (h^3)$. 


- El método de Runge-Kutta de segundo orden es para cada paso una aproximación de segundo orden, correcto hasta orden $h^2$ con error $\cal O (h^3)$.


- Por el contrario el método de Euler es de primer orden, correcto para cada paso hasta orden $h$ con error $\cal O (h^2)$.
  

### Ejemplo 7.2: aplicando el método de Runge-Kutta de orden dos

Aplicar el método de Runge-Kutta de orden dos para resolver la ecuación:

$$\frac{\text{d}x}{\text{d}t}=-x^3+\sin t,$$

en el intervalo $[0,10]$ usando 10, 50 y 100 puntos equiespaciados con la condición inicial $x(t=0)=0$.

Compararlo con el resultado obtenido con el método de Euler para 1000 pasos.

```python
def runge_kutta(a,b,N,f,x0):
    tp=np.linspace(a,b,N)
    h=(b-a)/N
    x=x0
    xp=[]
    for t in tp:
        xp.append(x)
        k1=h*f(x,t)
        k2=h*f(x+k1/2,t+h/2)
        x+=k2
    return tp,xp
tp,xp=runge_kutta(a,b,N,f,0)
plt.plot(tp,xp)
```

### 7.1.2.2 Método de Runge-Kutta de orden n

- Veamos como se puede obtener de forma general las ecuaciones del método Runge-Kutta a un orden arbitrario.


- Una vez más, dada la ecuación diferencial

    $$x'=\frac{\text{d}x}{\text{d}t}=f(x,t),$$

    expandimos la solución en serie de Taylor

    $$x(t+h)=x(t)+h\,x'+\frac{1}{2}\,h^2\,x''+\frac{1}{3!}\,h^2\,x^{(3)}+\frac{1}{4!}\,h^4\,x^{(4)}+\cdots,$$

    con

    $$x^{(n)}(t)=\frac{\text{d}^nx(t)}{\text{d}t^n}.$$


- Por otro lado, denotando como:

    $$f^{(n,m)}(x,t)=\frac{\partial^{n+m}f(x,t)}{\partial x^n\,\partial t^m},$$

    tenemos que cada término en el desarrollo de Taylor de nuestra solución puede escribirse como:

    \begin{align}
    x'(t)=&f(x,t),\nonumber\\
    x''(t)=&f^{(0,1)}(x,t)+f^{(1,0)}(x,t)f(x,t),\nonumber\\
    x^{(3)}(t)=&f^{(0,2)}(x,t)+2 f^{(1,1)}(x,t)f(x,t)+f^{(2,0)}(x,t)f(x,t)^2+f^{(1,0)}(x,t)^2f(x,t)+f^{(1,0)}(x,t)f^{(0,1)}(x,t),\nonumber\\
    \vdots\quad=&\qquad\qquad\vdots\qquad\qquad\qquad\;\;\;\,\vdots\qquad\qquad\qquad\quad\vdots\nonumber\\
    \end{align}


- De la misma forma, la solución en un punto $t+h$ también puede obtenerse combinando la evaluación de la función en varios puntos:

    $$x(t+h)=x(t)+\alpha_1\,k_1+\alpha_2\,k_2+\cdots+\alpha_m\,k_m,$$

    donde los valores de $k_i$ vienen dados por:

    \begin{align}
    k_1=&hf\left(x,t\right),\nonumber\\
    k_2=&hf\left(x+\nu_{21}k_1,t+\nu_{21}h\right),\nonumber\\
    k_3=&hf\left(x+\nu_{31}k_1+\nu_{32}k_2,t+\nu_{31}h+\nu_{32}h\right),\nonumber\\
    \vdots=&\qquad\vdots\qquad\qquad\qquad\vdots\nonumber\\
    k_m=&hf\left(x+\sum_{i=1}^{m-1}\nu_{mi}k_i,\sum_{i=1}^{m-1}t+\nu_{mi}h\right),\nonumber\\
    \end{align}

    con $\alpha_i$ y $\nu_{ij}$ parámetros que es necesario determinar.


- Con ello ya tenemos todos los elementos para obtener nuestra aproximación de Runge-Kutta a un orden dado.


- Por ejemplo, para recuperar el resultado de orden 2, tenemos por un lado:

    $$x(t+h)=x(t)+h f(x,t)+\frac{1}{2}h^2\left(f^{(0,1)}(x,t)+f^{(1,0)}(x,t)f(x,t)\right).$$

    y por otro

    $$x(t+h)=x(t)+\alpha_1 k_1+\alpha_2 k_2 =x(t)+\alpha_1 h f(x,t)+\alpha_2 hf\left(x+\nu_{21}k_1,t+\nu_{21}h\right).$$


- Desarrollando ahora $k_2$, tenemos:

    $$k_2=hf(x,t)+h^2 \nu_{21}\left(f^{(0,1)}(x,t)+f^{(1,0)}(x,t)f(x,t)\right),$$

    que nos permite escribir:

$$x(t+h)=x(t)+\left(\alpha_1+\alpha_2\right)hf(x,t)+h^2 \alpha_2\nu_{21}\left(f^{(0,1)}(x,t)+f^{(1,0)}(x,t)f(x,t)\right).$$


- Comparando ambas expresiones obtenemos que:

    $$\alpha_1+\alpha_2=1,\quad \alpha_2\nu_{21}=\frac{1}{2},$$

    que es un sistema de dos ecuaciones y tres incógnitas, lo cual nos permite fijar una libremente.


- Recapitulando, tenemos el sistema de ecuaciones

    \begin{align}
    \alpha_1+\alpha_2=&\,1,\nonumber\\
    \alpha_2\nu_{21}=&\,\frac{1}{2},\nonumber\\
    k_1=&\,hf(x,t),\nonumber\\
    k_2=&h\,f\left(x+\nu_{21}k_1,t+\nu_{21}h\right),\nonumber\\
    x(t+h)=&\,x(t)+\alpha_1 k_1+\alpha_2 k_2,\nonumber\\
    \end{align}

    que nos proporciona una formula iterativa para encontrar nuestra solución en cualquier punto.


- Por ejemplo, haciendo $\alpha_1=0$, tenemos que $\alpha_2=1$ y $\nu_{21}=1/2$, y nuestra solución es:

    \begin{align}
    k_1=&\,hf(x,t),\nonumber\\
    k_2=&h\,f\left(x+k_1/2,t+h/2\right),\nonumber\\
    x(t+h)=&\,x(t)+ k_2,\nonumber\\
    \end{align}

    que es el método del punto medio que hemos visto.


- Sin embargo hay otras opciones. Cogiendo $\alpha_1=1/2$, tenemos que $\alpha_2=1/2$ y $\nu_{21}=1$ que proporciona:

\begin{align}
k_1=&\,hf(x,t),\nonumber\\
k_2=&h\,f\left(x+k_1,t+h\right),\nonumber\\
x(t+h)=&\,x(t)+\frac{1}{2} k_1+\frac{1}{2}k_2,\nonumber\\
\end{align}

$\quad$ que es otro método de Runge-Kutta de orden 2 llamado método de Euler modificado.

### 7.1.2.3 Método de Runge-Kutta de cuarto orden

- Los algoritmos para los métodos de Runge-Kutta se vuelven más y más complicados a medida que aumentamos el orden.


- En general se considera que el de orden 2 es el que optimiza el cociente precisión versus complejidad.


- En particular, el algoritmo de cuarto orden más usado viene dado por el sistema de ecuaciones:

    \begin{align}
    k_1=&\,hf(x,t),\nonumber\\
    k_2=&h\,f\left(x+k_1/2,t+h/2\right),\nonumber\\
    k_3=&h\,f\left(x+k_2/2,t+h/2\right),\nonumber\\
    k_4=&h\,f\left(x+k_3,t+h\right),\nonumber\\
    x(t+h)=&\,x(t)+\frac{1}{6}\left(k_1+2k_2+2k_3+k_4\right).\nonumber\\
    \end{align}

    que es preciso hasta el orden $h^4$, es decir su error es ${\cal O}(h^5)$.


- El método de Runge-Kutta es el más usado para resolver ecuaciones diferenciales ordinarias, y en general el más fiable para problemas generales.

```python
def runge_kutta4(a,b,N,f,x0):
    tp=np.linspace(a,b,N)
    h=(b-a)/N
    x=x0
    xp=[]
    for t in tp:
        xp.append(x)
        k1=h*f(x,t)
        k2=h*f(x+k1/2,t+h/2)
        k3=h*f(x+k2/2,t+h/2)
        k4=h*f(x+k3/2,t+h/2)
        x+=1/6*(k1+2*k2+2*k3+k4)
    return tp,xp
tp,xp=runge_kutta4(a,b,N,f,0)
plt.plot(tp,xp)
```

## 7.1.3 Soluciones sobre rangos infinitos

- Hasta ahora hemos buscado la solución en un determinado intervalo finito, pero en ocasiones queremos encontrar una solución hasta $t=\infty$.


- Al igual que para la integrales, la forma de hacerlo es usar un cambio de variable. 


- Por ejemplo, haciendo:

    $$u=\frac{t}{1+t}\Rightarrow t=\frac{u}{1-u},$$

    que cumple que para $t\to\infty$ la nueva variables vale $u\to1$, nuestra ecuación diferencial se puede escribir como

    $$\frac{\text{d}x}{\text{d}u}\frac{\text{d}u}{\text{d}t}=f(x,t)\;\Rightarrow\; \frac{\text{d}x}{\text{d}u}=\frac{\text{d}t}{\text{d}u}f\left(x,\frac{u}{(1-u)}\right)=\left(1-u\right)^{-2}f\left(x,\frac{u}{(1-u)}\right).$$


- Por tanto, llamando:

    $$g(x,u)=\left(1-u\right)^{-2}f\left(x,\frac{u}{(1-u)}\right),$$

    nuestro problema inicial se puede escribir como:

    $$\frac{\text{d}x}{\text{d}u}=g(x,u),$$

    que es una nueva ecuación de primer orden. Al resolverla en el intervalo $u\in[0,1]$ estamos obteniendo nuestra ecuación original para $t\in[0,\infty)$.

# 7.2  Sistemas de ecuaciones diferenciales ordinarias

- Estudiemos ahora el caso en el que tenemos más de una variable (variable dependientes) dependiendo de un mismo parámetro (variable independiente), que en general asumiremos juega el papel del tiempo.


- Es decir, tenemos un sistema de ecuaciones diferenciales con dos variables $x(t)$, $y(t)$ que son exclusivamente función de $t$:

    $$\frac{\text{d}x}{\text{d}t}=f_x(x,y,t),\quad \frac{\text{d}y}{\text{d}t}=f_y(x,y,t),$$

    siendo $f_x$ y $f_y$ dos funciones, lineales o no lineales, de las variables $x,y,t$.


- En general, para un número arbitrario de variables y ecuaciones, el sistema se puede escribir en forma vectorial como:

    $$\frac{\text{d}\bf{r}}{\text{d}t}={\mathbf F}({\bf r},t),\quad\text{con}\quad{\bf r}=\left(x,y,\cdots\right),\quad\text{y}\quad{\mathbf F}({\bf r},t)=\left(f_x({\bf r},t),f_x({\bf r},t),\cdots\right),$$

    que nos permite interpretarla como una función de una sola variable y analizar su solución como hemos hecho hasta ahora. 


- Por ejemplo, desarrollando la solución en ${\bf r}(t+h)$ serie de Taylor tenemos:

    $${\bf r}(t+h)={\bf r}(t)+h\frac{{\text d}{\bf r}(t)}{\text{d}t}+{\cal O}\left(h^2\right)={\bf r}(t)+h\,{\mathbf F}({\bf r},t)+{\cal O}\left(h^2\right),$$

    que despreciando el término en $h^2$ nos lleva a la generalización del método de Euler para más de una variable. 


- De la misma forma, el método de Runge-Kutta de cuarto orden para sistemas de ecuaciones diferenciales vendrá dado por:

\begin{align}
{\bf k_1}=&h\,{\bf F}({\bf r},t),\nonumber\\
{\bf k_2}=&h\,{\bf F}({\bf r}+{\bf k_1}/2,t+h/2),\nonumber\\
{\bf k_3}=&h\,{\bf F}({\bf r}+{\bf k_2}/2,t+h/2),\nonumber\\
{\bf k_4}=&h\,{\bf F}({\bf r}+{\bf k_3},t+h),\nonumber\\
{\bf r}\left(t+h\right)=&r(t)+\,\frac{1}{6}({\bf k_1}+2{\bf k_2}+2{\bf k_3}+{\bf k_4}),\nonumber\\
\end{align}


- Estas ecuaciones pueden implementarse directamente en python haciendo uso de que es un lenguaje vectorial: permite operar con vectores, que pueden jugar el papel tanto de argumentos como resultado de una función.

### Ejemplo 7.5: Sistemas de ecuaciones diferenciales ordinarias

# 7.3 Ecuaciones diferenciales de segundo orden

- Hasta ahora hemos estudiado ecuaciones diferenciales de primer orden. Sin embargo, muchos problemas físicos involucran ecuaciones diferenciales dinámicas de segundo orden (aparecen derivadas segundas) o de orden superior.


- La forma general de una ecuación de segundo orden en una variable es

    $$\frac{\text{d}^2x}{\text{d}t^2}=f\left(x,\frac{\text{d}x}{\text{d}t},t\right),$$

    es decir una ecuación que puede involucrar, $x$, su derivada y el tiempo.


- La forma de lidiar con estas ecuaciones, es sin embargo, totalmente equivalente a los sistemas de ecuaciones ordinarias cuando uno tiene en cuenta que llamando

    $$\frac{\text{d}x}{\text{d}t}=y,$$

    nuestra ecuación se puede escribir como

    $$\frac{\text{d}y}{\text{d}t}=f(x,y,t),$$

    y por tanto, lo convierte en un sistema de dos ecuaciones con dos incógnitas.


- De forma análoga, para ecuaciones de orden superior tenemos de forma general el problema

    $$\frac{\text{d}^nx}{\text{d}t^n}=f\left(x,\frac{\text{d}x}{\text{d}t},\frac{\text{d}^2x}{\text{d}t^2},\cdots,\frac{\text{d}^{n-1}x}{\text{d}t^{n-1}},t\right),$$

    que podemos expresar como

    \begin{align}
    \frac{\text{d}x}{\text{d}t}=&y_1,\nonumber\\
    \frac{\text{d}y_1}{\text{d}t}=&y_2,\nonumber\\
    \frac{\text{d}y_2}{\text{d}t}=&y_3,\nonumber\\
    \vdots&\nonumber\\
    \frac{\text{d}y_{n}}{\text{d}t}=&f(x,y_1,y_2,\cdots,y_{n-1},t).\nonumber\\
    \end{align}


- De la misma forma, podemos generalizar un sistema de ecuaciones de orden 2 o superior tratando nuestras variables de forma vectorial; dado el sistema:

    $$\frac{\text{d}^2{\bf r}}{\text{d}t^2}=f\left({\bf r},\frac{\text{d}{\bf r}}{\text{d}t},t\right),$$

    podemos escribirlo como:

    $$\frac{\text{d}{\bf r}}{\text{d}t}={\bf s},\quad \frac{\text{d}{\bf s}}{\text{d}t}=f\left({\bf r},{\bf s},t\right).$$


- De forma general, si comenzamos con un sistema de $n$ ecuaciones de orden $m$ podremos convertirlas en un sistema de $m\times n$ ecuaciones de primer orden, que podremos resolver con los métodos que hemos estudiado hasta ahora.

# 7.4 Variando el paso: métodos con paso adaptativo

- Hasta ahora, todos los métodos que hemos empleado usan un paso de longitud fija que tiene que ser elegido a priori.


- Sin embargo, parece razonable asumir que uno obtendría resultados más precisos y eficientes si pudiésemos variar el paso a medida que el programa estuviese en ejecución, de forma que en cada momento se pudiese elegir el paso más óptimo. 


- Supongamos que la solución de la ecuación diferencial

    $$\frac{\text{d}\,x(t)}{\text{d}\,t}=f(x,t)$$

    viene dada por

```python
from IPython.display import display,Image
display(Image("paso.png",width=600))
```

- Mientras que en algunas zonas la solución varía muy  lentamente, comportamiento que se puede obtener con sólo unos pocos puntos, en otras zonas varia de forma muy abrupta, lo que requiere evaluar la función en muchos más puntos.


- El objetivo es por tanto variar el valor del paso $h$ a medida que vamos construyendo nuestra función, tomando un paso grande cuando la solución sea suave y más pequeño cuando varíe rápidamente.   


- Esto nos permitirá por un lado hacer nuestro programa más rápido y por otro lado más preciso pues el paso se adapta a las necesidades que impone la función. Esta es la idea del método de paso adaptativo. 


- Los métodos adaptativos tienen por tanto dos fases:

    1. Estimar el error asociado a un paso al construir nuestra solución.
    2. Compararlo con la tolerancia objetivo y aumentar o disminuir el paso para conseguir la precisión requerida.


## 7.4.1  Método de Runge-Kutta con paso adaptado

El proceso para obtener un método de Runge-Kutta con paso adaptado sería el siguiente:


- Escogemos un paso inicial, que típicamente será pequeño para ir a lo seguro.


- A continuación calculamos la solución en dos pasos consecutivos, por lo que si empezamos en el punto $x(t)$ terminaremos en $x(t+2h)$.


- Volvemos al punto inicial y realizamos un único paso, pero esta vez de longitud $2h$, que nos vuelve a llevar a $x(t+2h)$.  
  Al igual que con los métodos adaptativos para las integrales, comparando ambas estimaciones podemos obtener el error de nuestra estimación.

```python
display(Image("paso2.png",width=600))
```

- El método de Runge-Kutta de cuarto orden introduce un error ${\cal O}(h^5)$.   
  Por tanto, usando dos pasos de tamaño $h$, la solución de nuestro problema en $t+2h$ sera 
  
  $$x(t+2h)\simeq x_1+2c  h^5,$$
  
    siendo $x_1$ el valor que hemos obtenido usando el método de Runge-Kutta de cuarto orden, y c una constante que asumimos igual para ambos pasos.


- Por otro lado, tomando un paso de tamaño $2h$ se obtiene

  $$x(t+2h)=x_2+c  (2h)^5=x_2+32 ch^5.$$


- Comparando ambas expresiones podemos obtener el error asociado a nuestro paso de longitud $h$

    $$\epsilon=ch^5=\frac{1}{30}\left(x_1-x_2\right).$$


- El objetivo es por tanto hacer ese error igual a nuestra tolerancia objetivo.
  Llamando a esa tolerancia $\epsilon_\text{obj}$ que podemos asumir viene dada por un paso de tamaño $h'$ tenemos
      
    $$\epsilon_\text{obj}=ch'^5=ch^5\left(\frac{h'}{h}\right)^5=\frac{1}{30}(x_1-x_2)\left(\frac{h'}{h}\right)^5,$$


- Ahora bien nuestra tolerancia objetivo se puede expresar como una precisión por unidad de tiempo, $\delta$, de forma que tras un paso temporal $h'$ tengamos

    $$\epsilon_\text{obj}=h'\delta =\frac{1}{30}(x_1-x_2)\left(\frac{h'}{h}\right)^5\;\Rightarrow\,h'=h\left(\frac{30h\delta}{\vert x_1-x_2\vert}\right)^{1/4}=h\rho^{1/4},$$

    donde el cociente 

    $$\rho=\frac{30h\delta}{\vert x_1-x_2\vert},$$
  
    proporciona exactamente el cociente entre la precisión objetivo y el error de nuestra estimación actual $\epsilon$.

### 7.4.1.1 Algoritmo para el método de Runge-Kutta de paso adaptado

1. Definimos una precisión objetivo $\epsilon_\text{obj}=h\delta$.


2. Partiendo de nuestra condición inicial, realizamos dos pasos de anchura $h$, y uno de anchura $2h$ lo cual nos permite obtener el cociente $\rho$.


3. Si $\rho>1$ nuestra precisión es mayor que la objetivo. Nos movemos al punto $t+2h$, pero actualizamos el paso a $h'=h\rho^{1/4}$, de forma que no utilizamos pasos más pequeños de los necesarios.  
   Si $\rho<1$, nuestro resultado no esta alcanzando la precisión deseada. Volvemos al punto 2 pero usando como paso $h'$.


- En resumen, después de cada paso, dependiendo del valor de $\rho$, o aumentamos el valor del paso y nos movemos al siguiente punto o lo disminuimos y repetimos el mismo paso. 


- Para cada estimación tenemos dos valores, $x_1$ y $x_2$, aunque obviamente, al ser $x_1$ más preciso usaremos ese valor. 


- En general, los métodos de Runge-Kutta de paso adaptado involucran más esfuerzo que las que usan un paso fijo (hacemos como mínimo tres estimaciones de Runge-Kutta para obtener sólo dos valores). 


- Sin embargo, el esfuerzo merece la pena. 

    1. Nuestro resultados garantiza que cumple la precisión objetivo.
    2. El programa suele tardar menos tiempo en ejecutarse, pues permite optimizar el paso en aquellos intervalos donde la función es suave.
    
    
- Es posible que en un paso dado el valor de $x_1$ y $x_2$ casi coincidan de forma accidental. Si eso ocurriese, el valor de $h'$ podría hacerse muy grande.   
  Para evitar ese problema se suele definir un incremento máximo del valor del paso en cada iteración.  
  Por lo general se suele imponer que no aumente más del doble de su valor actual. 

## 7.4.2 Sistemas de ecuaciones diferenciales con paso adaptado

- Los métodos con paso adaptativo pueden generalizarse a sistemas de ecuaciones. 


- El único problema es que tendremos un error asociado a cada una de nuestras ecuaciones. 


- Dado el sistema de ecuaciones

    $$\frac{\text{d}x}{\text{d}t}=f_x(x,y,t),\quad \frac{\text{d}y}{\text{d}t}=f_y(x,y,t),$$

    nos encontraremos un error asociado a la ecuación para la variable $x$ y otro para $y$

    $$\epsilon_x=\frac{1}{30}(x_1-x_2),\quad \epsilon_y=\frac{1}{30}(y_1-y_2).$$


- El problema es por tanto definir como combinamos ambos errores para definir el cociente entre nuestro error y la tolerancia deseada. 


- La forma de hacerlo dependerá del problema:
<br/>
    
    - Si $x$ e $y$ representan coordenadas en un espacio de dos dimensiones el error total puede definirse como su error euclídeo $\sqrt{\epsilon_x^2+\epsilon_y^2}$.
    - Si el sistema proviene de una ecuación diferencial de dos variables, obviamente el único error relevante será el asociado a la variable original.

- Al igual que en el método de Romberg para resolver integrales, los métodos adaptativos no sólo nos aportan una forma de determinar el error de nuestro paso, si no que nos permiten mejorar la precisión de nuestra solución.

    que ahora es correcto hasta ${\cal O}(h^6)$.


- Este método es llamado de extrapolación local.


- Con toda la información de la que disponemos, nos permite obtener una estimación más precisa. 


- El único problema es que ahora no conocemos el error de nuestra estimación, sabemos que es menor que la estimación inicial, pero no por cuanto.


- Por supuesto el método se puede continuar, volver a buscar otro paso que nos permita estimar el error de ${\cal O}(h^6)$ y así sucesivamente.


- Es la llamada extrapolación de Richardson, que veremos en la próxima sección.

## 7.4.4 El método de Runge-Kutta-Fehlberg

# 7.5 Otros métodos para resolver ecuaciones diferenciales

- Los métodos de Runge-Kutta son robustos, precisos, fáciles de programar y proporcionan soluciones correctas en la mayoría de los casos. 


- Sin embargo no son de largo los únicos métodos disponibles. Aunque en general se usan menos que los métodos de Runge-Kutta suelen ser útiles en ciertos casos, en particular en aquellas situaciones en las que los métodos de Runge-Kutta no proporcionan una solución correcta.

## 7.5.1 Métodos multipaso

- El método de Euler es un método de un sólo paso, dado $x(t)$ obtenemos directamente el valor en $x(t+h)$ evaluando la función en $f(x,t)$.


- Los métodos de Runge-Kutta utilizan pasos intermedios para obtener una estimación más precisa en el punto $x(t+h)$.   
  Sin embargo, toda esa información es descartada una vez se da el siguiente paso


- Los métodos multipaso tratan de mantener la eficiencia de los métodos de Runge-Kutta, pero reusando la información utilizada en el paso anterior en lugar de descartarla.


- En general los métodos multipasos implican ecuaciones que relacionan la solución en una serie de puntos y el valor de su derivadas.   
  Por ejemplo, para el punto $x(t+\alpha_n h)$ 

    $$x(t+\alpha_n h)+\beta_{n-1} x(t+\alpha_{n-1} h)+\cdots+\beta_{0}x(t)=h\left(\gamma_n f(x(t+\alpha_n h),t+\alpha_n h)+\gamma_{n-1} f(x(t+\alpha_{n-1} h),t+\alpha_{n-1} h)+\cdots\gamma_0f(x(t),t)\right).$$


- Si el coeficiente $\gamma_n\neq0$, $x(t+\alpha_n h)$, el valor que queremos determinar, aparece en los dos lados de la ecuación,  
  son los llamados métodos multipaso implícitos. 


- Si  $\gamma_n=0$ el valor de nuestra solución en el último paso, $\alpha_n h$ se puede obtener explícitamente,   
  son los llamados métodos multipaso explícitos. 

### 7.5.1.1 Métodos de Adams-Bashford

- Son métodos explícitos, $\gamma_n=0$, donde los coeficientes $\beta_i$ se fijan de forma que $\beta_{n-1}=-1$ y el resto $\beta_{n-2}=\beta_{n-2}=\cdots=\beta_{0}=0$.


- Los coeficientes de $\gamma_{n-m}$, con $n>m\ge0$ y el tamaño de los pasos (valor de los coeficientes $\alpha_i$) son escogidos para asegurar una precisión a un orden dado utilizando una interpolación polinómica (método de Lagrange) de la derivada que luego es integrada analíticamente.


- Por ejemplo el método de Adams-Bashford de orden 4 es

    $$x(t+4h)=x(t+3h)+\frac{h}{24}\left(55f(x(t+3h),t+3h)-59f(x(t+2h),t+2h)+37f(x(t+h),t+h)-9f(x(t),t)\right)$$

    introduce un error ${\cal O}(h^5)$.


- Para empezar el algoritmo de Adams-Bashford necesitamos 4 puntos anteriores. 
  Son por tanto métodos que requiere utilizar cualquier otro, por ejemplo un método de Runge-Kutta, para establecer las condiciones iniciales.

### 7.5.1.2 Método de Adams-Moulton

- Es parecido al método de Adams-Bashford pero introduce una ecuación implícita. 


-  Los coeficientes $\beta_i$ se fijan de la misma forma $\beta_{n-1}=-1$ y $\beta_{n-2}=\beta_{n-2}=\cdots=\beta_{0}=0$ pero en este caso $\gamma_{n}\neq 0$.


- Al eliminar esta restricción, los métodos de Adam-Moulton son de orden $n+1$ usando sólo $n$ pasos.


- Por ejemplo, el método de  Adams-Moulton de cuarto orden viene dado por las ecuaciones:

    $$x(t+3h)=x(t+2h)+\frac{h}{24}\left(9f(x(t+3h),t+3h)+19f(x(t+2h),t+2h)-5f(x(t+h),t+h)+f(x(t),t)\right).$$

## 7.5.2 Método del salto de Rana

- Dada la ecuación diferencial

    $$\frac{\text{d}x}{\text{d}t}=f(x,t),$$

    ya hemos visto que el método de Runge-Kutta de segundo orden o del punto medio viene dado por las ecuaciones:

    \begin{align}
    x(t+h/2)=&x(t)+h/2f(x,t),\nonumber\\
    x(t+h)=&x(t)+hf(x(t+h/2),t+h/2),\nonumber
    \end{align}


- En cada paso, usa el método de Euler para calcular la solución en el punto medio $x(t+h/2)$ y usa ese resultado para la estimación en $x(t+h)$.


- Lo cual se puede representar gráficamente por medio del panel superior de la siguiente figura:

```python
display(Image("leapfrog.png",width=600))
```

- El método del salto de la rana (leapfrog) es una variante de esta idea, que se puede describir gráficamente por el panel inferior de la imagen de arriba.
 
 
- En la segunda iteración en vez de usar el método de Euler para calcular la solución en $x(t+3h/2)$, usa el método del punto medio partiendo del punto $x(t+h/2)$

    \begin{align}
    x(t+h/2)=&x(t)+h/2\,f(x,t),\nonumber\\
    x(t+h)=&x(t)+h\,f(x(t+h/2),t+h/2),\nonumber\\
    x(t+3h/2)=&x(t+h/2)+h\,f(x(t+h),t+h),\nonumber\\
    \end{align}

    resultado que podemos usar para calcular el siguiente paso:

    $$x(t+2h)=x(t+h)+h\,f(x(t+3h/2),t+3h/2),$$

    lo cual permite definir un algoritmo iterativo. 


- El método es también de orden 2, con un error ${\cal O}(h^3)$ y puede extenderse a sistemas de ecuaciones sustituyendo $x$ por el vector ${\bf r}$ y $f$ por el vector ${\bf F}$.

    \begin{align}
    {\bf r}(t+h)=&{\bf r}(t)+h\,{\bf F}({\bf r}(t+h/2),t+h/2),\nonumber\\
    {\bf r}(t+3h/2)=&{\bf r}(t+h/2)+h\,{\bf F}({\bf r}(t+h),t+h).\nonumber\\
    \end{align}


- Aunque al ser un método de orden dos no queda muy claro cual es su interés en comparación con el método de Runge-Kutta de orden 4, sus virtudes quedan claras cuando se aplican a una serie de problemas físicos que veremos a continuación.

### 7.5.2.1  Inversión temporal y conservación de energía

- El método del salto de la rana es invariante bajo inversión temporal: es decir, si partiendo del punto final usamos el método del salto de la rana con un paso $-h$, volveremos al punto inicial recorriendo todos los puntos intermedios. 


- Para verlo, sustituyamos primero $h\to -h$ en las ecuaciones del salto de rana, que nos permiten estudiar la solución hacía atrás en el tiempo. 

    \begin{align}
    x(t-h)=&\,x(t)-hf\left(x\left(t-\frac{1}{2}h\right),t-\frac{1}{2}h\right),\nonumber\\
    x\left(t-\frac{3}{2}h\right)=&\,x\left(t-\frac{1}{2}h\right)-hf(x(t-h),t-h),\nonumber\\
    \end{align}

    y posteriormente, para no obtener nuestra solución en un instante anterior al inicial, sustituyamos $t\to t+\frac{3}{2}h$

    \begin{align}
    x\left(t+\frac{1}{2}h\right)=&\,x\left(t+\frac{3}{2}h\right)-hf\left(x\left(t+h\right),t+h\right),\nonumber\\
    x\left(t\right)=&\,x\left(t+\frac{1}{2}h\right)-hf\left(x\left(t+\frac{1}{2}h\right),t+\frac{1}{2}h\right),\nonumber\\
    \end{align}

    que nos proporciona la solución de $x(t)$ y $x(t+h/2)$ a partir del valor de $x(t+h)$ y $x(t+3h/2)$.
    
    
- Sin embargo, estas soluciones coinciden exactamente con el resultado de ir hacía adelante en el tiempo: al resolver una ecuación diferencial con el método del salto de rana hacía atrás en el tiempo, con paso $-h$, obtenemos en cada instante $t$ la misma solución $x(t)$ que cuando vamos hacía adelante en el tiempo.


- Lo mismo no se aplica al método de Runge-Kutta, al empezar por el final del intervalo temporal y aplicar el método con un paso negativo no recuperaremos las condiciones iniciales. 


- Desde un punto de vista computacional, la invariancia temporal juega un papel fundamental para la conservación de la energía. 


- Para verlo, volvamos al problema del péndulo moviéndose en el vacío (sin rozamiento), cuyas ecuaciones de movimiento

    $$\frac{\text{d}^2\theta}{\text{d}t^2}=-\frac{g}{l}\sin\theta,$$

    satisfacen la conservación de la energía.

- Lo interesante es que el  movimiento del péndulo también es por si mismo invariante bajo reflexión temporal. El movimiento que ha hecho después de un periodo completo es igual si se mueve hacia delante o hacia atrás en el tiempo.


- Eso implica que al cabo de un ciclo completo la energía del sistema calculada con el método del salto de la rana debe volver al mismo valor y seguiría así para siempre, cosa que no sucede con el método de Runge-Kutta.

### Ejemplo 7.9: el método del salto de la rana

Aplicar el método del salto de la rana a la ecuación diferencial

$$\frac{\text{d}^2x}{\text{d}t^2}=\left(\frac{\text{d}x}{\text{d}t}\right)^2-x-5,$$

usando el método del salto de la rana en el intervalo $t\in[0,50]$, con un paso h=0.001 y usando las condiciones iniciales $x(0)=1$, $v(0)=0$.   
Representar la solución como función del tiempo.

```python
import numpy as np
import matplotlib.pyplot as plt
#Definición de las funciones:
def f(r):
    return np.array([r[1],r[1]**2-r[0]-5],float)
#Salto de rana
r=np.array([1,0],float)
tp=np.arange(0,50,1e-3)
h=1e-3
r2=r+h/2*f(r)
xp=[]
vp=[]
for t in tp:
    r+=h*f(r2)
    r2+=h*f(r)
    xp.append(r[0])
    vp.append(r[1])
plt.plot(tp,xp)
plt.show()
```

## 7.5.3 El método de Verlet

- Es un caso particular del método del salto de la rana para estudiar las ecuaciones del movimiento de un sistema físico. 


- Sea la ecuación de movimiento

  $$\frac{\text{d}^2x}{\text{d}t^2}=f(x,t),$$

  ya hemos visto que la podemos descomponer en un sistema de dos ecuaciones diferenciales ordinarias

  $$\frac{\text{d}x}{\text{d}t}=v,\quad\frac{\text{d}v}{\text{d}t}=f(x,t).$$


- En vez de tratarlas de forma vectorial, apliquemos ahora el método del salto de la rana a cada una de ellas.  


- Asumamos que las condiciones iniciales nos fijan $x(t)$, el valor de la solución en el instante $t$, y  $v(t+h/2)$, la velocidad en el instante $t+h/2$.   
  Aplicando el método de la rana tenemos entonces:
  
    \begin{align}  
    x(t+h)=&x(t)+hv(t+h/2),\nonumber\\
    v(t+3h/2)=&v(t+h/2)+hf(x(t+h),t+h).\nonumber\\
    \end{align}


- Que definen un sistema en el que sólo tenemos que conocer la solución en múltiples enteros de $h$ y la velocidad en múltiplos semienteros $h/2$.


- Esto supone una mejora con respecto aplicar el método del salto de la rana al vector $(x,v)$, pues en ese caso tendríamos que calcular la posición y velocidad en todos los puntos. 


- El único problema es puede ser que sólo conozcamos la velocidad en el instante $t$ no en $t+h/2$.   
  En ese caso, partiendo de $v(t)$, aplicamos Euler con paso $h/2$

    $$v(t+h/2)=v(t)+\frac{1}{2}h f(x(t),t),$$


- También es posible que necesitemos estimar algún observable que requiera conocer la posición y velocidad en el mismo instante, por ejemplo la energía total del sistema. En ese caso, aplicando Euler de nuevo con paso $h/2$ se tiene

    $$v(t+h)=v(t+h/2)+\frac{1}{2}h f(x(t+h),t+h).$$


- Con todas estas ecuaciones podemos definir un algoritmo para tratar con este tipo de problemas. Dadas las condiciones iniciales x(t) y v(t) tenemos

    \begin{align}
    v(t+h/2)=&v(t)+\frac{1}{2}h f(x(t),t)\nonumber\\
    x(t+h)=&x(t)+hv(t+h/2),\nonumber\\
    k=&hf(x(t+h),t+h),\nonumber\\
    v(t+h)=&v(t+h/2)+\frac{1}{2}k,\nonumber\\
    v(t+3h/2)=&v(t+h/2)+k\nonumber\\
    \end{align}

    que es el llamado método de Verlet. 


- El método se puede extender trivialmente a ecuaciones de movimiento en más de una dimensión, con sólo sustituir $x\to {\bf r}$ y $f\to {\bf F}$ siendo ${\bf r}$ y ${\bf F}$ los vectores de posiciones y funciones de nuestro sistema de ecuaciones.

## 7.5.4 El método del punto medio modificado

- Otra de las razones por las que el método del salto de rana es útil esta relacionado con el comportamiento de su error. 


- Sabemos que el método del salto de rana es de segundo orden, es decir, el error después de realizar un único paso es ${\cal O}(h^3)$. 


- Es decir, de forma general podemos definir el error del salto de rana en un paso dado como

    $$\epsilon(h)=c_3 h^3 +\dots,$$

    siendo $c_3$ una constante.


- Si partiendo de este nuevo punto hacemos ahora un paso hacía atrás en el tiempo, es decir, con longitud $-h$, el hecho de que el método del salto de rana sea invariante bajo reflexión temporal implica que tenemos que volver exactamente al mismo punto, y por tanto

    $$\epsilon(-h)=-\epsilon(h),$$

    es decir, $\epsilon(h)$ tiene que ser una función impar, o lo que es lo mismo, la expansión de Taylor del error sólo debe contener potencias impares de $h$

    $$\epsilon(h)=c_3 h^3 +c_5 h^5 +c_7 h^7 +\dots,$$

    siendo $c_5, c_7,\cdots$ constantes.


- Es importante notar, sin embargo, que $\epsilon(h)$ nos proporciona el error de un único paso. 


- Al analizar el método de Euler ya hemos visto que **después de realizar varios pasos en error acumulado es un orden de $h$ mayor**.


- En principio, eso implica que el error total del método del salto de rana, sólo debe contener potencias pares de $h$.


- Sin embargo, esto no es realmente correcto: 
<br/>

    - El primer paso del método de la rana es un paso de Euler, que introduce un error que no tiene una paridad dada. 
    - Lo cual hace que el error final del método de la rana tenga tanto potencias pares como impares. 
    
    
- Aunque a priori esto puede no parecer importante, un error par es la base del siguiente método que veremos, el método de Bulirsch-Stoer, que proporciona las soluciones más precisas.


- Para resolver este problema, el método del salto de rana se modifica de la siguiente forma. 

    1. Dado un problema definido en el intervalo $[t,t+H]$ con n paso de anchura $h=H/n$, escribamos en primer lugar el algoritmo para el método del salto de rana de la forma:
    
        \begin{align}
        x_0=&x(t),\nonumber\\
        y_1=&x(t+h/2)=x_0+\frac{h}{2}f(x_0,t),\nonumber\\
        \end{align}
    
         que representan las condiciones iniciales del problema, a partir del cual se puede construir la solución: 

        \begin{align}
           x_1=&x(t+h)=x_0+h f(y_1,t+h/2),\nonumber\\
           y_2=&x(t+3h/2)=y_1+hf(x_1,t+h),\nonumber\\
           x_2=&x(t+2h)=x_1+h f(y_2,t+3h/2),\nonumber\\
           \vdots\quad&\quad\;\,\vdots\qquad\qquad\;\vdots\nonumber\\
           y_{m+1}=&x\left(t+(m+1/2)h\right)=y_m+hf\left(x_m,t+mh\right),\nonumber\\
           x_{m+1}=&x\left(t+(m+1)h\right)=x_m+h f\left(y_{m+1},t+(m+1/2)h\right),\nonumber\\
         \vdots\quad&\qquad\quad\,\vdots\qquad\qquad\qquad\vdots\nonumber\\
           y_{n}=&x\left(t+H-h/2\right)=y_{n-1}+hf\left(x_{n-1},t+H-h\right),\nonumber\\
           x_{n}=&x\left(t+H\right)=x_{n-1}+h f\left(y_{n},t+H-h/2\right),\nonumber\\     
         \end{align}
 
 
   2. Los últimos dos puntos de la solución, $y_n=x(t+H-h/2)$ y $x_n=x(t+H)$, se obtienen a partir de $y_{n-1}$ y $x_{n-1}$, respectivamente.

## 7.5.5 Método de Bulirsch-Stoer

- El método de Bulirsch-Stoer es fruto de combinar:
    
    1. El método del punto medio modificado
    
    2. La extrapolación de Richardson (la base del método de Romberg para la integración numérica).
    
    
- El problema consiste, una vez más, en resolver la ecuación diferencial ordinaria

    $$\frac{\text{d}x}{\text{d}t}=f(x,t),$$

    en un intervalo $[t,t+H]$.


- Empezamos calculando la solución del problema en el punto final $x(t+H)$, usando el método del punto medio modificado, inicialmente con un único paso $h_1=H$.  
  Que nos proporciona el valor $R_{1,1}$ con un error ${\cal O}(h_1^2)$ 
  
    $$x(t+H)=R_{1,1}+c_1 h_1^2+{\cal O}(h_1^4),$$

    siendo $c_1$ una constante desconocida. 


- Una vez obtenido $R_{1,1}$ podemos volver al punto inicial y obtener una nueva solución, usando esta vez dos pasos de anchura $h_2=h_1/2=H/2$, dando lugar a

    $$x(t+H)=R_{2,1}+c_1 h_2^2+{\cal O}(h_2^4).$$


- Como ambas soluciones corresponden al valor exacto de la función, igualando obtenemos que

    $$R_{2,1}+c_1 h_2^2+{\cal O}(h_2^4)=R_{1,1}+c_1 h_1^2+{\cal O}(h_1^4)=R_{1,1}+4c_1 h_2^2+{\cal O}(h_2^4),$$

    que nos permite determinar el valor del error:

    $$c_1h_2^2=\frac{1}{3}\left(R_{2,1}-R_{1,1}\right),$$

    por lo que nuestra solución viene dada por

    $$x(t+H)=R_{2,1}+\frac{1}{3}\left(R_{2,1}-R_{1,1}\right)+{\cal O}(h_2^4)=R_{2,2}+{\cal O}(h_2^4),$$

    donde 

    $$R_{2,2}=R_{2,1}+\frac{1}{3}\left(R_{2,1}-R_{1,1}\right)$$ 

    es ahora una nueva estimación que es correcta hasta el orden ${\cal O}(h_2^4)$.


 - Podemos repetir el proceso eligiendo un nuevo paso $h_3=H/3$ que nos proporciona la estimación $R_{3,1}$ de forma que

    $$x(t+H)=R_{3,1}+c_1 h_3^2+{\cal O}(h_3^4)=R_{2,1}+\frac{9}{4}c_1 h_3^2+{\cal O}(h_3^4),$$

    que nos permite obtener el nuevo error  $c_1 h_3^2=4/5 \left(R_{3,1}-R_{2,1} \right)$ y por tanto, definir una nueva estimación

    $$R_{3,2}=R_{3,1}+\frac{4}{5}\left(R_{3,1}-R_{2,1} \right),$$

    tal que 

    $$x(t+H)=R_{3,2}+c_2h_3^4+{\cal O}(h_3^6)=R_{2,2}+ \frac{81}{16}c_2h_3^4+{\cal O}(h_3^6),$$

    que nos permite obtener una nueva estimación

    $$R_{3,3}=R_{3,2}+\frac{16}{65}\left(R_{3,2}-R_{2,2}\right),$$

    que coincide con la solución

    $$x(t+h)=R_{3,3}+{\cal O}(h_3)^6$$

    introduciendo un error ${\cal O}(h_3)^6$ introduciendo tan sólo 3 pasos. 


- De forma general, podemos continuar este procedimiento utilizando para cada paso el método del punto medio modificado con un intervalo adicional.    
  Llamando a la estimación de nuestro paso actual $R_{n,1}$, siguiendo el procedimiento anterior podemos calcular $R_{n,2}$, $R_{n,3}$,$\cdots$, con
  
    $$x(t,H)=R_{n,m}+c_m h_n^{2m}+{\cal O}\left(h_n^{2m+2} \right),$$
  
    siendo $c_m$ una constante desconocida, mientras que la estimación $R_{n-1,m}$ cumple

    $$x(t,H)=R_{n-1,m}+c_m h_{n-1}^{2m}+{\cal O}\left(h_{n-1}^{2m+2} \right),$$
  
  
- Sin embargo, teniendo en cuenta que $h_n=H/n$, se obtiene que

    $$h_{n-1}=\frac{n}{n-1}h_n,$$

    lo cual nos permite obtener que

    $$c_m h_n^{2m}=\frac{R_{n,m}-R_{n-1,m}}{\left[n/(n-1)\right]^{2m}-1},$$

    que nos permite definir la nueva estimación

    $$R_{n,m+1}=R_{n,m}+\frac{R_{n,m}-R_{n-1,m}}{\left[n/(n-1)\right]^{2m}-1}$$

- La primera columna corresponde al resultado de obtener el punto medio modificado con un número de puntos dado, mientras que el resto de valores  en una fila dada son fruto de la extrapolación de Richardson.


- Cada valor de la extrapolación sólo depende de dos estimaciones adicionales, tal y como se indica por las flechas.


- El último valor en cada fila corresponde a la extrapolación más precisa para un número de puntos dado.

### 7.5.5.1 Características del método de Bulirsch-Stoer

- El método de Bulirsch-Stoer implica aumentar el número de pasos hasta que el error de nuestro estimación

    $$c_m h_n^{2m}=\frac{R_{n,m}-R_{n-1,m}}{\left[n/(n-1)\right]^{2m}-1},$$

    es menor que nuestra tolerancia, que en general se define como un error por unidad de tiempo $\delta$, por lo que, en el punto $t+H$ es $\epsilon=\delta H$.


- A diferencia del método de Romberg para la integración numérica en el que el número de pasos se dobla en cada iteración, en el método de Bulirsch-Stoer el valor de $n$ se incrementa sólo en una unidad cada vez.  


- Formalmente, sólo calcula de forma precisa el valor de la solución en $x(t+H)$. Para cada punto intermedio sólo se obtienen estimaciones basadas en el punto medio modificado cuyo error es ${\cal O}(h^2)$.


- El método de Bulirsch-Stoer es en el fondo una expansión de $x(t+H)$ en potencias de $h_n=H/n$, por consiguiente sólo tiene sentido si la serie converge más o menos rápido. 


- Si uno necesita cientos o miles de iteraciones para obtener una precisión asumible el método no es una buena elección. 


- En la práctica eso implica que $H$ no puede ser muy grande. 


- Por tanto, si queremos resolver el problema en un intervalo dado $[a,b]$, es necesario dividirlo en un un número $N$ de subintervalos de anchura $H=(b-a)/N$, sobre los cuales apliquemos el método de Bulirsch-Stoer. 

### 7.5.5.2 Algoritmo del método de Bulirsch-Stoer

Definimos una precisión objetivo por unidad de tiempo y dividimos nuestro solución en $N$ subintervalos de longitud $H$. A cada uno de estos subintervalos le aplicamos el siguiente algoritmo:

1. Usamos el método del punto medio modificado para obtener la estimación $R_{1,1}$ de la solución en $x(t+H)$ usando un sólo paso.


2. Incrementamos el número de pasos $n$ en uno, y obtenemos una nueva estimación del punto medio modificado con ese número de puntos $R_{n,1}$.


3. Usamos la ecuación 

    $$R_{n,m+1}=R_{n,m}+\frac{R_{n,m}-R_{n-1,m}}{\left[n/(n-1)\right]^{2m}-1},$$

    para obtener los valores de $R_{n,2},\cdots,R_{n,n}$ correspondientes a la fila $n$.


4. Después de calcular la final completa comparamos el error de nuestra estimación

    $$c_m h_n^{2m}=\frac{R_{n,m}-R_{n-1,m}}{\left[n/(n-1)\right]^{2m}-1},$$

    con la precisión objetivo $\delta H$. Si el error es mayor que la tolerancia volvemos al punto 2. De otra forma nos movemos al siguiente subintervalo $H$.

- Se puede extender trivialmente a sistemas de ecuaciones con sólo sustituir $x$ por el vector ${\bf r}$, por lo que las estimaciones $R_{n,n}$ también serán vectores. 


- De la misma forma, se puede aplicar a ecuaciones diferenciales de orden superior con sólo transformarlas en sistemas de ecuaciones de orden 1. 


- De forma general da mejores resultados que un Runge-Kutta con paso adaptado con menos esfuerzo.


- Sin embargo, al ser un método basado en una serie perturbativa en $h$, las soluciones deben tener un comportamiento suave para que la convergencia de la serie sea rápida.


- Por tanto, ecuaciones diferenciales con comportamientos patológicos (grandes fluctuaciones, divergencias, $\cdots$) no son adecuados para este método.


- Cuando es aplicable, es considerado como uno de los mejores métodos para resolver ecuaciones diferenciales ordinarias.

### Ejemplo 7.11: el método de Bulirsch-Stoer y el péndulo no-lineal

Volver a resolver las ecuaciones del péndulo no lineal usando esta vez el algoritmo de Bulirsch-Stoer con una precisión por unidad de tiempo de $10^{-8}$.

```python
import numpy as np
import matplotlib.pyplot as plt
l=0.1
g=9.81
r=np.array([179/180*np.pi,0],float)
def f(r):
    theta=r[0]
    w=r[1]
    fw=-g/l*np.sin(theta)
    return np.array([w,fw],float)
    
delta=1e-8
a=0
b=10
H=10/100
def BS(f,r,a,b,delta):
    H=(b-a)/100
    tp=np.arange(a,b,H)
    thetap=[]
    for t in tp:
        thetap.append(r[0])
        n=1
        h=H/n
        r1=r+0.5*h*f(r)
        r2=r+h*f(r1)
        R1=np.empty([n,2],float)
        R1[0]=0.5*(r2+r1+0.5*h*f(r2))
        err=2*delta*H
   
        while err>delta*H:
            n+=1
            h=H/n
            r1=r+0.5*h*f(r)
            r2=r+h*f(r1)
            for i in range(n-1):
                r1+=h*f(r2)
                r2+=h*f(r1)
            R2=np.empty([n,2],float)
            R2[0]=0.5*(r2+r1+0.5*h*f(r2))
            for m in range (1,n):
                error=(R2[m-1]-R1[m-1])/((n/(n-1))**(2*m)-1)
                R2[m]=R2[m-1]+error
            err=abs(error[0])
            R1=np.copy(R2)
        r=R2[n-1]
    return tp,thetap
tp,r=BS(f,r,0,10,delta)

plt.plot(tp,r)    
```

### Ejemplo 7.12: lanzamiento de un proyectil en vertical

Sea una pelota lanzada en vertical desde el suelo. Resolver el problema de contorno sabiendo que la pelota debe alcanzar el suelo al cabo de 10 s. 

```python
import numpy as np
import matplotlib.pyplot as plt

def biseccion(x1,x2,f,eps):
    if f(x1)*f(x2)>0:
        return print('La función tiene el mismo signo en ambos puntos')
    else:
        while x2-x1>eps:
            xm=x1+(x2-x1)/2
            if f(xm)*f(x1)<0:
                x2=xm
            else:
                x1=xm
    return xm
    
def f(r,t):
    g=9.81
    fv=-g
    return np.array([r[1],fv],float)
N=1000
h=10/1000
tp=np.linspace(0,10,N)
x0=0
def H(v):
    r=np.array([x0,v],float)
    for t in tp:
        k1=h*f(r,t)
        k2=h*f(r+k1/2,t+h/2)
        k3=h*f(r+k2/2,t+h/2)
        k4=h*f(r+k3,t+h)
        r+=(k1+2*k2+2*k3+k4)/6
    return r[0]
v1=1
v2=100
h1=H(v1)
h2=H(v2)
eps=1e-6
while abs(h2-h1)>eps:
    vp=(v1+v2)/2
    hp=H(vp)
    if hp*h1>0:
        v1=vp
        h1=hp
    else:
        v2=vp
        h2=hp
vfinal=(v1+v2)/2
print(vfinal)
```

## 7.6.2 Problemas de autovalores

- Un caso especial de problema de contorno ocurre cuando la ecuación o el sistema de ecuaciones es lineal y homogénea, es decir, que todos los términos de la ecuación son proporcionales a la variable dependiente. Este es el caso de los problemas de autovalores. 


- El caso canónico es el de resolver la ecuación de Schrödinger independiente del tiempo para un determinado potencial, es decir

    $$-\frac{\hbar^2}{2m}\frac{\text{d}^2\psi(x)}{\text{d}x^2}+V(x)\,\psi(x)=E\psi(x),$$

    donde $\psi(x)$ es la función de onda, $V(x)$ el potencial y $E$ la energía total del sistema. 


- Supongamos por ejemplo el caso de un pozo infinito

    $$V(x)=\left\{\begin{array}{l l} 0&\text{para 0 <x <L,}\\\infty&\text{en cualquier otro punto,}\end{array}\right.$$

    siendo $L$ la longitud del pozo.


- Aunque este problema se puede resolver de forma analítica vamos a ver cual es su solución numérica. 


- Puesto que la probabilidad de encontrar la partícula fuera de la región 0<x<L es nula, las condiciones de contorno de nuestro problema son 

    $$\psi(0)=\psi(L)=0.$$


- Transformamos la ecuación en un sistema de ecuaciones de primer orden tenemos

    $$\frac{\text{d}\psi(x)}{\text{d}x}=\phi(x),\quad \frac{\text{d}\phi(x)}{\text{d}x}=\frac{2m}{\hbar^2}\left[V(x)-E\right]\psi(x).$$


- Para resolver el sistema usando cualquiera de los métodos que hemos estudiado hasta ahora debemos imponer condiciones iniciales. 


- Conocemos el valor de $\psi(0)$ pero no el de $\phi(0)=\left.\text{d}\psi(x)/\text{d}x\,\right\vert_{x=0}$.


- Podemos por tanto utilizar una estimación inicial y comprobar el valor de la función de onda en $\psi(L)$ que  de forma general no satisfará la condición de contorno. 


- Podría parecer que este problema es resoluble usando el método del disparo, pero lamentablemente no es el caso.


- Como el problema es lineal, al cambiar el valor de la derivada de forma que el valor de la función de onda en $x\to 0^+$ sea el doble que el primer intento, conseguiremos una nueva solución de la ecuación de Schrödinger, que será $2\,\psi(L)$.

```python
from IPython.display import display,Image
display(Image("square.png",width=600))
```

- El problema es que en general no existe solución a la ecuación excepto para ciertos valores de $E$, los autovalores de la función de onda. 


- Es justo por esto por lo que las energías de los sistemas cuánticos están cuantizadas, sólo hay un conjunto de energías permitidas. 


- El problema por tanto consiste en encontrar el valor de $E$ que nos permite resolver el problema de contorno,  es decir, podemos entender la ecuación de Schrödinger como una función de la energía $f(E)$ que nos proporciona el valor de la función de onda en $x=L$. 


- El objetivo es por tanto encontrar el valor de E tal que $\psi(L)=f(E)=0$.


- Lo interesante es notar que el el problema no depende del $\phi(0)$, el único efecto de cambiar esta condición es multiplicar la solución por una constante.


- El resultado final, será correcto excepto por un factor de normalización.


- Tradicionalmente esta valor se fija imponiendo que la función de onda este normalizada

    $$\int_{-\infty}^\infty\vert\psi(x)\vert^2\text{d}x=1.$$

## 7.6.3 Métodos de relajación

- Los métodos de relajación utiliza una estrategia opuesta a la del disparo. 


- Mientras que en el método de disparo partimos de una solución que satisface la ecuación diferencial pero no cumple las condiciones de contorno, en el método de la relajación partimos de una solución que satisface las condiciones de contorno pero no la ecuación diferencial. 


- Iterativamente partimos de esta solución y la modificamos hasta que también satisfaga la ecuación diferencial.  


- Como ya veremos en el siguiente tema, son métodos empleados de forma habitual en la solución de ecuaciones en derivadas parciales.     

### 7.6.3.1 El método de diferencias finitas

- Dividimos el intervalo de integración $(a,b)$ en $N$ subintervalos de longitud $h$. 


- Eso nos permite definir nuestra solución en un conjunto discreto de puntos (puntos de la malla) que denotamos $x_i$ con $x_i=0,\cdots,N$.


- Las derivadas de nuestra solución en la ecuación diferencial se calculan utilizando los métodos de diferencias finitas.


- Por ejemplo utilizando la aproximación del punto medio se tiene para la primera derivada que

    $$x'_i=\frac{x_{i+1}-x_{i-1}}{2h},$$

    mientras que para la segunda derivada el resultado sería

    $$x''_i=\frac{x_{i+1}-2x_i+x_{i-1}}{h^2}.$$


- De esta forma, la ecuación diferencial se puede escribir como $N+1$ ecuaciones simultáneas para $N+1$ incógnitas $x_0,\cdots,x_{N}.$



- Si las funciones son lineales, podremos resolver el sistema utilizando por ejemplo la descomposición $LU$.



- Si las ecuaciones diferenciales son no-lineales, las ecuaciones algebraicas serán también no-lineales y deberemos resolverlas por el método de Newton-Raphson, relajación,...


- Por ejemplo dada la ecuación de orden 2 y usando el método de relajación tendríamos

    $$\frac{\text{d}^2x}{\text{d}t^2}=f(x,t),$$

    definida en el intervalo $[0,H]$ con condiciones de contorno

    $$x(0)=x(H)=0,$$


1. Dividimos el intervalo $[0,H]$ en $N$ subintervalos.


2. Expresamos las derivadas segundas por medio de las diferencias centradas para discretizar la ecuación:

    $$\frac{x_{i+1}-2x_i+x_{i-1}}{h^2}-f(x_i,t_i)=0,$$

    que implica la ecuación

    $$x_i=\frac{1}{2}\left(x_{i+1}+x_{i-1}-h^2f(x_i,t_i)\right),$$


- Usamos el método del punto fijo o relajación para resolver la ecuación obteniendo las iteraciones

    $$x_i^{(n+1)}=\frac{1}{2}\left(x_{i+1}^{(n)}+x_{i-1}^{(n)}-h^2f(x_i^{(n)},t_i)\right),$$

    para todos los valores de i con $1\le i \le N-1$.


- Análogamente al comentario que hicimos en los métodos iterativos de resolución de ecuaciones lineales y no lineales, a veces, se puede acelerar la convergencia utilizando un método SOR (Succesive Over Relaxation), de sobrerelajación. 


- Es decir, si definimos nuestra nueva estimación como

    $$x_i^{(n+1)}=x_i^{(n)}+\Delta x_i^{(n)},$$

    el método de sobrerelajación intentan acelerar la convergería del problema haciendo 

    $$x_i^{(n+1)}=x_i^{(n)}+(1+\omega)\Delta x_i^{(n)}.$$

    siendo $\omega$ es el parámetro de sobre relajación que en general $\omega \in [-1, 1]$.
<br/>
     
    - $\omega > 0$ usados habitualmente para acelerar los procesos de convergencia lenta.  
    - $\omega < 0$ usados habitualmente para para conseguir la convergencia de procesos iterativos divergentes.


- Usando el método de sobrerelajación nuestra estimación de la iteración $n+1$ se modificaría como

    $$x_i^{(n+1)}=\frac{1+\omega}{2}\left(x_{i+1}^{(n)}+x_{i-1}^{(n)}-h^2f(x_i^{(n)},t_i)\right)-\omega x_i^{(n)},$$

# 7.7 Estabilidad y rigidez (stiffness)

## 7.7.1 problemas de condiciones iniciales

### 7.7.1.1 Estabilidad

- Un método de integración de un problema de condiciones iniciales es dice que es estable si los efectos de los errores locales no se acumulan catastróficamente, es decir los errores globales permanecen acotados.


- Si un método es inestable, el error global crecerá exponencialmente, eventualmente produciendo un overflow.


- La estabilidad depende de tres factores:
<br/>

    - Las ecuaciones diferenciales.
    
    - El método de resolución.
    
    - El valor del incremento h.
    
    
- Estudiar la estabilidad no es sencillo salvo en ecuaciones diferenciales lineales.

#### Ejemplo 7.15: estabilidad del método de Euler

Consideremos el problema:

$$\frac{\text{d}x}{\text{d}t}=-\lambda x,\quad x(0)=\alpha.$$    

donde $\lambda$ es una constante positiva. La solución analítica del problema es:

$$x(t)=\beta \text{e}^{-\lambda t}.$$

Resolvámosla por el método de Euler:

$$x(t+h)=x(t)-h\lambda\, x(t)=(1-h\lambda)\,x(t).$$


Si $\vert 1-h\lambda\vert>1$, el método será inestable pues $x$ crecerá de forma no acotada en cada paso numérico.


De esta forma el método sólo convergerá si

$$h<\frac{2}{\lambda}.$$

El problema se puede extender a sistemas de ecuaciones lineales, obteniéndose el resultado 

$$h<\frac{2}{\lambda_\text{max}},$$


siendo $\lambda_\text{max}$ el mayor autovalor de la matriz de coeficientes constantes $\Lambda$ que definen el sistema lineal correspondiente.

### 7.7.1.2 Rigidez

- Un problema de valores iniciales para un sistema de ecuaciones se dice que es rígido (stiff) si sólo es resoluble usando determinados métodos númericos y no otros. 


- Suele estar asociado a ecuaciones diferenciales vectoriales donde algunas componentes de la solución vectorial ${\bf r}(t)$ varían más rápidamente con $t$ que otras.


- La rigidez para un sistema de ecuaciones del tipo:

    $$\frac{\text{d}{\bf r}}{\text{d}t}=-\Lambda \bf r$$

    ocurrirá si hay una gran disparidad en la magnitud de los autovalores positivos $\lambda_i$ de la matriz $\Lambda$ donde la solución del sistema es

    $${\bf r}(x)=\sum_{i}^N{\alpha_i\text{e}^{-\lambda_it}}{\bf v}_i,$$

Un condensado de **Bose-Einstein** es un estado de la materia que se forma típicamente cuando un gas de bosones a densidades muy bajas se enfría (la llamada **trampa de potencial**) a temperaturas muy cercanas al cero absoluto.

En estas condiciones, y por debajo de una **temperatura crítica $\text{T}_c$**, una gran fracción de estos bosones se encuentran en su estado fundamental (**condensan**), haciendo que fenómenos cuánticos microscopios, tal como la interferencia de su función de onda, se conviertan en **fenómenos macroscópicos**.  

La ecuación de **Gross-Pitaevskii (GP)** es una **ecuación de Schrödinger no lineal** que describe la evolución de funciones de onda macroscópicas, y por tanto, nos permite describir las propiedades de un condensado de Bose-Einstein a temperaturas $T$ muy por debajo de la temperatura crítica de condensación $T_c$.


La ecuación GP incluye: 
<br>

Escribe una función definida por el usuario llamada digamos, step(r,t,H) que toma como argumentos el vector de posición $\vec{r} = (x,y)$ en un tiempo inicial $t$ y una longitud de intervalo $H$, y devuelve el nuevo valor de $\vec{r}$ en el tiempo $t+H$.  

Esta función debe realizar el cálculo de extrapolación de punto medio/Richardson modificado hasta que el cálculo converja a la precisión requerida o se alcance el número máximo $n=8$ de pasos de punto medio modificado.  Si no converge en ocho pasos, haga que su función se llame a sí misma, dos veces, para calcular por separado la solución para la primera y luego la segunda mitad del intervalo de $t$ a $t+H$, algo así:

\begin{align}
r1 =& \text{step}(r,t,H/2)\\
r2 =& \text{step}(r1,t+H/2,H/2),
\end{align}

Entonces $\it{estas}$ funciones pueden llamarse a sí mismas, y así sucesivamente, subdividiendo
el intervalo tantas veces como sea necesario para alcanzar la precisión requerida.

```python
import numpy as np
import matplotlib.pyplot as plt

####
def step(f,r,t,H,xp,yp):
    print('Recursó')
    tp=np.arange(0,2e9,H)
    delta=10/365.25/34/3600
    for t in tp:
        xp.append(r[0])
        yp.append(r[1])
        n=1
        h=H/n
        r1=r+0.5*h*f(r)
        r2=r+h*f(r1)
        R1=np.empty([n,len(r)],float)
        R1[0]=0.5*(r2+r1+0.5*h*f(r2))
        err=2*delta*H
   
        while (err>delta*H) & (n<8):
            n+=1
            h=H/n
            r1=r+0.5*h*f(r)
            r2=r+h*f(r1)
            for i in range(n-1):
                r1+=h*f(r2)
                r2+=h*f(r1)
            R2=np.empty([n,len(r)],float)
            R2[0]=0.5*(r2+r1+0.5*h*f(r2))
            for m in range (1,n):
                error=(R2[m-1]-R1[m-1])/((n/(n-1))**(2*m)-1)
                R2[m]=R2[m-1]+error
            err=abs(error[0])
            R1=np.copy(R2)
        if err<=delta*H:
            r=R2[n-1]
            return r
        else:
            r1=R2[n-1]
            r2=step(f,r1,t+H/2,H/2,xp,yp)
    return r,xp,yp
r=np.array([4e12,0,0,500],float)
def f(r):
    G=6.6743e-11
    M=1.989e30
    x=r[0]
    y=r[1]
    vx=r[2]
    vy=r[3]
    r3=(x**2+y**2)**1.5
    fvx=-G*M*x/r3
    fvy=-G*M*y/r3
    return np.array([vx,vy,fvx,fvy],float)
t=0
H=2e9
xp=[]
yp=[]
r1,xp,yp=step(f,r,t,H,xp,yp)
print(len(xp))
plt.plot(xp,yp,'r.')
```
